Non-Parametric test(also called distributition free test) usually indicates that population data does not have normal distribution which means you cannot make assumptions about the data which you can make when data is normally distributed (like symmetric graph, mean=mode=median, emperical rule etc).

If you don't have a graph availabe to spot parametric vs non-parametric, check the skewness and kurtosis of the distribution. Skewness and kurtosis for a normal distribution is roughly about 0 and 3 respectively (in some cases +/-2 and +/-7 respectively).

Non-parametric test across elementry stats is chi-square test. However, there are many more tests. Ex: In this data we'll be using *Mann Whitney test* which is non-parametric and alternative to the Two sample t-test however not always.

Disadvantages with non-parametric test is that they are time consuming as most of the calculation is done by hand. Also, critical value tables for many tests arenâ€™t included in many computer software packages. So, you need to check them manually.

### We have seen the affect of numerical columns but what about categorical columns?
#### Here we will introduce *Mann Whitney test*

#### It is used to compare differences between two independent groups when the dependent variable is either ordinal or continuous, but not normally distributed. 

### NOTE: Before you run this test make sure to go through the assumptions of Mann Whitney test.

```python
from scipy import stats
```

### Hypothesis:

***Null Hypothesis :*** Women Entrepreneurship Index values show a statistically significant difference between the countries that are develpoed and developing.

***Alternate Hypothesis :*** Women Entrepreneurship Index values do not show a statistically significant difference between the countries that are developed and developing.


```python
# If you try to serve 'Level of development' column directly to test than mann whitlney test won't run.
df_develop = df[df['Level of development'] == 'Developed']
df_developing = df[df['Level of development'] == 'Developing']
```

```python
a = stats.mannwhitneyu(list(df_develop['Women Entrepreneurship Index']), 
                       list(df_developing['Women Entrepreneurship Index']))
```

```python
print('u-statistic for Level of Development ', a[0])
print('p-value: ', a[1])
```

### Rejecting Null hypothesis as p-value is very small which means:

Women Entrepreneurship Index values do not show a statistically significant difference between the countries that are developed and developing.

## Hypothesis:
***Null Hypothesis :*** Women Entrepreneurship Index values show a statistically significant difference between the countries that are members of the European Union and not.

***Alternate Hypothesis :*** Women Entrepreneurship Index values do not show a statistically significant difference between the countries that are members of the European Union and not.


```python
df_mem = df[df['European Union Membership'] == 'Member']
df_non_mem = df[df['European Union Membership'] == 'Not Member']
```

```python
b = stats.mannwhitneyu(list(df_mem['Women Entrepreneurship Index']),
                      list(df_non_mem['Women Entrepreneurship Index']))
```

```python
print('u statistic for European Union Membership : ', b[0])
print('p-value: ', b[1])
```

### Rejecting Null hypothesis as p-value is small.
#### So, Women Entrepreneurship Index values do not show a statistically significant difference between the countries that are members of the European Union and not.
