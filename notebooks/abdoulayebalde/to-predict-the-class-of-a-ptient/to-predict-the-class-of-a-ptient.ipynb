**Context**
This dataset is originally from the National Institute of Diabetes and Digestive and Kidney Diseases. The objective of the dataset is to diagnostically predict whether or not a patient has diabetes, based on certain diagnostic measurements included in the dataset. Several constraints were placed on the selection of these instances from a larger database. In particular, all patients here are females at least 21 years old of Pima Indian heritage.

**Content**
The datasets consists of several medical predictor variables and one target variable, Outcome. Predictor variables includes the number of pregnancies the patient has had, their BMI, insulin level, age, and so on.

<h3> AIM </h3>
To build a machine learning model to accurately predict whether or not the patients in the dataset have diabetes or not?
In this question I am using DecisionTree and KNN algorithm because it is a classification problem

```python
df = pd.read_csv("/kaggle/input/pima-indians-diabetes-database/diabetes.csv")
df.head()
```

## importing libraries 

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

# Feature selection

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
X[:5] , y[:5]
```

```python
from sklearn.preprocessing import 
```

# spliting the data into trainset and test set 

```python
X_train,X_test,y_train,y_test = train_test_split(X.values,y.values, test_size = .2, random_state = 0)
```

```python
# let have a look at the size of training and testing data
print("Train size:", X_train.shape ,y_train.shape)
print('Test size :',X_test.shape,y_test.shape)
```

```python
# let do something funny first by visualizing the training  data as scatter plot
plt.figure(figsize=(15,7))
plt.scatter(X_train[:,0], X_train[:,1], c = y_train)
plt.figure(figsize=(15,7))
plt.scatter(X_train[:,2],X_train[:,3], c=y_train)
```

# now training the model

```python
tree = DecisionTreeClassifier(criterion="entropy", max_depth = 4)
tree.fit(X_train,y_train)
```

```python
# now let predict y
y_pred = tree.predict(X_test)
y_pred[:10]
```

# Evaluation of the model

```python
print("The accuracy score is :",accuracy_score(y_test,y_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
