# Analysis & visualisation

```python
df['continent_name'].value_counts()
```

```python
fig = px.box(df, x='continent_name',y='median_salary',width=1100, height=800, title="Median salaries across different continents")

fig.show()
```

```python
fig = px.box(df, x='continent_name',y='average_salary',width=1100, height=800, title="Average salaries across different continents")

fig.show()
```

```python
fig = px.box(df, x='continent_name',y='lowest_salary',width=1100, height=800, title="Lowest salaries across different continents")

fig.show()
```

```python
fig = px.box(df, x='continent_name',y='highest_salary',width=1100, height=800, title="Highest salaries across different continents")

fig.show()
```

```python
fig = px.bar(df, x='country_name',y='median_salary',width=1100, height=800, title="Median salaries across different countrys")

fig.show()
```

```python
top20_highest_salary_country=df.sort_values(by='highest_salary', ascending = False).head(20)
fig = px.bar(top20_highest_salary_country, x='country_name',y='highest_salary',width=1100, height=800, title="Top 20 Countrys with Hightest salaries")

fig.show()
```

```python
top20_median_salary_country=df.sort_values(by='median_salary', ascending = False).head(20)
fig = px.bar(top20_median_salary_country, x='country_name',y='median_salary',width=1100, height=800, title="Top 20 Countrys With Median salaries")

fig.show()
```

```python
top20_avwerage_salary_country=df.sort_values(by='average_salary', ascending = False).head(20)
fig = px.bar(top20_avwerage_salary_country, x='country_name',y='average_salary',width=1100, height=800, title="Top 20 Countrys With Average salaries")

fig.show()
```

```python
data=df.groupby('continent_name')['median_salary'].mean().reset_index()
data=data.sort_values(by='median_salary',ascending=False)
fig=px.bar(data,x='continent_name',y='median_salary',title='The Madian Salary in Each Continent ')
fig.show()
```

```python
data=df.groupby('continent_name')['average_salary'].mean().reset_index()
data=data.sort_values(by='average_salary',ascending=False)
fig=px.bar(data,x='continent_name',y='average_salary',title='The Average Salary in Each Continent ')
fig.show()
```

```python
data=df.groupby('continent_name')['highest_salary'].mean().reset_index()
data=data.sort_values(by='highest_salary',ascending=False)
fig=px.bar(data,x='continent_name',y='highest_salary',title='The Highest Salary in Each Continent ')
fig.show()
```

```python
# contres in europe and averag salary
country= df[df['continent_name']=='Europe']
fig = px.bar(country, x='country_name',y='average_salary',width=1100, height=800, title="Average salaries across Europe")

fig.show()
```

```python
# contres in Nouth Amrica and averag salary
country= df[df['continent_name']=='North America']
fig = px.bar(country, x='country_name',y='average_salary',width=1100, height=800, title="Average salaries across North America")

fig.show()
```
