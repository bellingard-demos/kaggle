# Analysis

```python
types=data['type'].value_counts()
types.reset_index()
```

```python
data.groupby('type')['country'].value_counts().reset_index()
```

```python
directors=data.groupby(['type','director'])['type'].value_counts().reset_index()
directors
```

```python
top_movie_10=directors[(directors['type']=='Movie')&(directors['director']!='Not Given')].sort_values(by='count',ascending=False).head(10)
top_movie_10
```

```python
top_TV_10=directors[(directors['type']=='TV Show')&(directors['director']!='Not Given')].sort_values(by='count',ascending=False).head(10)
top_TV_10
```

```python
data['country'].nunique()
```

```python
data['country'].unique()
```

```python
countrys=data.groupby(['type'])['country'].value_counts().reset_index()
countrys
```

```python
top10_country=countrys.sort_values(by='count',ascending=False).head(10)
top10_country
```

```python
data.head(10)
```

```python
data['release_year'].unique()
```

```python
data['release_year'].nunique()
```

```python
release_years=data.groupby('release_year')['release_year'].value_counts().reset_index()
release_years
```

```python
top10_years=release_years.sort_values(by='count',ascending=False).head(10)
top10_years
```

```python
top10_years=top10_years.sort_values(by='release_year')
top10_years
```

```python
years_director=data.groupby(['release_year','director'])['director'].value_counts().reset_index()
years_director
```

```python
top10_director_years=years_director.sort_values(by='count',ascending=False).head(0)
top10_director_years
```

```python
data['duration'].nunique()
```

```python
data['duration'].value_counts().head(20).reset_index()
```

```python
durations=data.groupby(['duration','type'])['duration'].value_counts().reset_index()
durations
```

```python
top5_duration=durations.sort_values(by='count',ascending=False).head(5)
top5_duration
```

```python
data['rating'].nunique()
```

```python
data['rating'].unique()
```

```python
rat=data.groupby(['type','rating'])['type'].value_counts().reset_index()
top10_rat=rat.sort_values(by='count',ascending=False).head(10)
top10_rat
```

```python
data['listed_in'].nunique()
```

```python
list_grop=data.groupby(['listed_in','type'])['type'].value_counts().reset_index()
list_grop
```

```python
top10_list=list_grop.sort_values(by='count',ascending=False).head(10)
top10_list
```

# visulaisation

```python
types
```

```python
types.plot(kind='pie',figsize=(15, 6),autopct='%1.1f%%',shadow=True,labels=['Movie','TV Show'],explode=[0,0.2] )
plt.title('Distribution of Shows Types')
plt.axis('equal') 
```

```python
fig = plt.figure() 

ax0 = fig.add_subplot(1, 2, 1) 
ax1 = fig.add_subplot(1, 2, 2) 
# Subplot 1: 
top_movie_10.plot(kind='bar',x='director',y='count',color='blue', figsize=(20, 6), ax=ax0) 
ax0.set_title('The Top 10 Movies Directors After Not Given')
ax0.set_xlabel('Movies Directors')
ax0.set_ylabel('Movies Counts')

# Subplot 2: 
top_TV_10.plot(kind='bar',x='director',y='count',color='g', figsize=(20, 6), ax=ax1) 
ax1.set_title('The Top 10 TV Show Directors After Not Given ')
ax1.set_xlabel('Movies TV Show')
ax1.set_ylabel('TV Show Counts')

plt.show()
```

```python
sns.barplot(data=top10_country,x='country',y='count',hue='type',alpha=0.8,width=0.6)
plt.ylabel('Count')
plt.xlabel('Contries')
plt.title('Top 10 Countries')
plt.xticks(rotation=90)
```

```python
top10_years.plot(x='release_year',y='count')
plt.ylabel('Count')
plt.xlabel('Year')
plt.title('Top 10 Release Year')
```

```python
top5=top5_duration['duration']
top5_duration.plot(kind='pie',y='count',labels=top5,figsize=(15, 6),autopct='%1.1f%%',shadow=False,explode=[0,0.2,0,0,0]) 
plt.legend(top5_duration['duration'])
plt.title('Top 5 Durations')
plt.axis('equal') 
```

```python
top10_list
```

```python
top10_list.plot(kind='bar',x='listed_in',y='count',color='b', figsize=(20, 8),alpha=0.7)
plt.ylabel('Count')
plt.xlabel('Liste Name')
plt.title('Top 10 listes')
plt.xticks(rotation=90)
```

```python
sns.barplot(data=top10_rat,x='rating',y='count',hue='type',alpha=0.8,width=0.6)
plt.ylabel('Count')
plt.xlabel('Rating')
plt.title('Top 10 Rating And Type')
plt.legend(loc='upper right') 
```
