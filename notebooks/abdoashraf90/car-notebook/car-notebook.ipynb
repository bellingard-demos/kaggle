# Analysis

```python
df['mpg'].nunique()
```

```python
df['mpg'].value_counts()
```

```python
df['mpg'].max()
```

```python
df['cylinders'].nunique()
```

```python
number_of_cylinders=df['cylinders'].value_counts()
number_of_cylinders
```

```python
df['horsepower'].nunique()
```

```python
top10_horsepower= df['horsepower'].value_counts().head(10).reset_index()
top10_horsepower=top10_horsepower.sort_values(by='count')
top10_horsepower
```

```python
df['origin'].nunique()
```

```python
num_of_origin=df['origin'].value_counts()
num_of_origin
```

```python
df['model_year'].nunique()
```

```python
years=df['model_year'].value_counts().reset_index()
years
```

```python
df.head()
```

```python
cylinder_num=df.groupby(['cylinders','model_year'])['model_year'].value_counts().reset_index()
cylinder_num=cylinder_num.sort_values(by='count',ascending=False).head(10)
cylinder_num
```

```python
df['displacement'].nunique()
```

```python
top10_displacement=df['displacement'].value_counts().reset_index().head(10)
top10_displacement
```

```python
sns.scatterplot(data=df,x='weight',y='acceleration')
plt.title('weight vs acceleration')
plt.xlabel('weight of the car')
plt.ylabel('acceleration of the car')
plt.show()
```

# visualisation

```python
number_of_cylinders.plot(kind='pie',figsize=(15, 6),autopct='%1.1f%%',shadow=False,labels=['4','8','6','3','5'],explode=[0.,0,0,0.5,0.3] )
plt.title('Average of Types Of Cylinders ')
plt.axis('equal') 
```

```python
sns.barplot(data=top10_displacement,x='displacement',y='count',alpha=0.8,width=0.6)
plt.ylabel('Count')
plt.xlabel('Displacement')
plt.title('Top 10  Displacement And Their Count ')
plt.xticks(rotation=90)
```

```python
sns.barplot(data=top10_horsepower,x='horsepower',y='count',alpha=0.8,width=0.6)
plt.ylabel('Count')
plt.xlabel('horsepower')
plt.title('Top 10 horsepower')
plt.xticks(rotation=90)
```

```python
num_of_origin.plot(kind='pie',figsize=(15, 6),autopct='%1.1f%%',shadow=False,labels=['Usa','Japan','Europe'],explode=[0,0,0.1] )
plt.title('Average of Types Of Cylinders ')
plt.axis('equal')
```

```python
sns.barplot(data=years,x='model_year',y='count',alpha=0.9)
plt.ylabel('Count')
plt.xlabel('model_year')
plt.title('Top Model Year and the Number of Cars')
```
