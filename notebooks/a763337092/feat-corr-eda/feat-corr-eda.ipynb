```python
train = pd.read_csv('/kaggle/input/jane-street-market-prediction/train.csv')
train.head()
```

```python
len(train)
```

```python
feat_cols = [f'feature_{i}' for i in range(130)]
```

```python
col_nan_dict = {}
for col in feat_cols:
    nan_num = len(train[train[col].isna()])
    col_nan_dict[col] = nan_num
    print(f'{col:11}: {nan_num:7}')
```

```python
col_nan_df = pd.DataFrame()
col_nan_df['col'] = col_nan_dict.keys()
col_nan_df['nan'] = col_nan_dict.values()
col_nan_df.head()
```

```python
same_nan_df = col_nan_df.groupby('nan')['col'].apply(list).reset_index()
same_nan_df
```

```python
for _nan, _col in same_nan_df.values:
    print(f'{_nan:6}:', _col)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def make_plots(Vs):
    col = 4
    row = len(Vs)//4+1
    plt.figure(figsize=(20,row*5))
    idx = train[~train[Vs[0]].isna()].index
    for i,v in enumerate(Vs):
        plt.subplot(row,col,i+1)
        n = train[v].nunique()
        x = np.sum(train.loc[idx,v]!=train.loc[idx,v].astype(int))
        y = np.round(100*np.sum(train[v].isna())/len(train),2)
        t = 'int'
        if x!=0: t = 'float'
        plt.title(v+' has '+str(n)+' '+t+' and '+str(y)+'% nan')
        plt.yticks([])
        h = plt.hist(train.loc[idx,v],bins=100)
        if len(h[0])>1: plt.ylim((0,np.sort(h[0])[-2]))
    plt.show()
Vs = ['feature_0', 'feature_1', 'feature_2', 'feature_5', 'feature_6', 'feature_37', 'feature_38', 'feature_39', 'feature_40', 'feature_41', 'feature_42', 'feature_43', 'feature_46', 'feature_47', 'feature_48', 'feature_49', 'feature_50', 'feature_51', 'feature_52', 'feature_53', 'feature_54', 'feature_57', 'feature_60', 'feature_61', 'feature_62', 'feature_63', 'feature_64', 'feature_65', 'feature_66', 'feature_67', 'feature_68', 'feature_69', 'feature_70', 'feature_71', 'feature_77', 'feature_83', 'feature_89', 'feature_95', 'feature_101', 'feature_107', 'feature_113', 'feature_119']
make_plots(Vs)
```

```python
def make_corr(Vs,Vtitle=''):
    cols = Vs
    plt.figure(figsize=(15,15))
    sns.heatmap(train[cols].corr(), cmap='RdBu_r', annot=True, center=0.0)
    if Vtitle!='': plt.title(Vtitle,fontsize=14)
    else: plt.title(Vs[0]+' - '+Vs[-1],fontsize=14)
    plt.show()
Vtitle = 'nan0'
make_corr(Vs,Vtitle)
```
