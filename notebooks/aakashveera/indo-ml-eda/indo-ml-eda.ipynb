```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from tqdm import tqdm
from PIL import Image
import random
import cv2
```

```python
df = pd.read_csv("../input/datathonindoml-2022/train_labels.csv")
columns = 4
rows = 1

class_dictionary = {
    0:'memo',
    1:'form',
    2:'email',
    3:'handwritten',
    4:'advertisement',
    5:'scientific_report',
    6:'scientific_publication',
    7:'specification',
    8:'file_folder',
    9:'news_article',
    10:'budget',
    11:'invoice',
    12:'presentation',
    13:'questionaire',
    14:'resume',
    15:'letter'
}
```

```python
def display_samples(image_id):
    file_ids = df[df['label']==image_id].sample(10)['id'].values

    fig=plt.figure(figsize=(35, 45))

    for index in range(columns*rows):
        img = cv2.imread("/kaggle/input/datathonindoml-2022/train/train/"+str(file_ids[index])+'.tif', cv2.IMREAD_GRAYSCALE)
        fig.add_subplot(rows, columns, index+1,label='this is a label')
        plt.imshow(img,cmap='gray')
        plt.axis('off');
```

```python
print(class_dictionary[0].upper(),end='\n')
display_samples(0)
```
