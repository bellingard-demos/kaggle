```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

# Simple preprocessing

```python
df = pd.read_csv("/kaggle/input/iris/Iris.csv", index_col=0)
df.head()
```

```python

label_encoder = LabelEncoder()

df['Species'] = label_encoder.fit_transform(df['Species'])
label_mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
df.head()
```

```python
label_mapping
```

```python
X = df.drop(columns="Species")
y = df["Species"]
```

#### To Ensure fairness of the test, both models will be trained and evaluated on same data

```python

# Assuming X contains the feature data and y contains the target labels
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
model_1 = DecisionTreeClassifier(random_state=42)

# Train the model on the training data
model_1.fit(X_train, y_train)

# Model 2: Decision tree with different parameters
model_2 = DecisionTreeClassifier(max_depth=3, min_samples_split=2, random_state=42)

# Train the model on the training data
model_2.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

# Predict the target labels using both models
y_pred_model1 = model_1.predict(X_test)
y_pred_model2 = model_2.predict(X_test)

# Calculate accuracy scores for both models
accuracy_model1 = accuracy_score(y_test, y_pred_model1)
accuracy_model2 = accuracy_score(y_test, y_pred_model2)

print("Accuracy for Model 1:", accuracy_model1)
print("Accuracy for Model 2:", accuracy_model2)
```

```python
import statsmodels.api as sm

# Number of correct predictions for each model
correct_predictions_model1 = accuracy_model1 * len(y_test)
correct_predictions_model2 = accuracy_model2 * len(y_test)

# Number of total predictions for each model
total_predictions_model1 = len(y_test)
total_predictions_model2 = len(y_test)

# Perform the two-sample proportions z-test
z_stat, p_value = sm.stats.proportions_ztest([correct_predictions_model1, correct_predictions_model2],
                                             [total_predictions_model1, total_predictions_model2])

print("Z-statistic:", z_stat)
print("P-value:", p_value)
```
