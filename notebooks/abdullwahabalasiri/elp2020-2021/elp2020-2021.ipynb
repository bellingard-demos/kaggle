```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
puk_df =pd.read_csv('../input/english-premier-league202021/EPL_20_21.csv')
```

```python
puk_df['MinsPerMatch'] = (puk_df['Mins'] / puk_df['Matches']).astype(int)
puk_df['GoalsPerMatch'] = (puk_df['Goals'] / puk_df['Matches']).astype(float)
```

```python
puk_df.head().round(2)
```

```python
total_goals = puk_df['Goals'].sum()
total_goals
```

```python
total_penaltygoals = puk_df['Penalty_Goals'].sum()
total_penaltygoals
```

```python
total_penaltyattemps= puk_df['Penalty_Attempted'].sum()
total_penaltyattemps
```

```python
plt.figure(figsize=(13,6))
pl_not_scored = puk_df['Penalty_Attempted'].sum() - total_penaltygoals
data = [pl_not_scored, total_penaltygoals]
labels = ['Penalties missed', 'Penalties Scored']
colors = sns.color_palette('Paired')
plt.pie(data,labels = labels , colors= colors, autopct = '%.0f%%')
plt.show()
```

```python
under20 = puk_df[puk_df['Age'] <= 20 ]
age25_20 = puk_df[ (puk_df['Age'] > 20) & (puk_df['Age'] <= 25) ]
age25_30 = puk_df[ (puk_df['Age'] > 25) & (puk_df['Age'] <= 30) ]
Above30 = puk_df[puk_df['Age'] > 30 ]
```

```python
x = np.array([under20['Name'].count(), age25_20['Name'].count(), age25_30['Name'].count(),  Above30['Name'].count()])
mylabels = ["<=20", ">20 & <=25", ">25 & <=30", ">30" ]
plt.title('Total Players With Age Groups', fontsize = 20)
plt.pie(x, labels = mylabels, autopct="%.1f%%")
plt.show()
```
