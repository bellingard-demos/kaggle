```python
teum = teum.drop(columns=['Country Code', 'Indicator Name','Indicator Code'])
```

```python
df = teum.melt(id_vars=['Country Name'], var_name='Year', value_name='Value')
df['Year'] = df['Year'].astype(str)

fig = px.line(df, x='Year', y='Value', color='Country Name', title= 'Mortality Rate by Country', height=1000)
fig.show()
```

```python
curious_cases = teum[teum['Country Name'].isin(['Rwanda','Sudan','Haiti', 'Bangladesh'])]
new_df = curious_cases.melt(id_vars=['Country Name'], var_name='Year', value_name='Value')
new_df['Year'] = new_df['Year'].astype(str)

fig = px.line(new_df, x='Year', y='Value', color='Country Name', title='Child Mortality Rates by Year for RWA, SDN, BGD and HTI', height=1000)

fig.show()
```

RWA - BOSTON (April 4, 2014)—The 1994 genocide in Rwanda resulted in the deaths of up to one million people and the displacement of millions more. In the years that followed, Rwanda was the poorest country in the world, with the highest child mortality and lowest life expectancy at birth anywhere. Less than one in four children were fully vaccinated. Many health workers had either been killed or fled the country, and health facilities across the country had been destroyed.

HTI - On January 12, 2010, an earthquake of magnitude 7.0 struck Haiti near its capital. The earthquake killed a total of 222,600 people and displaced 1.5 million people (IDP). Poor physical infrastructure in Haiti has been cited as a primary cause of the excess number of deaths from injuries in the disaster’s immediate aftermath . In the medium to long-term aftermath, however, poor public health infrastructure has been cited as a primary cause of excess mortality, exacerbated by weak governance of the Haitian state

https://equityhealthj.biomedcentral.com/articles/10.1186/s12939-016-0403-z#Sec1

SDN - The increase in child mortality in Sudan around 1984 can be associated with the beginning of the Second Sudanese Civil War in 1983​​. This conflict would have led to widespread disruption of healthcare services, food supplies, and general living conditions, significantly impacting child mortality rates.

BGD - The spike in the child mortality rate in Bangladesh in 1971 can be attributed to the Bangladesh Liberation War, which took place that year. The conflict led to widespread disruption, destruction of healthcare infrastructure, displacement of populations, and resulted in a humanitarian crisis, all of which would have drastically increased mortality rates, especially among vulnerable populations such as children

```python
teum['Mortality Rate'] = teum[2021.0]
merged_df = pd.merge(teum[['Country Name', 'Mortality Rate']], tedois, left_on='Country Name', right_on='TableName')

income_group_rates = merged_df.groupby('IncomeGroup')['Mortality Rate'].mean().reset_index()

fig = px.bar(income_group_rates, x='IncomeGroup', y='Mortality Rate', title='Average Child Mortality Rate by Income Group')
fig.show()
```
