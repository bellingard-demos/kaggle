## Model Building

```python
# creating a new dataframe(df01) to do modelling on it
# our aim is to keep it original dataset untouched
df01=df[['Gender', 'Age', 'Annual Income (k$)','Spending Score (1-100)']]
```

```python
df01.head()
```

```python
from sklearn import preprocessing
```

```python
label_encoder=preprocessing.LabelEncoder()
```

```python
df01.Gender= label_encoder.fit_transform(df01.Gender)
```

```python
df01.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
scaler=MinMaxScaler()
```

```python
df01.columns
```

```python
var= ['Gender', 'Age', 'Annual Income (k$)', 'Spending Score (1-100)']
```

```python
df01[var]=scaler.fit_transform(df01[var])
```

```python
df01.head()
```

```python
#finding optimal no of clusters using silhouette score
range_n_clusters = [2, 3, 4, 5, 6, 7]
for num_clusters in range_n_clusters:
    
    # intialise kmeans
    kmeans = KMeans(n_clusters=num_clusters, max_iter=50)
    kmeans.fit(df01)
    
    cluster_labels = kmeans.labels_
    
    # silhouette score
    silhouette_avg = silhouette_score(df01, cluster_labels)
    print("For n_clusters={0}, the silhouette score is {1}".format(num_clusters, silhouette_avg))
```

```python
# we will go with n_clusters=3
kmeans=KMeans(n_clusters=3, max_iter=50)
kmeans.fit(df01)
```

```python
df['cluster']= kmeans.labels_
```

```python
df.head()
```
