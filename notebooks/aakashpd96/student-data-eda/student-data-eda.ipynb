```python
import pandas as pd, numpy as np
import matplotlib.pyplot as plt, seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df= pd.read_csv('../input/students-performance-in-exams/StudentsPerformance.csv')
df.head()
```

```python
df.info()
```

```python
df['total_score']=df['math score']+df['reading score']+df['writing score']
```

```python
df.describe()
```


# EDA on Dataset

```python
df.groupby('gender')['total_score'].mean().plot(kind='bar')
```

```python
# comparing maths score among genders
df.groupby('gender')['math score'].mean().plot(kind='bar')
```

```python
# comparing reading score among genders
df.groupby('gender')['reading score'].mean().plot(kind='bar')
```

```python
# comparing writing score among genders
df.groupby('gender')['writing score'].mean().plot(kind='bar')
```

```python
df['peducation']=df['parental level of education']
df.drop('parental level of education', axis=1, inplace=True)
```

```python
df.groupby('test preparation course')['total_score'].mean().plot(kind='bar')
```

```python
df.groupby('peducation')['total_score'].mean().plot(kind='bar')
```

```python
# gender
df=df.replace({'male':1,'female':0})
```

```python
#race/ethinic group
df= df.replace({'group A':1, 'group B':2, 'group C':3, 'group D':4, 'group E':5})
```

```python
#lunch
df= df.replace({'free/reduced':0, 'standard':1})
```

```python
#test preparation course
df= df.replace({'none':0, 'completed':1})
```

```python
#peducation
df= df.replace({'some college':1, "associate's degree":2, 'high school':3, 'some high school':4, "bachelor's degree":5, "master's degree":6})
```
