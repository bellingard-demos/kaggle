## Content
The dataset contains a set of 150 records under 5 attributes - Petal Length, Petal Width, Sepal Length, Sepal width and Class(Species).

```python
import numpy as np, pandas as pd
```

```python
iris= pd.read_csv('../input/iris-flower-dataset/IRIS.csv')
iris.head()
```

```python
y= iris.pop('species')
```

```python
X= iris
```

### Model Building using Logistic Regression

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, test_size=0.3, random_state=100)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression(random_state=0, multi_class='ovr')
```

```python
model= clf.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```

94.3% accuracy on the train DataSet

```python
model.score(X_test,y_test)
```

We got 95.5% accuracy on the Test Dataset

```python
model.predict(X_train)
```

#### In order to get the **species of iris flower** use **model.predict(your_dataset)** and it will give you the species with an accuracy of around 95.5%
