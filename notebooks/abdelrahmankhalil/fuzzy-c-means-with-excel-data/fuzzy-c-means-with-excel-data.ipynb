# ****Scaling Data

```python
scalarModel = StandardScaler()
#X_train = scalarModel.fit_transform(X_train)
#X_test = scalarModel.fit_transform(X_test)
data = scalarModel.fit_transform(data)
```

```python
data = pd.DataFrame(data , columns = columns )
data
```

```python
data = data.values
data
```

# ****Apply the algorithm Fuzzy c Means

```python
fcmModel = FCM(n_clusters = 4)
fcmModel.fit(data)
center = fcmModel.centers
center
```

```python
#Calculating Prediction
pred = fcmModel.predict(data)
print('Predicted Value for fcmModel is : ' , pred)
pred.shape
```

# ****Draw the Data and centers

```python
# Visualising the clusters
plt.scatter(data[pred == 0, 0], data[pred == 0, 1], s = 10, c = 'r')
plt.scatter(data[pred == 1, 0], data[pred == 1, 1], s = 10, c = 'b' )
plt.scatter(data[pred == 2, 0], data[pred == 2, 1], s = 10, c = 'g' )
plt.scatter(data[pred == 3, 0], data[pred == 3, 1], s = 10, c = 'y' )
plt.scatter(center[:, 0], center[:, 1], s = 300, c = 'black',marker = '+')
plt.title('Clusters of customers')
plt.xlabel('X_label ')
plt.ylabel('y_label ')
plt.legend()
plt.show()
```
