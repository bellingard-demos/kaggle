# ****Delete empty rows

train_data.dropna(subset = ["Age"], inplace=True)

##delete empty rows from test_data
test_data.dropna(subset = ["Age"], inplace=True)
print(test_data.shape)

# ****Convert data from string to numbers

```python
oe=OrdinalEncoder()
##convert train_data
for col in train_data:
    train_data[col]=oe.fit_transform(np.asarray(train_data[col].astype('str')).reshape(-1,1))
##.........................................................................
##convert test_data
for col in test_data:
    test_data[col]=oe.fit_transform(np.asarray(test_data[col].astype('str')).reshape(-1,1))
```

```python
print(train_data.shape)
train_data.head()
```

```python
print(test_data.shape)
test_data.head()
```

# ****Spliting data to "X" and "y"

```python
y = train_data['Survived']
X = train_data.drop(columns = 'Survived')
print(X.shape)
print(y.shape)
```

# ****Scaling Data

```python
## Scaling Model 
## scale to train_data
scaler = StandardScaler()
X = scaler.fit_transform(X)
print(X.shape)
```

# ****Spliting data to X_train, y_train,X_test and y_test

```python
## spliting data into X_train , y_train,X_test,y_test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=44, shuffle =True)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

# ****Apply the RandomForestClassfier Algorithm

```python
RandomForestClassifierModel = RandomForestClassifier(n_estimators=100,max_depth=9, random_state=44,min_samples_split=10,min_samples_leaf=5
                                                    ,n_jobs = -1) 
RandomForestClassifierModel.fit(X_train,y_train)

```

```python
#Calculating Details
print('Random Forest Classifier Train Score is : ' , RandomForestClassifierModel.score(X_train, y_train))
print('Random Forest Classifier Test Score is : ' , RandomForestClassifierModel.score(X_test, y_test))
```

```python
#Calculating Prediction
y_pred =  RandomForestClassifierModel.predict(X_test)

print('Predicted Value for RandomForestClassifierModelis : ' , y_pred)

print(y_pred.shape)
```

```python
MAEValue = mean_absolute_error(y_test, y_pred, multioutput='uniform_average')
print("MAEValue is :" ,MAEValue)
```

```python
MSEValue= mean_squared_error(y_test, y_pred, multioutput='raw_values')
print("MSEValue is : " , MSEValue)
```

```python
confusionMatrix = confusion_matrix(y_test,y_pred)
print('confusionMatrix is : ',confusionMatrix)
# drawing confusion matrix
sns.heatmap(confusionMatrix, center = True)
plt.show()
```

```python
AccuracyScore = accuracy_score(y_test,y_pred)
print('AccuracyScore is : ',AccuracyScore)
```

```python
FScore = f1_score(y_test,y_pred)
print('FScore is ',FScore)
```
