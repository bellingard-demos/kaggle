# ****    Import library 

```python
import numpy as np 
import pandas as pd
import matplotlib .pyplot as plt 
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from fcmeans import FCM
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

# ****Read Data

```python
data = pd.read_csv('/kaggle/input/audit-data/audit_data.csv')
```

```python
data.shape
```

```python
data.head()
```

# ****Convert Data from Supervised to Unsupervised 

```python
## Delete the Output from data 
data = data.drop('Risk',axis = 1)
data.shape
```

```python
## delete the column containg the null data

data = data.drop('Money_Value',axis = 1)
data.shape
```

```python
data.isnull().sum()
```

# **** Convert string data to numerical number

```python

le = LabelEncoder()
le.fit(data['LOCATION_ID'])
le.transform(data['LOCATION_ID'])
data['LOCATION_ID'] = le.transform(data['LOCATION_ID'])
```

```python
columns = data.columns
columns
```

# ****Correlation

```python
plt.figure(figsize=(15,10))
sns.heatmap(data.corr( ),annot=True)
```

# ****Apply the Fuzzy C Means Algorithm

```python
data = data.values
```

```python
fcmModel = FCM(n_clusters = 7)
fcmModel.fit(data)
center = fcmModel.centers
center
```

# ****Prediction

```python
#Calculating Prediction
pred = fcmModel.predict(data)
print('Predicted Value for fcmModel is : ' , pred)
pred.shape
```

```python
data = pd.DataFrame(data , columns = columns )
data
```

```python
# add the cluster column to the dataframe 
data['cluster'] = pred
data.head(20)
```

# ****Draw The Data and Center

```python
data = data.values
```

```python
# Visualising the clusters
# Draw the data
plt.scatter(data[pred == 0, 0], data[pred == 0, 1], s = 10, c = 'r')
plt.scatter(data[pred == 1, 0], data[pred == 1, 1], s = 10, c = 'b' )
plt.scatter(data[pred == 2, 0], data[pred == 2, 1], s = 10, c = 'g' )
plt.scatter(data[pred == 3, 0], data[pred == 3, 1], s = 10, c = 'pink' )
plt.scatter(data[pred == 4, 0], data[pred == 4, 1], s = 10, c = 'aqua' )
plt.scatter(data[pred == 5, 0], data[pred == 5, 1], s = 10, c = 'purple' )
plt.scatter(data[pred == 6, 0], data[pred == 6, 1], s = 10, c = 'y' )
#plt.scatter(data[pred == 7, 0], data[pred == 7, 1], s = 10, c = 'grey' )
# Draw the center
plt.scatter(center[:, 0], center[:, 1], s = 500, c = 'black',marker = '+')
plt.title('Clusters ')
plt.xlabel('X_label ')
plt.ylabel('y_label ')
plt.legend()
plt.show()
```
