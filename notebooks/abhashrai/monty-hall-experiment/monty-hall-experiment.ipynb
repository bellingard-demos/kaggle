```python
import pandas as pd
import random

pd.reset_option('display.max_columns')
```

Suppose:

    DoorA --> index:1, prize:'goat'
    DoorB --> index:2, prize:'car'
    DoorC --> index:3, prize:'goat'
    
**Note: The doors and prizes behind each door in this experiment remain the same. Only player choices is random!**

```python
# set number of epochs and players to your liking
no_of_epochs = 50
no_of_players = 10000
```

```python
total_swap_wins = []
total_swap_losses = []
total_no_swap_wins = []
total_no_swap_losses = []

total_when_swapped_total_plays = []
total_when_not_swapped_total_plays = []



for a in range(0, no_of_epochs):
    swap_wins = 0
    swap_losses = 0
    no_swap_wins = 0
    no_swap_losses = 0

    when_swapped_total_plays = 0
    when_not_swapped_total_plays = 0

    for i in range(0, no_of_players):
        random_door_choice = random.randint(1, 3)
        
        if random_door_choice == 1:
        # Here Monty reveals door C and asks if player wants to swap

            # Case 1: Player switches door (Player gets the car and wins)
            swap_wins += 1
            when_swapped_total_plays += 1

            # Case 2: Player doesn't switch door (Player gets the goat and looses)
            no_swap_losses += 1
            when_not_swapped_total_plays += 1

        elif random_door_choice == 2:
        # Here Monty reveals door A and asks if player wants to swap

            # Case 1: Player switches door (Player gets the goat and looses)
            swap_losses += 1
            when_swapped_total_plays += 1

            # Case 2: Player doesn't switch door (Player gets the car and wins)
            no_swap_wins += 1
            when_not_swapped_total_plays += 1

        elif random_door_choice == 3:
        # Here Monty reveals door A and asks if player wants to swap

            # Case 1: Player switches door (Player gets the car and wins)
            swap_wins += 1
            when_swapped_total_plays += 1

            # Case 2: Player doesn't switch door (Player gets the goat and looses)
            no_swap_losses += 1
            when_not_swapped_total_plays += 1
    
    total_swap_wins.append(swap_wins)
    total_swap_losses.append(swap_losses)
    total_no_swap_wins.append(no_swap_wins)
    total_no_swap_losses.append(no_swap_losses)
    total_when_swapped_total_plays.append(when_swapped_total_plays)
    total_when_not_swapped_total_plays.append(when_not_swapped_total_plays)
```

```python
df = pd.DataFrame({
    'Wins (Switched)': total_swap_wins, 
    'Losses (Switched)': total_swap_losses,
    'Total plays (Switched)': total_when_swapped_total_plays,
    'Wins (Not Switched)': total_no_swap_wins, 
    'Losses (Not Switched)': total_no_swap_losses, 
    'Total plays (Not Switched)': total_when_not_swapped_total_plays,
})

df
```
