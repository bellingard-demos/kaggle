```python
import pandas as pd
import re

def extract_number_before_keyword(text, keyword):
    pattern = rf'(\d+)\s+{re.escape(keyword)}'
    match = re.search(pattern, text)
    if match:
        return match.group(1)
    else:
        return None

# Read the Excel file into a Pandas DataFrame
file_path = '/kaggle/input/zewwww/New Microsoft Excel Worksheet.xlsx'
data = pd.read_excel(file_path)

# Keywords to find
keywords = ["P-PALLET", "PALLET", "C-BOX"]

# Initialize columns for each keyword with NaN
for keyword in keywords:
    data[keyword] = float('NaN')

# Select the column containing the text
text_column = data['ALL_PALLETS_INFO']  # Replace with your actual column name

# Extract the number before each keyword for each text in the column
for keyword in keywords:
    numbers = []
    for t in text_column:
        number = extract_number_before_keyword(str(t), keyword)  # Ensure the text is converted to string
        numbers.append(number)
    data[keyword] = numbers



data.head(50)
```

```python
# Define the path to save the CSV file in the writable directory (/kaggle/working/)
csv_file_path = '/kaggle/working/data.csv'

# Save the DataFrame to a CSV file in the writable directory
data.to_csv(csv_file_path, index=False)
```
