# As we can see from the plot above, linear regression will not be the best fit. so, we will try using Polynomial Regression


```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
#Sorting the X_train and Y_train in ascending order for no error in ploting the curve
df.sort_values('Temperature (°C)',inplace=True)
```

```python
#split the data
x = df.iloc[:,:-1].values
y = df.iloc[:,1:].values
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=20)
```

```python
# after etration the 4th degree is the best
poly = PolynomialFeatures(degree=4)
x_poly=poly.fit_transform(X_train)
```

```python
Regression=LinearRegression()
Regression.fit(x_poly,y_train)
```

```python
plt.scatter(X_train, y_train)

# Predict using the model and the transformed training features
y_pred_train = Regression.predict(x_poly)

# Sort X_train and y_pred_train based on X_train for a smoother line plot
sorted_indices = X_train.ravel().argsort()
X_train_sorted = X_train[sorted_indices]
y_pred_train_sorted = y_pred_train[sorted_indices]

plt.plot(X_train_sorted, y_pred_train_sorted, color='r')
plt.title("Sales of the Ice Cream at Different Temperatures")
plt.xlabel('Temperature (°C)')
plt.ylabel('Ice Cream Sales (units)')
plt.text(-1.5, 35, 'Training validation', style='italic')
plt.show()
```

```python
plt.scatter(X_test,y_test)

# Create polynomial features for X_test
xpoly_test = poly.transform(X_test)
y_pred_test = Regression.predict(xpoly_test)

# Sort X_test and y_pred_test based on X_test for a smoother line plot
sort_indices = X_test.ravel().argsort()
X_test_sorted = X_test[sort_indices]
y_pred_test_sorted = y_pred_test[sort_indices]


plt.plot(X_test_sorted,y_pred_test_sorted,color="g")
plt.title("Sales of the Ice Cream at Different Temperatures")
plt.xlabel('Temperature (°C)')
plt.ylabel('Ice Cream Sales (units)')
plt.text(-1.5, 35, 'Testing validation', style='italic')
plt.show()
```

```python
#Checking the model performance for training  and testing sets
print('Train Score: ', Regression.score(x_poly,y_train))
print('Test Score: ', Regression.score(xpoly_test,y_test))
```
