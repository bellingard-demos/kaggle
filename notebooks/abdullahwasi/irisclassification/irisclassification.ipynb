### Implemented without label encoding

```python
# Dataset used is the IRIS dataset
df = pd.read_csv('../input/iris/Iris.csv') # loading dataset
```

```python
# dropping Id
df.drop("Id" ,axis =1 , inplace=True)
print(df.head())
print(df.shape)
```

```python
df.isnull().sum() #checking for null values
```

```python
df.info()
```

```python
print(df.describe())
x= df.drop(['Species'] , axis =1)  # input parameters
print(x)
y = df['Species']  #dependent variable (required paramenter)
print(y)
print(df.Species.unique())   # list all species
```

```python
import sklearn
from sklearn.model_selection import train_test_split
# splitting data for training and testing
x_train , x_test , y_train , y_test = train_test_split(x ,y , test_size=0.2 , random_state = 0)

acc_list = []
```

```python
df.corr() # correlation between different parameters
```

```python
sns.pairplot(df, hue='Species');
```

## Logistic Regression (with sag solver)

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver='sag') 
model.fit(x_train,y_train )    #training model
y_pred = model.predict(x_test)    # performing predictions
from sklearn.metrics import confusion_matrix , accuracy_score
names = ['Iris-setosa', 'Iris-versicolor' ,'Iris-virginica']
cm = confusion_matrix(y_test , y_pred , labels =names )
print(cm)


cmtx = pd.DataFrame(
    confusion_matrix(y_test, y_pred, labels=names), 
    index=['pred:{:}'.format(x) for x in names], 
    columns=['true:{:}'.format(x) for x in names]
)
print(cmtx)


print("Train data accuracy:",accuracy_score(y_true = y_train, y_pred=model.predict(x_train)))
print("Test data accuracy:",accuracy_score(y_true = y_test, y_pred=y_pred))

ac= accuracy_score(y_true = y_test, y_pred=y_pred)
acc_list.append(ac)
```
