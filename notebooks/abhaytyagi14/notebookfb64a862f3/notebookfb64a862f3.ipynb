```python
import re
import json
from PyPDF2 import PdfReader

# Load resume PDF 
pdf_file = "/kaggle/input/resume2/AnujSinghRawat_Resume.pdf"
pdf_reader = PdfReader(pdf_file)
resume_text = ""
for page in pdf_reader.pages:
    resume_text += page.extract_text()

# Define regex patterns
name_pattern = r'([^0-9]+)'
email_pattern = r'([\w.+-]+@[a-zA-Z0-9.-]+\.[a-zA-Z]{2,4})'
phone_pattern = r'((\+91)?[-\s]?(\d{3}[-\s]?\d{3}[-\s]?\d{4}))'

# Initialize extracted details  
details = {
    "name": "",
    "email": "", 
    "phone": ""
}

# Extract name
name_match = re.search(name_pattern, resume_text)
if name_match:
    details["name"] = name_match.group(1)

# Extract email 
email_match = re.search(email_pattern, resume_text)
if email_match:
    details["email"] = email_match.group(1)

# Likewise for phone number
phone_match = re.search(phone_pattern, resume_text)
if phone_match:
    details["phone"] = phone_match.group(1)

# Extract education
edu_pattern = r'(\d{4} -\d{4}).*?([^\n]+)' 
education = re.findall(edu_pattern, resume_text, re.DOTALL)
details["education"] = [{"period": edu[0], "institution": edu[1]} 
                         for edu in education]

# Extract skills
skills_pattern = r'(?i)(\b[A-Za-z-]+\b)'
raw_skills = re.findall(skills_pattern, resume_text)
details["skills"] = [skill.lower() for skill in raw_skills]

# Save extracted details as JSON
with open("resume_data.json", "w") as f:
    json.dump(details, f)
```
