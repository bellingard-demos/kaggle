# Data Visualization ( Movies ) 
# Abdallah Mosa 10/26/2022

```python
import seaborn as sns
import matplotlib.pyplot as plt
cf = pd.read_csv('../input/top-500-movies-budget/top-500-movies.csv').head(50)
cf
```

# The most popular movies are drama movies

```python
plt.figure( figsize = (15,10))
sns.barplot(x="genre",y="rank",data=cf)

plt.xlabel("Kind Of Movie",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.title("Number of movie types",fontsize = 20)
```

# In 2011, it was the highest year in production costs 

```python
plt.figure( figsize = (15,10))
sns.pointplot(x = "year", y = "production_cost", data = cf)
```
