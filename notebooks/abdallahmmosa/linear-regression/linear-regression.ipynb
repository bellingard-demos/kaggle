```python
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
df = pd.read_csv('../input/car-price-prediction/CarPrice_Assignment.csv')
df.head()
```

# Show the description and the info of the data set

```python
df.describe()
```

```python
df.info()
```

```python
x=df['enginesize'].values
x
```

```python
y = df['horsepower'].values
y
```

```python
plt.scatter(x,y,color='black')
plt.xlabel('enginesize')
plt.ylabel('horsepower')
plt.plot
```

```python
x = x.reshape(-1,1)
y = y.reshape(-1, 1)
```

```python
x_train, x_test, y_train, y_test  = train_test_split(x,y,train_size=.8, test_size=.2, random_state=400 )
```

```python
plt.scatter(x_train,y_train,color='red')
plt.xlabel('enginesize ')
plt.ylabel('horsepower')
plt.plot
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test) *100
```

```python
y_predict = lr.predict([[110]])
y_predict
```

```python
y_predict=lr.predict(x_test)
y_predict
```

```python
plt.scatter(x_train, y_train, color='blue')
plt.plot(x_test, y_predict, color='red')
plt.xlabel('enginesize')
plt.ylabel("horsepower")
plt.title("for x and y train (enginesize-horsepower)")
plt.plot
```

```python
plt.scatter(x_test, y_test, color='red')
plt.plot(x_test, y_predict, color='green')
plt.xlabel('enginesize')
plt.ylabel("horsepower")
plt.title("test x and y (enginesize-horsepower)")
plt.plot
```

# conclusions 

 1. Horsepower is highly dependent on enginesize but not mainly because there are engines with less enginesize and high horsepower
 2. Most engines have horsepower between 80-175 hp
