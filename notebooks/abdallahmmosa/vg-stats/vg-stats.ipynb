```python
df= pd.read_csv("../input/videogamesales/vgsales.csv")
df
```

# 1- Which company is the most common video game publisher?
 

```python
top_one_company = df["Publisher"].value_counts().head(1)
top_one_company
```

# 2-What’s the most common platform?


```python
top_one_platform = df["Platform"].value_counts().head(1)
top_one_platform
```

# 3-What about the most common genre?

```python
top_one_genre = df["Genre"].mode()[0]
top_one_genre
```

# 4-What are the top 20 highest grossing games?

```python
top_20_games = df[["Name","Global_Sales"]].sort_values("Global_Sales",ascending=False).head(20)
top_20_games
```

# 5- For North American video game sales, what’s the median?
#### a)Provide a secondary output showing ten games surrounding the median sales output.
#### b)Assume that games with same median value are sorted in descending order.

```python
north_american_sales = df["NA_Sales"].median()
print(f"The median of  North american sales ={north_american_sales}")
sort = df.loc[df["NA_Sales"] == north_american_sales].head(10)
sort
```

# 6-For the top-selling game of all time, how many standard deviations above/below the mean are its sales for North America?

```python
MEAN = df['NA_Sales'].mean()
STD = df['NA_Sales'].std()
NAtop = df.sort_values('NA_Sales', ascending=False)['NA_Sales'][0]
print((NAtop - MEAN)/STD)
```

# 7-The Nintendo Wii seems to have outdone itself with games. How does its average number of sales compare with all of the other platforms?

```python
wi = df[df['Platform'] == 'Wii']['Global_Sales'].mean()
others = df[df['Platform'] != 'Wii'].Global_Sales.mean()
print(wi/others)
print("wi platforms = ", wi )
print("the other platforms = ",others)
```

# 8. What is the total sales of PS3 platform?


```python
ps3 = df.loc[df["Platform"] == "PS3"]["Global_Sales"].sum()
print(f"total sales = {ps3}")
```

# 9- what’s the median for EU Sales

```python
north_american_sales = df["EU_Sales"].median()
print(f"The median of  EU sales ={north_american_sales}")
```

# 10- What is the total sales of Nintendo 64 platform?¶



```python
N64  = df.loc[df["Platform"] == "N64"]["Global_Sales"].sum()
print(f"total sales = {N64}")
```
