# Class Board 

```python

class ChessBoard:

    def __init__(self):
        self.board=  np.zeros((8,8,3))
        self.black = (0,0,0)
        self.white = (1,1,1)
        self.blue = (0,1,1)
        self.red = (1,.2,0)
        self.red_p_row= None
        self.red_p_col=None
        self.blue_p_row= None
        self.blue_p_col=None 

        
    def show_board(self):
        '''
        this function just to  design the chess board on white and black 
        '''
        for i in range(8):
            for j in range(8):
                if (j%2==0 and i%2==0) or j%2!=0 and i%2!=0:
                    self.board[i][j]=self.white

    def add_red(self,row,col):
        '''
        this function to recolor the location that the red queen set onit 
        
        '''
        self.board[row][col]=self.red
        self.red_p_row= row
        self.red_p_col=col
    def add_blue(self,row,col):
        '''
        this function to recolor the location that the blue  queen set onit 
        '''
        self.board[row][col]=self.blue
        self.blue_p_row= row
        self.blue_p_col=col

    def is_under_attack(self):
        '''
        this function to check if the queens is under attack through each other 
        '''
        if self.red_p_col == self.blue_p_col:
           
            return True
        if self.red_p_row == self.blue_p_row :
            
            return True
        if abs(self.red_p_row - self.blue_p_row)==abs(self.red_p_col - self.blue_p_col):
           
            return True 
        
        else : 
            
            return  False  
        
    def render(self):
        plt.imshow(self.board)
        
```
