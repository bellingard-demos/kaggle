```python
grouped = df.groupby('Continent')
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(15, 15))
sns.histplot(df, x='Pop2022', hue='Continent', ax=axes[0][0], kde=True)
axes[0][0].set_title("Population distribution by continent in 2022")
sns.histplot(df, x='Pop2023', hue='Continent', ax=axes[0][1], kde=True)
axes[0][1].set_title("Population distribution by continent in 2023")
sns.histplot(df, x='growthRate', hue='Continent', ax=axes[0][2], kde=True)
axes[0][2].set_title("Growth rate distribution by continent")
sns.barplot(x=grouped['Pop2022'].sum().index,
           y=grouped['Pop2022'].sum(),
           ax=axes[1][0])
sns.barplot(x=grouped['Pop2023'].sum().index,
           y=grouped['Pop2023'].sum(),
           ax=axes[1][1])
sns.barplot(x=grouped['growthRate'].mean().index,
           y=grouped['growthRate'].mean(),
           ax=axes[1][2])

axes[1][0].set_title("Population means for each continent in 2022")
axes[1][1].set_title("Population means for each continent in 2023")
axes[1][2].set_title("Growth Rate for each continent")

for cont1, cont2, cont3 in zip(axes[1][0].containers, axes[1][1].containers, axes[1][2].containers):
    axes[1][0].bar_label(cont1, size=15, color='black')
    axes[1][1].bar_label(cont2, size=15, color='black')
    axes[1][2].bar_label(cont3, size=15, color='black')
    
sns.boxplot(x=df['Continent'], y=df['Pop2022'], ax=axes[2][0])
axes[2][0].set_title("Population boxplots by continent in 2022")
sns.boxplot(x=df['Continent'], y=df['Pop2023'], ax=axes[2][1])
axes[2][1].set_title("Population boxplots by continent in 2023")
sns.boxplot(x=df['Continent'], y=df['Pop2022'], ax=axes[2][2])
axes[2][2].set_title("Growth rate boxplots by continent")
    
plt.tight_layout()
plt.show()
```

```python
def bottom_show(top):
    top_df1 = df.iloc[-top:, :]
    fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(20, 10))
    sns.barplot(top_df1,x='City', y='Pop2022', ax=axes[0])
    axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=45)
    sns.barplot(top_df1,x='City', y='Pop2023', ax=axes[1])
    axes[1].set_xticklabels(axes[1].get_xticklabels(), rotation=45)
    sns.barplot(top_df1,x='City', y='growthRate', ax=axes[2])
    axes[2].set_xticklabels(axes[2].get_xticklabels(), rotation=45)
    plt.suptitle("Top {} least populates cities in 2022 and in 2023\nand their growth rates".format(top), size=25)
    plt.tight_layout()
    plt.show()
```

```python
def top_cities(top, bottom=False):
    top_df = df.iloc[:top, :]
    fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(20, 10))
    sns.barplot(top_df,x='City', y='Pop2022', ax=axes[0])
    axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=45)
    sns.barplot(top_df,x='City', y='Pop2023', ax=axes[1])
    axes[1].set_xticklabels(axes[1].get_xticklabels(), rotation=45)
    sns.barplot(top_df,x='City', y='growthRate', ax=axes[2])
    axes[2].set_xticklabels(axes[2].get_xticklabels(), rotation=45)
    plt.suptitle("Top {} populates cities in 2022 and in 2023\nand their growth rates".format(top), size=25)
    plt.tight_layout()
    plt.show()
    if bottom:
        bottom_show(top)
```

```python
top_cities(10, True)
```

```python
def plot_barplot(x, y, axs):
    sns.barplot(x=x, y=y, ax=axs)
    for container in axs.containers:
        axs.bar_label(container, size=10, color='black')
        
    axs.set_xticklabels(axs.get_xticklabels(), rotation=50)

def top_countries(top):
    grouped = df.groupby('Country')
    sums22 = grouped['Pop2022'].sum()
    s22_df = pd.DataFrame(sums22).sort_values(by='Pop2022', ascending=False)
    sums23 = grouped['Pop2023'].sum()
    s23_df = pd.DataFrame(sums23).sort_values(by='Pop2023', ascending=False)
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 10))
    plot_barplot(s22_df.index[:top], s22_df.iloc[:top, 0], axes[0])
    plot_barplot(s23_df.index[:top], s23_df.iloc[:top, 0], axes[1])
    plt.suptitle("{} most populated countires".format(top, top),
                size=20)
    plt.tight_layout()
    plt.show()
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 10))
    plot_barplot(s22_df.index[-top:], s22_df.iloc[-top:, 0], axes[0])
    plot_barplot(s23_df.index[-top:], s23_df.iloc[-top:, 0], axes[1])
    plt.suptitle("{} least populated countires".format(top, top),
                size=20)
    plt.tight_layout()
    plt.show()
```

```python
top_countries(10)
```
