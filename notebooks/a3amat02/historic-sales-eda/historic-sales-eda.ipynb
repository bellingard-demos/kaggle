```python
def barplots(df, x, y):
    f,ax=plt.subplots(1,2,figsize=(25,10))
    Group_data = df.groupby(y)
    sns.barplot(x = Group_data[x].mean().index, 
                y = Group_data[x].mean().values,
                ax= ax[0],  palette = 'mako')
    for container in ax[0].containers:
        ax[0].bar_label(container,color='black',size=15)
    ax[0].set_xticklabels(Group_data[x].mean().index, 
                          rotation = 50)
    
    ax[0].set_ylabel("Mean values grouped by {}".format(y))

    palette_color = sns.color_palette('summer')
    plt.pie(x = df[y].value_counts(),
            labels=df[y].value_counts().index,
            autopct='%.0f%%',shadow=True,
            colors= palette_color)
    ax[1].set_title("Pie chart")
    
    plt.suptitle("{} barplots grouped by {} and pie charts".format(x.capitalize(), y), size=20)
    plt.show()
```

# Barplots grouped by categorical values and pie charts

```python
df['value'].hist(bins=10)
```

```python
barplots(df, 'value', 'year')
```

```python
barplots(df, 'value', 'region')
```

```python
barplots(df, 'value', 'powertrain')
```
