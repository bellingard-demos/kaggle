# Autoregression training and evaluation pipeline

User enters currency and model is trained for predicting this exact currency

```python
def train_for_currency(df, x):
    group = df[df['Currency'] == x.lower()]
    sample = group.iloc[:, -1].values
    train, test = sample[:int(len(sample)*0.8)], sample[int(len(sample)*0.8):]
    window = 29
    model = AutoReg(train,lags=29)
    model_fit = model.fit()
    predictions = model_fit.predict(start=len(train),end=len(train)+len(test)-1,dynamic=False)
    mae = mean_absolute_error(predictions, test)
    mse = mean_squared_error(predictions, test)
    print('MAE: {}\nMSE: {}'.format(mae, mse))
```

```python
train_for_currency(df, 'Bitcoin')
```
