```python
active_df = df[df['Rocket_Status'] == "StatusActive"].copy()
```

## Analysis of Active only Space Missions

As we can see, that China dominates when it comes to the number of Space Missions.
Three biggest active players are:
1. The USA(if we consider all of the US companies, of which at least 4 in top 8)
2. China
3. India
4. Russia

```python
plt.figure(figsize=(8, 6))
counts = active_df['Organisation'].value_counts()
lbs = active_df['Organisation'].value_counts().index
plt.pie(counts, autopct="%0.2f%%")
plt.tight_layout()
plt.legend(lbs)
```

```python
plt.figure(figsize=(8, 6))
counts = active_df['Mission_Status'].value_counts()
lbs = active_df['Mission_Status'].value_counts().index
plt.pie(counts, autopct="%0.2f%%")
plt.tight_layout()
plt.legend(lbs)
```
