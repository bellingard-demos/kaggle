```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/kaggle/input/minecraft-piracy-dataset/piracydataset.csv")
```

```python
df.shape
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

# Overall frequencies

```python
ticks = [x for x in range(0, int(max(df['age'])), 3)]
df['age'].plot.hist(bins=20)
plt.xticks(ticks)
```

```python
over = df['version'].value_counts()
labs = over.index
plt.pie(over, labels=labs, autopct="%0.2f%%")
```

```python
over
```

```python
sns.pairplot(df, vars=['age'], hue='version')
```

```python
all_clients = df['client'].value_counts()
labs_clients = all_clients.index
plt.pie(all_clients, labels=labs_clients, autopct="%0.2f%%")
```

# Cracked version analysis

```python
cracked = df[df['version'] == 'Cracked'].copy()
```

```python
ticks = [x for x in range(0, int(max(cracked['age'])), 3)]
cracked['age'].plot.hist(bins=20)
plt.xticks(ticks)
```

## Cracked versions Data Distribution of different clients among ages

```python
sns.pairplot(cracked, vars=['age'], hue='client')
plt.legend(cracked['client'].value_counts().index[::-1])
```

## Ratio of clients and frequencies

```python
counts = cracked['client'].value_counts()
labs = counts.index
plt.pie(counts,labels=labs,autopct="%0.2f%%")
```

```python
counts
```

# Paid version analysis

```python
paid = df[df['version'] == 'Paid Version'].copy()
```

```python
ticks = [x for x in range(0, int(max(paid['age'])), 3)]
paid['age'].plot.hist(bins=20)
plt.xticks(ticks)
```

```python
sns.pairplot(cracked, vars=['age'], hue='client')
plt.legend(cracked['client'].value_counts().index[::-1])
```

# Conclusion

There is no clear distinction that we could use to determine whether player is using Pirate version or Paid version. We would not tell that by age or gaming versions as filtered data remains the same.

But we know for sure, judging from overall, cracked and paid version histograms - minecraft is very popular among teenagers, especially among in the range between 14-20 years old.
