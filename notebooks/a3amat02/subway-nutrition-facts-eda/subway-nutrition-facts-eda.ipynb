```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/kaggle/input/subway-nutrition/SubwayMenuNutrition.csv")
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
categories = df['Category'].value_counts()
categories
```

<div>
    <h1 style="text-align:center;
               font-size:48px">
        All Subway food categories listed
    </h1>
</div>

```python
plt.pie(categories, autopct='%0.2f%%', labels=categories.index)
plt.show()
```

<div>
    <h1 style="text-align:center;
               font-size:48px">
        Average values for each nutritional fact per category in Subway
    </h1>
</div>

```python
cols = df.columns[2:]
fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(20, 10))

grouped = df.groupby('Category')

k = 0
for i in range(2):
    for j in range(4):
        means = grouped[cols[k]].mean()
        sns.barplot(x=means.index, y=means, ax=axes[i][j])
        for container in axes[i][j].containers:
            axes[i][j].bar_label(container, size=12)
        axes[i][j].set_xticklabels(axes[i][j].get_xticklabels(), rotation=45)
        axes[i][j].set_title("Mean value of {} per category".format(cols[k]))
        k += 1
plt.tight_layout()
plt.show()
```
