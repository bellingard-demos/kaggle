```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/kaggle/input/obesity-classification-dataset/Obesity Classification.csv")
```

```python
def plots(df, x, y):
    group = df.groupby(y)
    fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(20, 10))
    sns.histplot(df, x=x, kde=True, color='g', ax=axes[0])
    sns.barplot(x=group[x].mean(),
               y=group[x].mean().index,
               ax=axes[1])
    
    for container in axes[1].containers:
        axes[1].bar_label(container, size=15, color='black')
        
    axes[2].pie(df[y].value_counts(),
               labels=df[y].value_counts().index,
               autopct='%0.2f%%')
    
    plt.tight_layout()
    
    plt.show()
```

```python
numericals = ['Age', 'Height', 'Weight', 'BMI']
```

# Barplots of mean values grouped by categorical values

```python
for i in numericals:
    plots(df, i, 'Label')
```

# Split by gender

```python
for i in numericals:
    plots(df[df['Gender'] == 'Male'], i, 'Label')
```

```python
for i in numericals:
    plots(df[df['Gender'] == 'Female'], i, 'Label')
```

```python
sns.pairplot(df, vars=numericals, hue='Label', corner=True)
```

# Establishing differences in numerical parameters between men and women

```python
sns.pairplot(df,vars=numericals, hue='Gender', corner=True)
```
