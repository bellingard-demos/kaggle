```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/kaggle/input/covid-19-india-statewise-2023/IndiaCovidCasesByStateUT.csv")
```

```python
def process_int(x):
    x = x.replace(",","")
    return int(x)
df['Total Cases'] = df['Total Cases'].apply(process_int)
df['Active'] = df['Active'].apply(process_int)
df['Discharged'] = df['Discharged'].apply(process_int)
df['Deaths'] = df['Deaths'].apply(process_int)
```

```python
def process_perc(x):
    x = x[:-1]
    return float(x)

df['Active Ratio'] = df['Active Ratio'].apply(process_perc)
df['Discharge Ratio'] = df['Discharge Ratio'].apply(process_perc)
df['Death Ratio'] = df['Death Ratio'].apply(process_perc)
```

```python
def top_plot(df, column, ax, top=10):
    dt = df.sort_values(column, ascending=False)
    sns.barplot(dt, x=dt.iloc[:top, 0], y=dt.loc[:top, column], ax=ax)
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45, fontsize=8)
    ax.set_title(f"Top {top} states sorted by {column}")
```

```python
print("Amount of total cases in 2023", sum(df['Total Cases']))
```

```python
fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(15, 5))
top_plot(df, 'Total Cases', axes[0], 10)
top_plot(df, 'Active', axes[1], 10)
top_plot(df, 'Discharged', axes[2], 10)
top_plot(df, 'Deaths', axes[3], 10)
plt.tight_layout()
plt.show()
```
