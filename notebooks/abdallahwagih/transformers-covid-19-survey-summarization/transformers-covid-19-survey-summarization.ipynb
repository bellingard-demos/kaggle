```python
model = pipeline(task="summarization", model="t5-small", min_length=50, max_length=120, truncation=True)
```

```python
print(df["Description"][0])
```

```python
descriptions = df["Description"].tolist()
description_summary = model(descriptions[:])
```

```python
summary = []
for i in range(len(description_summary)):
    summary.append(description_summary[i]['summary_text'])
```

```python
summary[:5]
```

```python
df['description_summary'] = summary
```
