```python
from transformers import pipeline
```

```python
# Load the mBART translation model
translation_model = pipeline("translation", model="facebook/mbart-large-50-many-to-many-mmt")
```

```python
source_language = input('Select Source Language ["ar", "en", "fr", "de", "es", "zh", "ja"]')
target_language = input('Select Target Language ["en", "fr", "de", "es", "zh", "ja", "ar"]')
```

```python
text_to_translate = input("Enter Text to Translate")
```

```python
translation_result = translation_model(text_to_translate, src_lang=source_language, tgt_lang=target_language)
```

```python
print(f"{text_to_translate}")
print(f"{translation_result[0]['translation_text'][3:]}")
```
