```python
import numpy as np
import pandas as pd
from DeepImageSearch import Load_Data, Search_Setup
```

## **Load data from folder**

```python
image_list = Load_Data().from_folder(['../input/landscape-pictures'])
```

## **Get model**

```python
# Set up the search engine, You can load 'vit_base_patch16_224_in21k', 'resnet50' etc more then 500+ models
st = Search_Setup(image_list=image_list, model_name='vgg19', pretrained=True, image_count=100)
```

## **Indexing data**

```python
# Index the images
st.run_index()
```

```python
# Get metadata
metadata = st.get_image_metadata_file()
```

## **Get similar imags**

```python
# for searching, you need to give the image path and the number of the similar image you want
st.get_similar_images(image_path=image_list[0], number_of_images=10)
```

## **Plot similar images**

```python
# Plot similar images
st.plot_similar_images(image_path=image_list[0], number_of_images=9)
```
