## **Import transformers**

```python
from transformers import pipeline
```

## **Get model**

```python
# Get model
model = pipeline("question-answering")
```

## **Set article text**

```python
sentence = input('Please paste your article :')
```

## **Set question**

```python
question = input("Questions from this article?")
```

## **Get answer**

```python
answers = model(question=question, context=sentence)
```

```python
answers['answer']
```
