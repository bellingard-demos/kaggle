# **EDA(Exploratory Data Analysis)**
EDA stands for Exploratory Data Analysis. It is a type of data analysis that involves exploring and summarizing data sets to gain insights into the patterns, relationships, and distributions of the data. EDA data analysis is used to identify outliers, missing values, and other anomalies in the data, as well as to identify trends and patterns that can help guide further analysis or inform decision-making. EDA data analysis can involve a range of techniques, including visualization, descriptive statistics, and statistical modeling, and is often used in fields such as data science, business analytics, and social science

# **Data Visualization**

second, we will use matplotlib and seaborn what is matplotlib and seaborn ?

**Matplotlib**: is an open-source data visualization library for Python. It provides a variety of customizable plots and charts for visualizing data in a clear and concise manner. Matplotlib can be used to create line plots, scatter plots, bar plots, histograms, 3D plots, and many other types of visualizations. integrates well with other scientific Python libraries such as Pandas and SciPy. Matplotlib can also be used to create interactive visualizations and animations. It is widely used in fields such as data science, engineering, finance, and social sciences for data exploration and presentation.


**Seaborn**: is a Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating informative and attractive statistical graphics. Seaborn includes a variety of plot types such as scatter plots, line plots, bar plots, heatmaps, and more. It also offers advanced features such as color palettes, statistical functions, and themes that make it easy to create professional-looking visualizations. Seaborn is commonly used in data science, machine learning, and statistical analysis to explore and communicate data insights.

## **Sales of games from 1980 to 2020**

```python
#Recall import seaborn as sns for use seaborn
sns.lineplot(x='Year', y='NA_Sales',data=data,label='North America')
sns.lineplot(x='Year', y='JP_Sales',data=data,label='Japan')
sns.lineplot(x='Year', y='EU_Sales',data=data,label='Europe')
#Recall import matplotlib.pyplot as plt for use matplotlib
plt.xlabel("years")
plt.title("Sales of games from 1980 to 2020")
plt.ylabel("USD in millions")
#add value in list
plt.legend()
plt.show()
```

Here I have created a line chart for game sales from 1980 to 2020 for sales in North America, sales in Europe, and sales in Japan. We note that America has the highest sales, followed by Japan, then Europe

## **Correlations between features in dataset**

```python
#Recall import seaborn as sns for use seaborn
corr_matrix = data.corr()
top_corr_features = corr_matrix.index
plt.figure(figsize=(8,6))
#plot heat map
g=sns.heatmap(data[top_corr_features].corr(),annot=True,cmap="RdYlGn")
plt.title("Correlations between features in dataset")
```

Here I have created correlations for each feature in the data set except for the categorical features. We note that most of the data have a weak relationships with Rank and Yaer and strong relationships with Sales in North America, Sales in Europe,Sales in JapanSales




### **What is market share for top 10 Platforms**

```python
gradable_counts=data.Platform.value_counts().nlargest(10)
# Plot a pie chart
gradable_counts.plot(kind='pie', title='Top 10 Platforms', figsize=(8,8),autopct='%.2f %%')
plt.legend()
plt.show()
```

Here I have created a pie chart of the percentage of the top 10 selected platforms. Note the platforms DS, PS2, PS3, and Wii have the largest percentage of games, and platforms PC, XB, and GBA have the lowest  percentage of games



## **What is market share for all Genre**

```python
best_Genre=data.Genre.value_counts()
best_Genre.plot(kind='bar', title='Top Genres', figsize=(8,6))
plt.legend()
plt.show()
```

Here I have created a bar chart best Genres Note the Genres Action, Sports, and Misc have the largest number of games, and Fighting, Strategy, and Puzzle  have the lowest number of games
