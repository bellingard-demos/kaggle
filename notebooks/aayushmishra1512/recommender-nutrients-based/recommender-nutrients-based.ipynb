```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
df  =pd.read_csv('/kaggle/input/nutrition-details-for-most-common-foods/nutrients_csvfile.csv')
```

```python
df
```

```python
df.dtypes
```

```python
df=df.replace("t",0)
df=df.replace("t'",0)
df.head()
```

```python
df=df.replace(",","", regex=True)
df['Fiber']= df['Fiber'].replace("a","", regex=True)
df['Calories'][91]=(8+44)/2
```

```python
df['Grams']=pd.to_numeric(df['Grams'])
df['Calories']=pd.to_numeric(df['Calories'])
df['Protein']=pd.to_numeric(df['Protein'])
df['Fat']=pd.to_numeric(df['Fat'])
df['Sat.Fat']=pd.to_numeric(df['Sat.Fat'])
df['Fiber']=pd.to_numeric(df['Fiber'])
df['Carbs']=pd.to_numeric(df['Carbs'])
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
cat = pd.get_dummies(df['Category'])
cat
```

```python
df = pd.concat([df,cat],axis = 1)
```

```python
df.head()
```

```python
df.drop('Category',axis =1 ,inplace = True)
```

```python
df
```

```python
rec = df.pivot_table(columns='Food',values=['Calories','Fat','Protein','Carbs','Fiber','Sat.Fat'])
rec
```

```python
rec1 = rec['Fortified milk']
```

```python
similar = rec.corrwith(rec1)
```

```python
corr = pd.DataFrame(similar,columns=['Correlation'])
corr.dropna(inplace=True)
```

```python
corr.sort_values('Correlation',ascending = False).head(10)
```

```python
(corr['Correlation']> 0.999000).sort_values(ascending = False).head(10)
```
