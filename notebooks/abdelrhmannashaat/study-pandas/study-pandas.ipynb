```python
# i can convert map to dataframe
weather_data = {
    'day' : ['1/1/2107','1/2/2107','1/3/2107','1/4/2107','1/5/2107','1/6/2107'],
    'temperature' : [32,35,28,24,32,31],
    'windspeed' : [6,7,2,7,4,2],
    'evvent' : ['Rain','Sunny','Snow','Snow','Sunny','Sunny']
}
df = pd.DataFrame(weather_data)
df
```

```python
df['temperature'].mean()
```

```python
df['temperature'].std()
```

```python
# get day with minimum temperature
df['day'][df['temperature'] == df['temperature'].min()]
```

```python
# change index 
df.set_index('day',inplace=True)
df
```

```python
# add new column
df['NewData'] = np.arange(1,7)
df
```

```python
df.drop('NewData' , inplace=True , axis = 1) # reomve column , (axis 1 column , axis 0 row)
df
```

```python
# handlling missing data
new_data = pd.read_csv('/kaggle/input/pracrice-pandas/weather_data.csv') 
new_data
```

```python
new_data1 = new_data.fillna(0)
new_data1
```

```python
 # fill nan with previous value
new_data2 = new_data.fillna(method='ffill')
new_data2
```

```python
# fill nan with next value
new_data3 = new_data.fillna(method='bfill')
new_data3
```

```python
# make pandas excpect missing values
new_data4 = new_data.interpolate() 
new_data4
```
