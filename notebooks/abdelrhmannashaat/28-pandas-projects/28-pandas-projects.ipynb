```python
# What is the type of the columns?
crime_df.dtypes
```

```python
# Convert the type of the column Year to datetime64
crime_df['Year'] = pd.to_datetime(crime_df['Year'])
crime_df['Year'].dtypes
```

```python
# Set the Year column as the index of the dataframe
crime_df['Year'] = crime_df['Year'].astype(int)
crime_df.set_index('Year',inplace = True)
crime_df
```

```python
# Delete the Total colum
crime_df.drop(columns=['Total'],inplace = True)
crime_df
```

```python
# Group the year by decades and sum the values
def get_decade(year):
    return (year // 10) * 10
crime_df['total'] = crime_df['Violent'] + crime_df['Property'] + crime_df['Murder'] + crime_df['Forcible_Rape'] + crime_df['Aggravated_assault'] + crime_df['Burglary'] + crime_df['Larceny_Theft'] + crime_df['Vehicle_Theft']
crime_df.groupby(get_decade)['total'].sum()
```

```python
# What is the most dangerous decade to live in the US?
crime_df.groupby(get_decade)['total'].sum().idxmax()
```
