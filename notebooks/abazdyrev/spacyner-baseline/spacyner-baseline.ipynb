```python
import pandas as pd
import spacy
```

```python
nlp = spacy.load("uk_core_news_md")

df_test = pd.read_csv('/kaggle/input/nlp-ua-locations-extractions/test.csv')
df_test.sample(5)
```

```python
df_test['locations'] = df_test.text.apply(
    lambda x: [str(e) for e in nlp(x).ents if e.label_ == 'LOC']
)
df_test.sample(5)
```

```python
df_test[['text_id', 'locations']].to_csv('spacy_baseline.csv', index=False)
```
