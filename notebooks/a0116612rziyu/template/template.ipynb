```python
data = pd.read_csv("/kaggle/input/hdbresale/train.csv")
data
```

```python
data["storey"] = data["storey_range"].str[:2]
x = data[["lease_commence_date","floor_area_sqm","storey"]].values
y = data["resale_price"].values.reshape(-1, 1)
print(x)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression().fit(x, y)
r_sq = model.score(x, y)
print(r_sq)
```

```python
test = pd.read_csv("/kaggle/input/hdbresale/test.csv")
test
```

```python
test["storey"] = test["storey_range"].str[:2]
x_pred = test[["lease_commence_date","floor_area_sqm","storey"]].values
y_pred = model.predict(x_pred)
print(y_pred)
```

```python
submission = pd.DataFrame.from_dict({'id': range(1, len(y_pred)+1), 'resale_price': y_pred[:,0]})
submission
```

```python
submission.to_csv('submission.csv', index=False)
```
