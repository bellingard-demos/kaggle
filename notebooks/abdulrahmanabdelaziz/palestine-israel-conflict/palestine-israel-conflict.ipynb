```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r'/kaggle/input/fatalities-in-the-israeli-palestinian-conflict/fatalities_isr_pse_conflict_2000_to_2023.csv')
df.shape
```

```python
df.sample(10)
```

```python
df['data_of_event'] = pd.to_datetime(df['date_of_event'])
df['date_of_death'] = pd.to_datetime(df['date_of_event'])
```

```python
df.info()
```

```python
ax = sns.countplot(x=df['citizenship'])
plt.title('# of people killed')
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
df['event_location'].describe()
```

```python
plt.figure(figsize=(10,6),dpi=100)
ax = sns.countplot(x=df['event_location_district'])
plt.title('# People killed in District Location')
for bar in ax.containers:
    ax.bar_label(bar)
plt.xticks(rotation =90)    
```

```python
plt.figure(figsize=(10,6),dpi=100)
ax = sns.countplot(x=df['event_location_region'])
plt.title('Event Location Region')
for bar in ax.containers:
    ax.bar_label(bar)
plt.xticks(rotation =90)    
```

```python
ax = sns.countplot(x=df['type_of_injury'])
for bar in ax.containers:
    ax.bar_label(bar)
plt.xticks(rotation=90)    
```

```python
plt.title('Distribtuion of Ages')
sns.histplot(df['age'],bins=20,kde=True)
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x=df['event_location_region'],hue=df['event_location_district'])
```

```python
death_by_type =df.groupby('killed_by')['name'].count().sort_values(ascending=False).reset_index()
sns.barplot(data = death_by_type ,y='killed_by',x='name',palette='viridis')
```

```python
df['date_of_death'] = pd.to_datetime(df['date_of_death'])

# Now, you can use the resample function
fig = plt.figure(figsize=(10, 8))
fig.add_subplot(211)
df.resample('Y', on='date_of_death')['name'].count().plot(cmap='viridis', title='Total recorded casualties by year')
plt.show()
```

```python
df['date_of_event'] = pd.to_datetime(df['date_of_event'])
ax = df.assign(year_of_event=lambda x: x.date_of_event.dt.year).groupby('year_of_event').count().plot(
    kind='barh',y=['event_location', 'took_part_in_the_hostilities','place_of_residence'] ,
    title='Number of Fatalities by Year',
    cmap='viridis', figsize=(10, 9))
plt.show()
```
