```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import warnings
warnings.filterwarnings('ignore')
print('setup completed!')
```

```python
ax = sns.countplot(x='Species',data = df)
for bar in ax.containers:
    ax.bar_label(bar,fmt='%.0f',label_type='edge', fontsize=12, weight='bold')
```

```python
numeric_col = list(df.select_dtypes(exclude='object').columns)
for col in numeric_col:
    sns.histplot(x=df[col],kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
for col in numeric_col:
    plt.figure(figsize=(4,1),dpi=100)
    sns.boxplot(x= df[col])
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
# Calculate the lower and upper limits
Q3 = df['SepalWidthCm'].quantile(0.75)
Q1 = df['SepalWidthCm'].quantile(0.25)
iqr = Q3 - Q1
lower_limit = Q1 - 1.5 * iqr
upper_limit = Q3 + 1.5 * iqr

# Identify outliers
outliers = (df['SepalWidthCm'] < lower_limit) | (df['SepalWidthCm'] > upper_limit)

# Create a new DataFrame containing only the outliers
outlier_df = df[outliers]

# Print or do further analysis with the outlier_df
outlier_df
```
