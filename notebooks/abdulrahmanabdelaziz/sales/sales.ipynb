```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
print('setup compeleted')
```

```python
df = pd.read_csv('/kaggle/input/videogamesales/vgsales.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True,axis=0)
```

```python
df.isnull().sum()
```

```python
df[df.duplicated()] # No duplicate rows
```

```python
df.describe().round(2)
```

```python
plt.figure(figsize=(10,8),dpi=100)
df['Genre'].value_counts().sort_values(ascending =False).plot(kind='bar')
```

```python
df['Publisher'].value_counts()
```

```python
df['Publisher'].nunique()
```

```python
plt.figure(figsize=(7,10),dpi=100)
df.pivot_table(values =['NA_Sales','EU_Sales','JP_Sales','Other_Sales'],
                  index ='Genre',
              aggfunc='mean').plot(kind='bar',color=['red','green','orange'])
plt.title('Average sales per Genre')
plt.show()
```

```python
df.head()
```

```python
df['Year']= pd.to_datetime(df['Year'])
```

```python
df.head()
```

```python
df['year'] = df['Year'].dt.year
```

```python
df['year']
```

```python
df.pivot_table(values='Global_Sales',index='Genre',aggfunc='sum').sort_values(by=['Global_Sales'],ascending=False)
```

```python
plt.figure(figsize=(7,9),dpi=100)
df.pivot_table(values='Global_Sales',index='Genre',aggfunc='sum').sort_values(by=['Global_Sales'],ascending=False).plot(kind='bar',color='green')
plt.title('Sum Of Sales By GENRE',size=20)
```

```python
plt.figure(figsize=(10,5))
df['Platform'].value_counts().plot(kind='bar',color=['orange'])
```

```python
df.groupby('Platform')['NA_Sales'].sum()
```

```python
plt.figure(figsize=(12,9),dpi=100)
df.groupby('Platform')['NA_Sales'].sum().plot(kind='bar',color=['black'])
plt.title('Sum of Na_sales per platform',size=15)
plt.xlabel('platform',size=15)
plt.title('Sum of sales',size=15)
```
