### 
### Pay_Method = 0: Digital Wallets, 1: Card, 2: PayPal, 3: Other

```python
df['Pay_Method'].value_counts()
```

```python
plt.figure(figsize=(12,6),dpi =100)
sns.countplot(x= df['Pay_Method'])
plt.xlabel('payment Method',size=20)
```

```python
count_of_Digital_Wallets=df['Pay_Method'].value_counts()[0]
count_of_Card =df['Pay_Method'].value_counts()[1]
count_of_paypal = df['Pay_Method'].value_counts()[2]
other = df['Pay_Method'].value_counts()[3]
data = [count_of_Digital_Wallets,count_of_Card,count_of_paypal,other]
explode = [0.0,0.0,0.1,0.0]
labels =['Digital','Card','Paypal','Others']
colors = ['red', 'yellow', 'orange', 'purple']
fig, ax = plt.subplots(figsize=(7, 7))
ax.pie(data ,labels =labels ,autopct='%1.1f%%',colors = colors,
       explode=explode,shadow=True,startangle=90, wedgeprops={'edgecolor': 'white'})
ax.set_title('Payment Method')
ax.legend(title='Payment', loc='best')
plt.show()
```

## Browser = 0: Chrome, 1: Safari, 2: Edge, 3: Other

```python
plt.figure(figsize=(8,5),dpi =100)
sns.countplot(x= df['Browser'])
plt.xlabel('Browser',size=20)
```

```python
chrome = df['Browser'].value_counts()[0]
safari = df['Browser'].value_counts()[1]
Edge= df['Browser'].value_counts()[2]
other = df['Browser'].value_counts()[3]
data = [chrome,safari,Edge,other]
explode = [0.0,0.0,0.1,0.0]
labels =['chrome','safari','Edge','Others']
colors = ['red', 'yellow', 'orange', 'purple']
fig, ax = plt.subplots(figsize=(7, 7))
ax.pie(data ,labels =labels ,autopct='%1.1f%%',colors = colors,
       explode=explode,shadow=True,startangle=90, wedgeprops={'edgecolor': 'white'})
ax.set_title('Browsers used')
plt.show()
```
