```python
import pandas as pd
import pandas_profiling as pf
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
 
import warnings
warnings.filterwarnings('ignore')
print('Setup Completed')
```

```python
df = pd.read_csv('/kaggle/input/salary-data/Salary_Data.csv')
```

```python
pf.ProfileReport(df)
```

```python
sns.histplot(df['Age'],kde=True,color='orange')
```

```python
df['Education Level'].value_counts()
```

```python
plt.figure(figsize=(10,6),dpi=100)
df['Education Level'].value_counts().plot(kind='bar',color='g')
plt.title('Education Level ratio',size=16)
```

```python
df['Years of Experience'].unique()
```

```python
plt.figure(figsize=(8,6),dpi=100)
sns.regplot(x=df['Years of Experience'],y = df['Salary'],color='red')
plt.title('Relationship between experience and salary',size=20)
plt.xlabel('Years of Experience',size=12)
plt.ylabel('Salary',size=12)
plt.show()
```

```python
df.groupby('Job Title').max()['Salary'].sort_values(ascending =False)
```

```python
plt.figure(figsize=(14,6),dpi=100)
df.groupby('Job Title').max()['Salary'].sort_values(ascending =False).head(10).plot(kind='bar')
plt.title('Top Ten unique salaries of positions',size=20)
plt.xlabel('job title',size=20)
plt.ylabel('salary',size=20)
plt.show()
```

```python
sns.kdeplot(df['Salary'],shade=True)
```

```python
sorted_df = df.sort_values('Salary', ascending=False)
top_10_df = sorted_df.head(10)

num_data_points = len(top_10_df)
fig, ax = plt.subplots(figsize=(10, num_data_points * 0.3))

sns.boxplot(y=top_10_df['Job Title'], x=top_10_df['Salary'], ax=ax)

ax.set_title('Top 10 Salaries by Job')
ax.set_ylabel('Job Title')
ax.set_xlabel('Salary ($)')
ax.set_xticklabels(ax.get_xticks(), rotation=45)

plt.tight_layout()
plt.show()
```
