```python
df.drop(columns=["Id","Species"], inplace=True)
```

**PCA**

```python
pca = PCA()
pca.fit(df)

ratio= pca.explained_variance_ratio_
cum_ratio=np.cumsum(ratio)
x= list(range(len(ratio)))
plt.bar(x, ratio)
plt.plot(x, cum_ratio, 'r-')
plt.show()
```

```python
NB_CLUSTERS = 10
```

```python
data = np.zeros((NB_CLUSTERS-1,2))
for j in range(2,NB_CLUSTERS+1):
    km = KMeans(n_clusters=j)
    km.fit(df)
    lab = km.labels_
    score = silhouette_score(df,lab)
    data[j-2,:] = (j,score)
    print("Nombre clusters : ", j, "Score : ", score)
```

```python
rs = RobustScaler()
NUMERIC_COLS = [c for c in df.columns if c not in ["cut", "color", "clarity"]]
df[NUMERIC_COLS] = rs.fit_transform(df[NUMERIC_COLS])
```
