```python
df.plot(kind="scatter", x="SepalLengthCm", y="SepalWidthCm")
```

```python
import seaborn as sns

sns.FacetGrid(df, hue="Species", palette="husl") \
   .map(plt.scatter, "SepalLengthCm", "SepalWidthCm") \
   .add_legend()
```

```python
#plotting in sets using different colors

sns.FacetGrid(df, hue="Species", palette="hls") \
   .map(plt.scatter, "SepalLengthCm", "SepalWidthCm") \
   .add_legend()


sns.FacetGrid(df, hue="Species", palette="Set2") \
   .map(plt.scatter, "SepalLengthCm", "SepalWidthCm") \
   .add_legend()

sns.FacetGrid(df, hue="Species", palette="cubehelix") \
   .map(plt.scatter, "SepalLengthCm", "SepalWidthCm") \
   .add_legend()
```

```python
# boxplot
sns.boxplot(x="Species", y="PetalLengthCm", palette="husl", data=df)
```

```python
# Line Graph
sns.FacetGrid(df, hue="Species", palette="husl").map(sns.kdeplot, "PetalLengthCm") \
   .add_legend()
```

```python
# Pairplot
sns.pairplot(df.drop("Id", axis=1), hue="Species", palette="husl")
```

```python
# Histplot
sns.histplot(df, x="SepalLengthCm",y="SepalWidthCm",hue="Species",palette="husl")
```

```python
#Histplot :: Histogram
import matplotlib.pyplot as plt
plt.hist(df["SepalLengthCm"], bins=20, color="green")
plt.title("Sepal Length in cm")
plt.xlabel("Sepal Length")
plt.ylabel("Count")
```

```python
plt.hist(df["PetalWidthCm"], bins=20, color="orange")
plt.title("Petal Width in cm")
plt.xlabel("Petal Width")
plt.ylabel("Count")
```

```python
plt.hist(df["PetalLengthCm"], bins=20, color="blue")
plt.title("Petal Length in cm")
plt.xlabel("Petal Length")
plt.ylabel("Count")
```

```python
plt.hist(df["SepalLengthCm"], bins=20, color="green")
plt.hist(df["SepalWidthCm"], bins=20, color="blue")
plt.hist(df["PetalWidthCm"], bins=20, color="orange")
plt.hist(df["PetalLengthCm"], bins=20, color="pink")
```

```python
# Distplot
sns.FacetGrid(df, hue="Species").map(sns.distplot,"PetalLengthCm").add_legend()
sns.FacetGrid(df, hue="Species").map(sns.distplot,"PetalWidthCm").add_legend()
sns.FacetGrid(df, hue="Species").map(sns.distplot,"SepalLengthCm").add_legend()
sns.FacetGrid(df, hue="Species").map(sns.distplot,"SepalWidthCm").add_legend()
```

```python
plt.plot(df.Id, df["SepalLengthCm"],"yo")
plt.plot(df.Id, df["PetalWidthCm"],"g^")
plt.plot(df.Id, df["PetalLengthCm"],"bs")
plt.plot(df.Id, df["SepalWidthCm"],"ro")
plt.show()
```

```python
# pie chart to check the distribution of species/classes
plt.subplots(1,1,figsize=(10,8))
df["Species"].value_counts().plot.pie(explode=[0.01,0.01,0.01],autopct='%1.1f%%')
plt.show()
```
