```python
married_counts = df['Married'].value_counts()
plt.bar(married_counts.index, married_counts.values)
plt.xlabel('Married or not')
plt.ylabel('Married Count')
```

```python
sex = df['sex'].value_counts()
plt.bar(married_counts.index, married_counts.values)
plt.xlabel('sex')
plt.ylabel('depressed')
```

```python
Age_counts = df['Age'].value_counts()
plt.bar(Age_counts.index, Age_counts.values)
plt.title('Depression Age')
plt.xlabel('Age')
plt.ylabel('depressed                  ')
plt.show()
```

```python
Number_children_counts = df['Number_children'].value_counts()
plt.bar(Number_children_counts.index, Number_children_counts.values)
plt.title('Depression Number_children')
plt.xlabel('Number_children')
plt.ylabel('depressed')
plt.show()
```

```python
education_level_counts = df['education_level'].value_counts()
plt.bar(education_level_counts.index, education_level_counts.values)
plt.title('Depression education_level')
plt.xlabel('education_level')
plt.ylabel('depressed')
plt.show()
```
