```python
from statsmodels.tsa.seasonal import seasonal_decompose

def decompose(df, column_name):
    df = df.asfreq("d")
    df[column_name].interpolate(inplace=True)
    
    result_mul = seasonal_decompose(df[column_name], model='multiplicative', extrapolate_trend='freq')
    result_add = seasonal_decompose(df[column_name], model='additive', extrapolate_trend='freq')

    plt.rcParams.update({'figure.figsize': (30, 15)})
    result_mul.plot().suptitle('Multiplicative Decompose', fontsize=40)
    result_add.plot().suptitle('Additive Decompose', fontsize=40)
    
    plt.show()
    
    return result_mul, result_add
```

```python
result_mul, result_add = decompose(df, 'Open')
```

```python
df_reconstructed = pd.concat([result_add._seasonal, result_add.trend, result_add.resid, result_add.observed], axis = 1)
df_reconstructed.columns = ['seas', 'trend', 'resid', 'actual_values']
df_reconstructed
```
