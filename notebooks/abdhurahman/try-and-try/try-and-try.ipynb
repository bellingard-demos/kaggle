# Abourt this Dataset
* Username: The username of the user who provided the rating.
* Anime ID: A unique identifier for each anime title.
* My Score: The rating given by the user for the anime on a scale of 1 to 10.
* User ID: A unique identifier for each user.
* Gender: The gender of the user.
* Title: The title of the anime.
* Type: The type of the anime (e.g., TV series, movie, OVA, etc.).
* Source: The source material of the anime (e.g., manga, light novel, original, etc.).
* Score: The average score of the anime based on all user ratings.
* Scored By: The number of users who have scored the anime.
* Rank: The ranking of the anime.
* Popularity: The popularity rank of the anime.
* Genre: The genre(s) of the anime.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/kaggle/input/myanimelist-dataset/final_animedataset.csv')
```

TOP 50 ANIME

```python
df = df.drop_duplicates(subset=['title'], keep='first')
topanime = df[['title', 'score', 'scored_by', 'rank']].sort_values(by="rank",ascending=True)
topanime.head(10)
```

```python
import seaborn as sns

topanime = topanime.head(20)
sns.set(style='whitegrid', palette='Blues')
plt.figure(figsize=(12, 10))
graph = sns.barplot(y='score', x='title', data=topanime)
plt.title('Top 20 anime By Rank', fontsize=20)
plt.xticks(rotation=90, ha='right')
for p in graph.patches:
    height = p.get_height()
    graph.text(p.get_x() + p.get_width() / 2, height, f'{height:.1f}', ha='center', va='bottom')
plt.tight_layout()
plt.show()
```
