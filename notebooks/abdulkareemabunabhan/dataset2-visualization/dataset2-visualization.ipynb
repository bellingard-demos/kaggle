```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv("/kaggle/input/netflix-shows/netflix_titles.csv")
df.head()
```

```python
count_type = sns.countplot(data=df, x="type")
count_type.set_xticklabels(count_type.get_xticklabels(), rotation=45, ha="right")
count_type.set_xlabel("Movies and T.V shows")
count_type.set_ylabel("Count")
count_type.set_title("Count of Netflix realeses")
plt.tight_layout()
plt.show()
```

#### Insight graph 1
<p>this visualization shows that Netflix produce much more movies than T.V shows and it is a step in the strip to show how much Netflix produce each year<p>

```python
df_filtered = df[df["release_year"] >= 2004]

count_by_year_type = df_filtered.groupby(["release_year", "type"]).size().unstack()

fig, ax = plt.subplots(figsize=(12, 6))

num_years = len(count_by_year_type)

bar_width = 0.35

years = count_by_year_type.index
indices = range(num_years)

ax.bar(indices, count_by_year_type["Movie"], width=bar_width, label="Movies")

ax.bar([i + bar_width for i in indices], count_by_year_type["TV Show"], width=bar_width, label="TV Shows")

ax.set_xlabel("Year")
ax.set_ylabel("Count")
ax.set_title("Count of TV Shows and Movies by Year (Starting from 2004)")

ax.set_xticks([i + bar_width/2 for i in indices])
ax.set_xticklabels(years)

ax.legend()

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

#### Insight graph 2
<p>this visualization shows the yearly production of each of movies and TV shows and it shows that each year the count of production for each TV shows and movies is getting higher and that the movies is always more then TV shows <p>

```python
df_filtered = df.dropna(subset=["director", "type"])

count_by_director_type = df_filtered.groupby(["director", "type"]).size().unstack()

count_by_director_type["Total"] = count_by_director_type.sum(axis=1)
count_by_director_type.sort_values(by="Total", ascending=False, inplace=True)
count_by_director_type.drop("Total", axis=1, inplace=True)

top_directors = count_by_director_type.head(10)

colors = sns.color_palette("muted", n_colors=len(top_directors.columns))

fig, ax = plt.subplots(figsize=(12, 8))

for i, col in enumerate(top_directors.columns):
    ax.barh(top_directors.index, top_directors[col], color=colors[i], label=col)

ax.set_xlabel("Count")
ax.set_ylabel("Director")
ax.set_title("Count of TV Shows and Movies by Director")

ax.legend(loc="upper right")

plt.tight_layout()
plt.show()
```
