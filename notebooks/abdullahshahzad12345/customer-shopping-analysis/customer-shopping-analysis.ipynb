# Data Visualization:

- Is there any relation between numerical columns like Age and purchase amount ?

- Which location generate the Maximum revenue

```python
To10highRevenueLocation = df.groupby('Location')['Purchase Amount (USD)'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(8,5))
sns.barplot(x = To10highRevenueLocation.index , y = To10highRevenueLocation.values)
plt.xticks(rotation=45)
plt.show()
```
