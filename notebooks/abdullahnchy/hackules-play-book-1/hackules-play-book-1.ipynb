### Matrix Algebra in Numpy

Generate 2 matrix and execute sum and multiplication operation between them

```python
# write your code here
mat1 = np.array([[1,2,3],[3,4,5]])
mat2 = np.array([[1,2,3],[2,3,4]])

mat_sum = mat1 + mat2
mat_product = mat1 * mat2

mat_sum
```

```python
mat_product
```

Suppose A and B is two numpy matrix. what is the difference between A*B and np.dot(A,B)

```python
# write your answer as comment
A = mat1
B = mat2
A*B
```

```python
np.dot(A,B)
```

```python
# clearly, 
# A*B is one-one multiplication while np.dot(A,B) is matrix multiplication where the number of rows of the first matrix must equal the number of columns of the second matrix
```
