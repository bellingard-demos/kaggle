# Data preparation

```python
x = df.drop('price', axis=1)
y = df.price
```

```python
X = p.get_dummies(x)
X = X.astype(int)
X
```

```python
# spliting the data for ML
from sklearn.model_selection import train_test_split

x_train , x_test , y_train , y_test = train_test_split( X,y, test_size=0.2, shuffle=True, random_state= 65)

# scalleing 
from sklearn.preprocessing import MinMaxScaler

scaller = MinMaxScaler()

X_train = scaller.fit_transform(x_train)
X_test = scaller.fit_transform(x_test)
```

```python
X_train.shape , X_test.shape
```

# Linear Regression

```python
LR_model = linear_model.LinearRegression()
LR_model.fit(X_train, y_train)
pred = LR_model.predict(x_test)
mae = mean_absolute_error(y_test, pred)
mse = mean_squared_error(y_test, pred)
print('MAE = ',mae)
print('MSE = ',mse)
```

you see , linear regression did'nt work as we thought guys , let's try other algorithms 

# DecisionTreeRegressor

```python
model_tree = DecisionTreeRegressor()
model_tree.fit(X = X_train, y=y_train)
predict = model_tree.predict(X_test)
mae = mean_absolute_error(y_test, predict)
mse = mean_squared_error(y_test, predict)
print("MAE=", mae)
print("RMSE=", np.sqrt(mse))
```

# RandomForestRegressor 

```python
model = RandomForestRegressor(n_estimators=200, random_state=7575)
model.fit(X_train, y_train)
pred_test = model.predict(X_test)
pred_test = model_tree.predict(X_test)
mae = mean_absolute_error(y_test, pred_test)
mse = mean_squared_error(y_test, pred_test)
print("MAE=", mae)
print("RMSE=", np.sqrt(mse))
```
