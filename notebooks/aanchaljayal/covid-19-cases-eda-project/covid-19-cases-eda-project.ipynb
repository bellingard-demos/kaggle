# Exploratory Analysis & Visualization

**1.Top 10 countries of Death Cases**

```python
DeathCaseinCountries = df[['Country/Region','Deaths']].sort_values(by=['Deaths'],ascending=False).head(10)
fig = px.bar(DeathCaseinCountries, x='Deaths', y='Country/Region',color='Country/Region', title='Top 10 Countries of Deaths Cases')
fig.show()
```

**2.Top 10 recovered countries**

```python
plt.style.use('seaborn-poster')
plt.figure(figsize=(15,5))
plt.title("Top 10 Recovered Countries ")     
plt.xlabel("Country/region")                          
df["Recovered"].value_counts().head(10).plot(kind="barh",color="b")      
plt.show()
```

**3.Top 10 confirmed cases by country**

```python
plt.style.use('Solarize_Light2')
plt.figure(figsize=(12,8))
sns.countplot(data=df.head(10),y="Confirmed")                    #Note here we are using countplot as sns which is an in-nuilt library in python that uses mtplotlib underneath to plot graphs.
plt.xticks(rotation='vertical')                                    #Note use of ticks for a better view of diagram.
plt.title('Top 10 Confirmed Cases By Country',fontsize=20)            #Note labelling the data
plt.xlabel('Confirmed Cases')                               
plt.show()
```

**4.Countries in WHO Region**

```python
whoRegionCountries_df = df.groupby(['WHO Region'])[['Country/Region']].count()
whoRegionCountries_df.set_axis(['Total Countries'], axis= 'columns',inplace=True)
whoRegionCountries_df.sort_values(['Total Countries'], ascending=False)
fig = px.pie(whoRegionCountries_df, values='Total Countries',names=whoRegionCountries_df.index, title='Countries in WHO Region', hole= 0.5)
fig.show()
```
