# Exploratory Analysis & Visualization

**1.top 5 channels with most likes to singles video**

```python
df1 = df.groupby('channel_title').count().sort_values('likes',ascending = False)['video_id'].head()
plt.figure(figsize = (8,4))
df1.plot.bar().bar_label(df1.plot.bar().containers[0])
plt.ylim(0,350)
plt.xticks(rotation = 75)
plt.show()
```

**2.Most views for Youtube Channel**

```python
df2 = df.groupby('channel_title').sum().sort_values('views',ascending = False)['views'].head()
plt.figure(figsize = (8,4))
df2.plot.bar().bar_label(df2.plot.bar().containers[0])
plt.xticks(rotation = 75)
plt.ylabel('in Billions')
plt.ylim(0,2000000000)
plt.title('Channels with most views')
plt.show()
```

**3.Most likes for youtube channel**

```python
df3 = df.groupby('channel_title').sum().sort_values('likes',ascending = False)['likes'].head()
plt.figure(figsize = (8,4))
df3.plot.bar().bar_label(df3.plot.bar().containers[0])
plt.ylim(0,60000000)
plt.xticks(rotation = 75)
plt.ylabel('in 10 Millions')
plt.title('Channels with most Likes')
plt.show()
```

**4.Top 5 Channels with most Dislikes**

```python
df4 = df.groupby('channel_title').sum().sort_values('dislikes',ascending = False)['dislikes'].head()

plt.figure(figsize = (8,4))
df4.plot.line(marker = 'o')
plt.ylim(0,8000000)
plt.xticks(rotation = 75)
plt.ylabel('in Millions')
plt.title('Channels with most Dislikes')
plt.show()
```
