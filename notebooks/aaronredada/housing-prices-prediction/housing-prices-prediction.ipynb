## Model Development

```python
x = data ['bedrooms']

y = data ['price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
model = LinearRegression()
```

```python
x_train = x_train.to_numpy().reshape(-1, 1)
y_train = y_train.to_numpy().reshape(-1, 1)
```

```python
model.fit(x_train, y_train)
```

# Model Evaluation
