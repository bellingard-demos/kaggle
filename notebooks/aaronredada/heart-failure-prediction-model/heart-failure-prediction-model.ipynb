## Splitting the Data

```python
x = data [["age", "sex", "chol", "trestbps"]]

y = data ["target"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

## Model Training

```python
model = LogisticRegression()
```

```python
model.fit(x_train.values, y_train.values)
```

## Model Evaluation

```python
heartfailure = model.predict([[40, 1, 246, 131]]) #using numbers for ([[age, sex, cholesterol, and resting blood pressure]])


if heartfailure == 1:
    print("Subject is Likely to have Heart Failure")
elif heartfailure == 0:
    print("Subject is Not likely to have Heart Failure")

```
