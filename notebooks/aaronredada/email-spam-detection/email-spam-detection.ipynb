### Import Libraries

```python
#import of data and analysis
import numpy as np
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt

#library for splitting the data
from sklearn.model_selection import train_test_split

#convert textual data into numerical data
from sklearn.feature_extraction.text import TfidfVectorizer

#to remove stopwords
import nltk

#algorithm for model training
from sklearn.linear_model import LogisticRegression

#testing of model and evaluation
from sklearn.metrics import accuracy_score, confusion_matrix
```

### Import and Analysis of Data

```python
data = pd.read_csv("/kaggle/input/practice/ML/14_naive_bayes/spam.csv") #reads the csv or excel file

print(data)
```

```python
data.shape #dimensions of the data
```

```python
data.describe()
```

```python
x = data["Message"]
y = data["Category"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,train_size=0.8) #splits the data from train and testing
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
feature_extraction = TfidfVectorizer(min_df=1, stop_words='english', lowercase=True)
```

```python
x_train_features = feature_extraction.fit_transform(x_train) #fit_transfom for the train data and transform for the test data

x_test_features = feature_extraction.transform(x_test)
```

```python
x_test_features.shape
```

```python
x_train_features.shape
```

## Model Testing


```python
#Use logistic regression algorithm to train model
```

```python
model = LogisticRegression()
model.fit(x_train_features, y_train)
```

```python
#prediction and testing of the model on training data
```

```python
pred_on_training_data = model.predict(x_train_features)
accuracy_on_training_data = accuracy_score(y_train, pred_on_training_data)

print("Accuracy of the Model on Training Data: ", accuracy_on_training_data)
```

```python
#prediction and testing of the model on testing data
```

```python
pred_on_testing_data = model.predict(x_test_features)
accuracy_on_testing_data = accuracy_score(y_test, pred_on_testing_data)

print("Accuracy of the Model on Testing Data: ", accuracy_on_testing_data)
```
