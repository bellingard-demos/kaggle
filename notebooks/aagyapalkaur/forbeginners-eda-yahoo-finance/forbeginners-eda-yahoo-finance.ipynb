# Import all the necessary libraries

```python
# import the libraries
import pandas as pd
import numpy as np
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
# Read the dataset
dataframe= pd.read_csv("/kaggle/input/yahoo-finance-industries-dataset/yahoo Finanace_industry - Stock Market.csv")
dataframe.head()
```

```python
dataframe.tail()
```

# Check the shape of the dataset

```python
# check the shape of the dataset
dataframe.shape
```

# Check the null Values in the Dataset

```python
# Check is there any null value in the dataset
dataframe.isnull().sum()
```

```python
# check Is there any duplicate value in the dataset
dataframe.duplicated().sum()
```

```python
# Lets check the unique value of market cap
dataframe["market cap"].unique()
```

># Here we notice that market cap is in string format and the values are in Billions, lets convert into float type

```python
dataframe["Market_Cap"]= dataframe["market cap"].str.replace(',',"")
dataframe["Market_Cap"]=dataframe["Market_Cap"].str.replace('B',"").astype(float)
dataframe.drop("market cap", inplace=True, axis=1)
```

```python
# Lets check the Values of Price Column
dataframe["Price"].unique()
```

># Price is also in string format, lets convert into float type

```python
dataframe["Price"]= dataframe["Price"].str.replace(',',"").astype(float)
dataframe["Price"]
```

```python
dataframe["Name"].unique()
```

```python
# Number of unique values in the Name column
dataframe["Name"].nunique()
```

># Here we notice that there are total 44 Unique Company Names

```python
dataframe["Name"]= dataframe["Name"].str.replace(',', "")
dataframe["Name"]
```

```python
dataframe.drop(["Symbol"], axis=1, inplace=True)
```

```python
# gets the dataframe info
dataframe.info()
```

```python
# describe the dataset in terms of statistics
dataframe.describe()
```

# Lets visualize the Correlation Matrix

```python
correlation_matrix = dataframe.drop("Name", axis=1).corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='Reds', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

># Here we observe that Price column has a high corrleation with PE_ratio with the value of 0.78, which means that if the Price of Stock is high, then the Companies earns high Profit.

# Exploratory Data Analysis

> # Univariate Analysis-- Consider only one variable at a time

```python
# Let see the count of Comapny Name Column
plt.figure(figsize=(12,12))
sns.countplot(dataframe, x=dataframe["Name"])
plt.xticks(rotation=90)
plt.show()
```

```python
# Lets see the numerical distribution of Market_Cap
plt.figure(figsize=(5,5))
sns.histplot(dataframe, x=dataframe["Market_Cap"], bins=10, kde=True, color="red")
plt.show()
```

```python
# Lets see the numerical distribution of Change
plt.figure(figsize=(5,5))
sns.histplot(dataframe, x=dataframe["Change"], bins=5,kde=True, color="blue")
plt.show()
```

```python
# Lets see the numerical distribution of PE_ratio
plt.figure(figsize=(5,5))
sns.histplot(dataframe, x=dataframe["PE_ratio"], bins=5,kde=True, color="green")
plt.show()
```

# Above we figure out that there are 15 values missing in the PE Ratio, lets handle the missing values first

```python
dataframe["PE_ratio"].fillna(dataframe["PE_ratio"].median(), inplace=True)
```

```python
plt.figure(figsize=(5,5))
sns.histplot(dataframe, x=dataframe["PE_ratio"], bins=5,kde=True, color="green")
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.histplot(dataframe, x=dataframe["Price"], bins=5,kde=True, color="brown")
plt.show()
```

> # Bivariate Analysis -- Consider two Variable at a time

>1. In the below plots we see the lineplot of Market_Cap and Change in stock price with respect to Companies Names.
2. Market_Cap defines that how big the size of the Comapnies are,in terms of stock/share they had.

```python
# Line graph
column=['Market_Cap','Change']
color=['green', 'blue']
for i in range(len(column)):
    plt.figure(figsize=(10, 6))
    sns.lineplot(data=dataframe, x='Name', y=column[i], color=color[i])
    plt.title(f"Line Plot of {column[i]}")
    plt.xlabel("Company Name")
    plt.xticks(rotation=90)
    plt.ylabel(column[i])
    plt.show()
```

```python
fig = px.scatter(dataframe, x="Market_Cap", y="PE_ratio",color="Name", title="Market_Cap Vs PE_Ratio")
fig.show()
```

> # Here we can observe that as the size of the companies increases their PE_ratio decreases.

```python
fig = px.scatter(dataframe, x="Market_Cap", y="Change",color="Name", title="Market_Cap Vs Change in Stock Price")
fig.show()
```

> # Here we figure out the maximum change in stock price of BlockRock Inc. whose Market_Cap fluctuate between 98.394 and 98.484 Billion. As the value of Market_Cap increase we observe the small change in stock price.

```python
fig = px.scatter(dataframe, x="Market_Cap", y="Price",color="Name", title="Market_Cap Vs Price")
fig.show()
```

> # We notice the similar trend here, as the size of Companies increases their Stock Price value decreases.

```python
fig = px.scatter(dataframe, x="Price", y="PE_ratio",color="Name", title="Price Vs PriceEarning_Ratio")
fig.show()
```

# Here We can see that as the the Price of Stock increases the Comapnies earn high Profit.
