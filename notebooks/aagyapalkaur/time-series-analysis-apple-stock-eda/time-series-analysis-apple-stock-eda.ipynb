# Simple Moving Average:
1. **A Simple Moving Average (SMA) is a useful tool in analyzing stock prices. It helps in smoothing out the price data and identifying trends, making it easier to understand how a stock's price is moving over time.**
2. **In the below code we set the rolling window to 50, which considered the average of prevoius 50 stock prices and move that widow through the Open Stock Price to get the series of Average Prices.**

```python
columns=["Open","Close"]

for col in columns:
    simple_moving_average_stock =dataset[col].rolling(window=100).mean()
    plt.figure(figsize=(6,6))
    simple_moving_average_stock.plot(label="Simple Moving Average", color="blue")
    dataset[col].plot(label=col, color="red")
    plt.xticks(rotation=0)
    plt.title(f"Moving Average of {col} Price", size=15)
    plt.legend()
    plt.show()
```

> **Here we can clearly see that Open, Close stock price and SMA follows each other. So, its shows the sign of upward trend in Open stock Price over the time.**
