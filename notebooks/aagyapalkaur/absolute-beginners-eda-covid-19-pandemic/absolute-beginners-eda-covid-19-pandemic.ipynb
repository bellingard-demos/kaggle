```python
dataframe= pd.read_csv("/kaggle/input/covid-19-coronavirus-pandemic-dataset/Covid Live.csv")
dataframe.head()
```

```python
# Lets drop the columns "#" that just show the id
dataframe.drop("#", axis=1, inplace=True)
```

```python
# Getting the list of all the columns
columns= dataframe.columns.tolist()
columns
```

```python
# Lets make the columns name simple
columns[0]="Country"
columns[1]="Total Cases"
columns[2]="Total Deaths"
columns[3]= "New Deaths"
columns[4]="Total Recovered"
columns[5]="Active Cases"
columns[6]="Serious Cases"
columns[7]="Total Cases/1M pop"
columns[8]="Deaths/1M pop"
columns[9]="Total Tests"
columns[10]="Test/1M pop"
dataframe.columns=columns
dataframe.head()
```

```python
def get_num_cat_columns(dataframe):
    categorical_cols=dataframe.select_dtypes(include="object").columns
    numerical_cols=dataframe.select_dtypes(exclude="object").columns
    
    return categorical_cols, numerical_cols
```

```python
categorical_cols, numerical_cols= get_num_cat_columns(dataframe)
```

          >
     Now,lets detect the Outliers in the Dataset
</p>
</div>

```python
def Winsorization_Method(columns, dataframe, a, b):
    outliers=[]

    for col in columns:
        q1= np.percentile(dataframe[col], a)
        q2= np.percentile(dataframe[col],b)
        
        for pos in range(len(dataframe)):
            if dataframe[col].iloc[pos]>q2 or dataframe[col].iloc[pos]<q1:
                outliers.append(pos) 
                
    outliers= set(outliers)                   # remove the duplicates from the outliers
    outliers= list(outliers)
    
    ratio= round(len(outliers)/len(dataframe)*100, 2)                       # Ratio of outliers
    dataframe.drop(dataframe.index[outliers], inplace=True)    # remove the outliers from the  dataset
    
    
    
    return ratio,dataframe
```

```python
ratio, dataframe= Winsorization_Method(numerical_cols,dataframe,a=0.5, b=99.5)
```

```python
print(f"Ratio of outliers detected in the dataset: {ratio}")
```
