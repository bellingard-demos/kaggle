```python
import pandas as pd
train= pd.read_csv('/kaggle/input/ipba-16-regression-graded-paris-house/train.csv')
test = pd.read_csv("/kaggle/input/ipba-16-regression-graded-paris-house/test.csv")
print(train.shape)
train.head()
```

```python
test.set_index("id",inplace=True)
```

```python
#splitting train data into x(all independent variables) and y(dependent variable)
x=train.drop(["price"],axis=1)
y=train["price"]
x
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
q1=x_train['floors'].quantile(0.25)
q3=x_train['floors'].quantile(0.75)
iqr=q3-q1
lw=q1-1.5*iqr
uw=q3+1.5*iqr
print(lw,uw)
floors_o=x_train.index[(x_train['floors']>uw) | (x_train['floors']<lw)].to_list()
floors_o
```

```python
q1=x_train['squareMeters'].quantile(0.25)
q3=x_train['squareMeters'].quantile(0.75)
iqr=q3-q1
lw=q1-1.5*iqr
uw=q3+1.5*iqr
print(lw,uw)
squareMeters_o=x_train.index[( x_train['squareMeters']>uw) | (x_train['squareMeters']<lw) ].tolist()
squareMeters_o
```

```python
q1=x_train['made'].quantile(0.25)
q3=x_train['made'].quantile(0.75)
iqr=q3-q1
lw=q1-1.5*iqr
uw=q3+1.5*iqr
print(lw,uw)
made_o=x_train.index[ (x_train['made']>uw) | (x_train['made']<lw) ].tolist()
made_o
```

```python
q1=x_train['basement'].quantile(0.25)
q3=x_train['basement'].quantile(0.75)
iqr=q3-q1
lw=q1-1.5*iqr
uw=q3+1.5*iqr
print(lw,uw)
basement_o=x_train.index[ (x_train['basement']>uw) | (x_train['basement']<lw) ].tolist()
basement_o
```

```python
q1=x_train['attic'].quantile(0.25)
q3=x_train['attic'].quantile(0.75)
iqr=q3-q1
lw=q1-1.5*iqr
uw=q3+1.5*iqr
print(lw,uw)
attic_o=x_train.index[ (x_train['attic']>uw) | (x_train['attic']<lw) ].tolist()
attic_o
```

```python
q1=x_train['garage'].quantile(0.25)
q3=x_train['garage'].quantile(0.75)
iqr=q3-q1
lw=q1-1.5*iqr
uw=q3+1.5*iqr
print(lw,uw)
garage_o=x_train.index[ (x_train['garage']>uw) | (x_train['garage']<lw) ].tolist()
garage_o
```

```python
outliers=floors_o+squareMeters_o+made_o+basement_o+attic_o+garage_o
print(len(outliers))
```

```python
x_train_final=x_train.drop(index=outliers).reset_index(drop=True)
y_train_final=y_train.drop(index=outliers).reset_index(drop=True)
print(x_train_final.shape)
print(y_train_final.shape)
```

```python
print(x_train['cityCode'].nunique())
x_train['cityCode'].value_counts()
#dropping city code variable as there are too many unqiue values for this caregorical variable for one hot encoding
x_trainf=x_train.drop(columns='cityCode')
test=test.drop(columns='cityCode')
```

```python
import statsmodels.formula.api as smf
reg1=smf.ols("price~squareMeters+numberOfRooms+hasPool+cityPartRange+made+isNewBuilt+attic+garage+hasStorageRoom+hasYard+floors+numPrevOwners+hasStormProtector+basement+hasGuestRoom", data=x_trainf1)
```

```python
results=reg1.fit()
print(results.summary())
```

```python
reg2=smf.ols("price~squareMeters+numberOfRooms+hasPool+cityPartRange+made+isNewBuilt+attic+garage+hasStorageRoom+hasYard+floors", data=x_trainf1)
results1=reg2.fit()
print(results1.summary())
```

```python
reg3=smf.ols("price~squareMeters+numberOfRooms+garage+hasStorageRoom+floors", data=x_trainf1)
results2=reg3.fit()
print(results2.summary())
```

```python
reg4=smf.ols("price~squareMeters+hasStorageRoom+floors", data=x_trainf1)
results3=reg4.fit()
print(results3.summary())
```

```python
from sklearn.metrics import mean_squared_error,r2_score
predictions=results3.predict(x_trainf1)
mse = mean_squared_error(y_train_final, predictions)
rmse=((mse)**(0.5))
print(mse)
print(rmse)
```

```python
actuals=x_trainf1['price']
plt.plot(predictions,"r")
plt.plot(actuals,"b")
```

```python
residuals=results3.resid
type(residuals)
residualsdf=pd.DataFrame(residuals)
residualsdf.rename(columns={0:"res"},inplace=True)
plt.scatter(residualsdf,predictions)
plt.xticks([])
```

```python
import sklearn.metrics as metrics
mae=metrics.mean_absolute_error(actuals,predictions)
mae
import numpy as np
np.mean(abs((actuals-predictions)/actuals))
```
