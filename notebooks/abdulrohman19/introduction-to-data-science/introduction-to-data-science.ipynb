> **5USER DEFINED FUNCTION**

Sebuah program yang besar biasanya disusun atas sejumlah fungsi. Sebuah fungsi berisi sejumlah pernyataan yang dikemas dengan sebuah nama. Selanjutnya, nama ini dapat dipanggil beberapa kali dalam program dan tentu saja cara ini dapat mengurangi duplikasi kode.

```python
#Mendefinisika fungsi
def maks(a, b):
    "akan muncul terbesar dari a dan b"
    if a > b:
        return print(a)
    else :
        return print(b)
```

```python
maks(0,1)
```

```python
#Argument bernama
def true(a, b, c):
    print(a, b, c)
    
true(a=1, b=2, c=3)
true(c=3, a=1, b=2)
true(b=2, a=1, c=3)
```

```python
#Perulangan
def replay(st, jum = 5):
    return st*jum

print(replay('yes'))
print(replay('yes',1))
```

```python
#Bentuk Argument
def ukuran(*arg):
    maks=arg[0]
    for nilai in arg[1:]:
        if nilai > maks:
            maks = nilai
    return maks
```

```python
ukuran(2000, 546534, 546546, 657465)
```

```python
def tes(m):
    if(m==0) or (m==1):
        return 1
    else :
        return m * tes(m-1)
    
tes(5)
```

Berkaitan dengan operasi file, Python menyediakan kemudahan untuk
melakukan pemanipulasian data pada file, seperti untuk menyimpan data,

tanpa melalui pengimporan modul

```python
def bro_what(do_file):
    y = True
    try:
        bro = open[do_file]
        bro.close()
    except IOError:
        y = false
            
    return y

    print(bro_what("first.ipynb"))
    print(bro_what("None"))
```
