```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("../input/grayscale/gray.jpeg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
img1 = 255 - img
plt.imshow(img1)
```

```python
gamma = 2
img1 = np.power(img, gamma)
plt.imshow(img1)
```

Log Transformation

```python
img1 = np.uint8(np.log1p(img))
```

```python
thresh = 2
img2 = cv2.threshold(img1, thresh, 200, cv2.THRESH_BINARY)[1]
plt.imshow(img2)
```
