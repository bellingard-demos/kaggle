# Membuat kerangka Data/ DataFrame dari tugas DataFrame

```python
import pandas as pd
import numpy as np
mcu_data = {'Names': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'], 'Scores': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19], 'Attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1], 'Quality': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}
df_mcu = pd.DataFrame(mcu_data)
df_mcu
```

#  Menampilkan siswa yang memenuhi syarat berdasarkan skor menggunakan kualifikasi baris'Yes' pada kolom 'Quality'

```python
df_mcu.loc[(df_mcu['Quality'] == 'yes'), :]
```

#  menambahkan kolom baru 'GRADE '

```python
print(df_mcu[df_mcu['Scores'].isnull()])
```

```python
df_mcu['Grade'] = ('SMA 1','SMA 2', 'SMA3','SMA 1','SMA 2', 'SMA3','SMA 1','SMA 2', 'SMA3','SMA3')
df_mcu
```

```python
df_mcu['Grade'] = np.where((df_mcu.Scores < 21 ), 
                  'F', np.where((df_mcu.Scores >= 0) & (df_mcu.Scores <= 11.0), 
                  'D', np.where((df_mcu.Scores >= 13.0) & (df_mcu.Scores <= 13.0), 
                  'C', np.where((df_mcu.Scores >= 14.0) & (df_mcu.Scores <= 12.0), 
                  'B', np.where((df_mcu.Scores >= 13.0) & (df_mcu.Scores <= 14.5), 
                  'A', 'No Marks')))))
df_mcu
```

# menambahkan nama siswa baru bernama jack

```python
mod_df = df_mcu.append({'Names': 'Jack', 'Scores':20, 'Attempts': 1.0 ,'Quality' :'yes', 'Grade':'SMA 1'}, ignore_index=True)
print('Modified DataFrame')
print(mod_df)
```

#  Menambahkan Kolom baru yang bernilai 0 atau 1
