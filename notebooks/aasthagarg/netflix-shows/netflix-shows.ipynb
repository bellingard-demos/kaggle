```python
df = df[df['date_added'].notna()]
```

```python
df['country'] = df['country'].fillna(df['country'].mode()[0])
```

```python
df['country'].mode()[0]
```

```python
df['year_added'] = df['date_added'].apply(lambda x: x.split(" ")[-1])
```

```python
df['year_added'].head()
```

```python
df['month_added']= df['date_added'].apply(lambda x: x.split(" ")[0])
```

```python
df['month_added'].head()
```

```python
ratings_ages = {
    'TV-PG': 'Older Kids',
    'TV-MA': 'Adults',
    'TV-Y7-FV': 'Older Kids',
    'TV-Y7': 'Older Kids',
    'TV-14': 'Teens',
    'R': 'Adults',
    'TV-Y': 'Kids',
    'NR': 'Adults',
    'PG-13': 'Teens',
    'TV-G': 'Kids',
    'PG': 'Older Kids',
    'G': 'Kids',
    'UR': 'Adults',
    'NC-17': 'Adults'
}
```

```python
df['target_ages'] = df['rating'].replace(ratings_ages)
df.head()
```

```python
df['principal_country'] = df['country'].apply(lambda x: x.split(",")[0])
```

```python
df.dtypes
```

```python
df['type'] = pd.Categorical(df['type'])

df['target_ages'] = pd.Categorical(df['target_ages'], categories=['Kids', 'Older Kids', 'Teens', 'Adults'])

df['year_added'] = pd.to_numeric(df['year_added'])
```

```python
df['genre'] = df['listed_in'].apply(lambda x :  x.split(',')) 
df['genre'].head()
```

```python
#Comparison of Tv Shows and Movies
movies_count = df[df.type == "Movie"]
tvshows_count =df[df["type"] == "TV Show"]
```

```python
import seaborn as sns
plt.figure(figsize=(10,8))
sns.set_style("white")
sns.countplot(x="type", data=df, palette="viridis")
```

```python
#Movies and TV Shows Content Comparison
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, plot, init_notebook_mode, iplot
init_notebook_mode(connected=True)
colors = ["#809fff", "#66ff66"]
trace = go.Pie(labels=["Movies", "Tv Show"], 
               values=[movies_count.type.count(), tvshows_count.type.count()],
               hoverinfo="label+percent", textinfo="label+percent", marker=dict(colors=colors,
                line=dict(color="#2d2d2d", width=2)))

fig = go.Figure(data=[trace])
iplot(fig)
```

```python
#Content In Different Countries
df_counts = df.country.value_counts()
df_counts
```

```python
trace = go.Bar(x=df_counts.index[:15], y=df_counts, marker=dict(
                  opacity=0.8,
                  color=np.arange(15)
              ))
fig = go.Figure(data=[trace])

fig.update_layout(title="Top 15 Countries by Content")
fig.update_xaxes(title="Country")
fig.update_yaxes(title="Count")

iplot(fig)
```

We can observe that netflix produces most of it's content in USA followed by India and UK

```python
#Movies released by directors
df_director =df.dropna(subset=["director"])
director_counts = df_director.director.value_counts()
director_counts
```

```python
trace = go.Pie(labels=director_counts.index[:10], values=director_counts,
               hoverinfo="label+percent", textinfo="label")

fig = go.Figure(data=[trace])
fig.update_layout(title="Director by % Content")

iplot(fig)
```

```python
#Year wise content released on netflix
year_wise_content = df.release_year.value_counts().index[:20]
year_wise_content
plt.figure(figsize=(12,10))
sns.set_style("dark")
sns.countplot(data=df, y="release_year", order=year_wise_content, palette="viridis")
```

```python
#Average Movie Duration
movies_count['duration']=movies_count['duration'].str.replace(' min','')
movies_count['duration']=movies_count['duration'].astype(str).astype(int)
movies_count['duration']
# movie duration distribution
plt.figure(figsize=(10,8))
sns.set_style("darkgrid")
sns.kdeplot(data=movies_count["duration"], shade=True)
```

We can observe that the maximum movies on netflix was of duration 80-100 minutes
