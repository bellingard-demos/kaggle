```python
import matplotlib.pyplot as plt
```

```python

class ChessBoard:
    board = []
    white = (1,1,1)
    blue_position = []
    red_position = []
    """
    creates a chess board by creating a matrix of black cells from the np library then changes the color of some cells to white.
    """
    def __init__(self):
        ChessBoard.board = np.zeros((8, 8, 3))
        for i in range(8):
            for j in range(8):
                if (i%2!=j%2):
                    ChessBoard.board[i][j] = ChessBoard.white
    """
    changes the color of a given (row and column) cell to blue.
    """                
    def color_blue(self,row, col):
        ChessBoard.blue_position = [row,col]
        ChessBoard.board[row, col] = (0, 0, 1)
    """
    changes the color of a given (row and column) cell to red.
    """  
    def color_red(self,row, col):
        ChessBoard.red_position = [row,col]
        ChessBoard.board[row, col] = (1,.2,0)
    """
    renders the chess board with the original positions of the queens in red and blue.
    """      
    def render(self):
        ChessBoard.board[7, 3] = (1,.2,0)
        ChessBoard.board[0, 3] = (0, 0, 1)
        plt.imshow(ChessBoard.board)
    """
    returns True," red is under attack" when the queens are aligned horizontally,verticaly or diagonally and returns False when the said cases 
    did not happen
    """    
    def is_under_attack(self):
        if ChessBoard.blue_position[0] == ChessBoard.red_position[0]:
            return True," red is under attack"
        if ChessBoard.blue_position[1] == ChessBoard.red_position[1]:
            return True," red is under attack"
        if abs(ChessBoard.blue_position[1] - ChessBoard.red_position[1]) == abs(ChessBoard.blue_position[0] - ChessBoard.red_position[0]):
            return True," red is under attack"
        return False
        
```

```python
horizontally_alligned = ChessBoard()
horizontally_alligned.color_blue(1,1)
horizontally_alligned.color_red(1,6)
print(horizontally_alligned.is_under_attack())
plt.imshow(horizontally_alligned.board)
plt.show()
```
