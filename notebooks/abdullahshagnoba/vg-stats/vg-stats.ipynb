## For the top-selling game of all time, how many standard deviations above/below the mean are its sales for North America?

```python
max_Global_Sales = subset_data_frame["Global_Sales"].max()
row_for_max_global_sales = subset_data_frame[subset_data_frame["Global_Sales"]==max_Global_Sales]
max_NA_sales_obj = row_for_max_global_sales["NA_Sales"]
max_NA_sales_value = max_NA_sales_obj[0]
max_NA_sales_value
mean_sales_NA = subset_data_frame["NA_Sales"].mean()
mean_sales_NA
std_for_sales_NA = subset_data_frame["NA_Sales"].std()
std_for_sales_NA
result = (max_NA_sales_value-mean_sales_NA)/std_for_sales_NA
result
```

## The Nintendo Wii seems to have outdone itself with games. How does its average number of sales compare with all of the other platforms?

```python
filter_platforms = subset_data_frame.groupby("Platform")["Global_Sales"]
filter_platforms
average_of_Wii = filter_platforms.mean()["Wii"]
# print(average_of_Wii)
average_of_averages = filter_platforms.mean().mean()
average_of_averages
if average_of_Wii>average_of_averages:
    print("the average of Wii is biger than the averages of other platforms")
else :
    print("the average of Wii is less than the averages of other platforms")
```

## Come up with 3 more questions that can be answered with this data set.
1. What region has the highest average sales?
2. What is the game with the highest global sales?
3. What is the least common game publisher based on global sales

#### 1. What region has the highest average sales?

```python
mean_sales_NA = subset_data_frame["NA_Sales"].mean()
# print(mean_sales_NA)
mean_EU_Sales = subset_data_frame["EU_Sales"].mean()
# print(mean_EU_Sales)
mean_JP_Sales = subset_data_frame["JP_Sales"].mean()
# print(mean_JP_Sales)
if mean_sales_NA > mean_EU_Sales and mean_sales_NA > mean_JP_Sales:  
    print("The region of highest sales is NA_Sales:",max(mean_sales_NA,mean_EU_Sales,mean_JP_Sales))
if mean_EU_Sales > mean_sales_NA and mean_EU_Sales > mean_JP_Sales:  
    print("The region of highest sales is EU_Sales:",max(mean_sales_NA,mean_EU_Sales,mean_JP_Sales))   
if mean_JP_Sales > mean_sales_NA and mean_JP_Sales > mean_EU_Sales:  
    print("The region of highest sales is JP_Sale:",max(mean_sales_NA,mean_EU_Sales,mean_JP_Sales))    
```
