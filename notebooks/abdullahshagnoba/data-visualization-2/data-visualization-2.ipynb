# Second data set

## My insight for this data set is to compare between artists and songs by the popularity.
### 1. First we will compare the songs by thier popularity with a line plot.
### 2. Then we will compare the artists by thier popularity with a bar plot.

```python
df=pd.read_csv("/kaggle/input/eurovision-2023-betting-odds/data/spotify/2023-04-13-spotify-streaming.csv")
df.head()
```

# Top ten popular songs

```python

plt.figure(figsize=(12, 12))
sns.lineplot(x="popularity", y="song",data=df.head(10) ,  color='blue')
plt.title('Top ten popular songs')
plt.xlabel('popularity')
plt.ylabel('song')
plt.show()
```

```python
top_artists = df[['artist', 'popularity']]
top_artists
plt.figure(figsize=(10, 6))
sns.barplot(x='popularity', y='artist', data=top_artists.head(10), palette='viridis')
plt.title('top_artists')
plt.xlabel('popularity')
plt.ylabel('artist')
plt.show()
```
