## Here is the Quick Analysis of the Video Game sales dataset.


```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
VGsales = pd.read_csv("/kaggle/input/videogamesales/vgsales.csv")
```

# Perform data preparation & cleaning

```python
VGsales.head()
```

```python
VGsales.describe()
```

```python
VGsales.describe(include = ['O'])
```

```python
VGsales.info()
```

> #### We have null value in Year, Publishers. 

```python
VGsales.isnull().sum()
```

> #### There 16598 rows. So removing the null values wont affect the dataset. 

```python
VGsales1 = VGsales.dropna()
VGsales1.shape
```

```python
VGsales1.isnull().any()
```

>#### Since we have just 4 records in total for the year 2017 & 2020. We are removing it. 

```python
VGsales1 = VGsales1[VGsales['Year'] != 2020]
VGsales1 = VGsales1[VGsales['Year'] != 2017]
VGsales1.shape
```

# Perform exploratory analysis & visualization

> ### *How is publishing of video games in the market change over years?*

```python
year_trend = VGsales1.Year.value_counts().to_frame().reset_index().rename(columns={'index': 'Year', 'Year':'Count'})
year_trend.head()
plt.figure(figsize = (10, 6))
sns.barplot(data = year_trend, x = 'Year', y = 'Count')
plt.xticks(rotation=90);
```

****Observation :**** Upon analysing the year,Large number of Video game are published on 2009.

> ### *Comparative study about the sales in JP, NA, EU and other places as the years passby*

```python
JP_Y = VGsales1.groupby(['Year']).sum()['JP_Sales'].to_frame().reset_index()
Other_Y = VGsales1.groupby(['Year']).sum()['Other_Sales'].to_frame().reset_index()
EU_Y = VGsales1.groupby(['Year']).sum()['EU_Sales'].to_frame().reset_index()
NA_Y = VGsales1.groupby(['Year']).sum()['NA_Sales'].to_frame().reset_index()

sns.lineplot(data = JP_Y, x = 'Year',y = 'JP_Sales')
sns.lineplot(data = Other_Y, x = 'Year',y = 'Other_Sales')
sns.lineplot(data = EU_Y, x = 'Year',y = 'EU_Sales')
sns.lineplot(data = NA_Y, x = 'Year',y = 'NA_Sales')
plt.legend(['JP', 'Other', 'EU', 'NA']);
```

 ****Observation:**** Sales in NA is largest. 

> ### *Comparative study about the sales of platform in JP, NA, EU and other places*

```python
JP_P = VGsales1.groupby(['Platform']).sum()['JP_Sales'].to_frame().reset_index()
Other_P = VGsales1.groupby(['Platform']).sum()['Other_Sales'].to_frame().reset_index()
EU_P = VGsales1.groupby(['Platform']).sum()['EU_Sales'].to_frame().reset_index()
NA_P = VGsales1.groupby(['Platform']).sum()['NA_Sales'].to_frame().reset_index()
Global_P = VGsales1.groupby(['Platform']).sum()['Global_Sales'].to_frame().reset_index()


sns.barplot(data = Global_P, y = 'Platform',x = 'Global_Sales', order =  Global_P.sort_values(by = ['Global_Sales'], ascending  = False)['Platform'][:10]);

fig, ax = plt.subplots(2,2, figsize = (10,10))
sns.barplot(ax = ax[0,0],data = Other_P, y = 'Platform',x = 'Other_Sales', order =  Other_P.sort_values(by = ['Other_Sales'], ascending  = False)['Platform'][:10]);
sns.barplot(ax = ax[0,1],data = JP_P, y = 'Platform',x = 'JP_Sales', order =  JP_P.sort_values(by = ['JP_Sales'], ascending  = False)['Platform'][:10]);
sns.barplot(ax = ax[1,0],data = EU_P, y = 'Platform',x = 'EU_Sales', order =  EU_P.sort_values(by = ['EU_Sales'], ascending  = False)['Platform'][:10]);
sns.barplot(ax = ax[1,1],data = NA_P, y = 'Platform',x = 'NA_Sales', order =  NA_P.sort_values(by = ['NA_Sales'], ascending  = False)['Platform'][:10]);
```

****Observation:****
* Though DS release large number of Video game. PS2 has huge global sales market.
* Superingly DS has good market in japan which constitute less fraction of sale in total global sale.

> ### *Comparative study about the sales in JP, NA, EU and other places with respect to genre*

```python
JP_G = VGsales1.groupby(['Genre']).sum()['JP_Sales'].to_frame().reset_index()
Other_G = VGsales1.groupby(['Genre']).sum()['Other_Sales'].to_frame().reset_index()
EU_G = VGsales1.groupby(['Genre']).sum()['EU_Sales'].to_frame().reset_index()
NA_G = VGsales1.groupby(['Genre']).sum()['NA_Sales'].to_frame().reset_index()
Global_G = VGsales1.groupby(['Genre']).sum()['Global_Sales'].to_frame().reset_index()

sns.barplot(data = Global_G, y = 'Genre',x = 'Global_Sales', order =  Global_G.sort_values(by = ['Global_Sales'], ascending  = False)['Genre'][:5]);

fig, ax = plt.subplots(2,2, figsize = (10,10))
sns.barplot(ax = ax[0,0], data = JP_G, y = 'Genre',x = 'JP_Sales', order =  JP_G.sort_values(by = ['JP_Sales'], ascending  = False)['Genre'][:5]);
sns.barplot(ax = ax[0,1], data = Other_G, y = 'Genre',x = 'Other_Sales', order =  Other_G.sort_values(by = ['Other_Sales'], ascending  = False)['Genre'][:5]);
sns.barplot(ax = ax[1,0], data = EU_G, y = 'Genre',x = 'EU_Sales', order =  EU_G.sort_values(by = ['EU_Sales'], ascending  = False)['Genre'][:5]);
sns.barplot(ax = ax[1,1], data = NA_G, y = 'Genre',x = 'NA_Sales', order =  NA_G.sort_values(by = ['NA_Sales'], ascending  = False)['Genre'][:5]);

```

****Observation:**** Globally people prefer Action, Sports and shooter. 

> ### *Comparative study about the sales in JP, NA, EU and other places for different publishers*

```python
JP_Pl = VGsales1.groupby(['Publisher']).sum()['JP_Sales'].to_frame().reset_index()
Other_Pl = VGsales1.groupby(['Publisher']).sum()['Other_Sales'].to_frame().reset_index()
EU_Pl = VGsales1.groupby(['Publisher']).sum()['EU_Sales'].to_frame().reset_index()
NA_Pl = VGsales1.groupby(['Publisher']).sum()['NA_Sales'].to_frame().reset_index()
Global_Pl = VGsales1.groupby(['Publisher']).sum()['Global_Sales'].to_frame().reset_index()

sns.barplot(data = Global_Pl, y = 'Publisher',x = 'Global_Sales', order =  Global_Pl.sort_values(by = ['Global_Sales'], ascending  = False)['Publisher'][:5]);

fig, ax = plt.subplots(2,2, figsize = (12,12))
sns.barplot(ax = ax[0,0], data = JP_Pl, y = 'Publisher',x = 'JP_Sales', order =  JP_Pl.sort_values(by = ['JP_Sales'], ascending  = False)['Publisher'][:5]);
sns.barplot(ax = ax[0,1], data = Other_Pl, y = 'Publisher',x = 'Other_Sales', order =  Other_Pl.sort_values(by = ['Other_Sales'], ascending  = False)['Publisher'][:5]);
sns.barplot(ax = ax[1,0], data = EU_Pl, y = 'Publisher',x = 'EU_Sales', order =  EU_Pl.sort_values(by = ['EU_Sales'], ascending  = False)['Publisher'][:5]);
sns.barplot(ax = ax[1,1], data = NA_Pl, y = 'Publisher',x = 'NA_Sales', order =  NA_Pl.sort_values(by = ['NA_Sales'], ascending  = False)['Publisher'][:5]);
```

****Observation:**** Nintendo captures the global market though it publish less video games compared to top 5 publishers.

> ### *What are the Top 10 platform selling highest number of video games?*

> ### *What are the Top genre of the video games being published?*

```python
sns.countplot(y = VGsales.Genre, order = VGsales.Genre.value_counts().index[:] );
```

> ### *Who are the Top 10 publishers publishing highest number of video games?*

```python
sns.countplot(y = VGsales.Publisher, order = VGsales.Publisher.value_counts().index[:10] );
```

**As Ninento is top publisher lets analyze it further.** 

> ### *What are the top 3 most genre of Nintendo ?*

```python
NVG = VGsales1.groupby(['Publisher', 'Genre']).count().reset_index()[['Publisher','Genre', 'Rank']].rename(columns={'Rank':'Count'})
Action = NVG[NVG['Publisher'] == 'Nintendo']
sns.barplot(data = Action, y = 'Genre',x = 'Count', order =  Action.sort_values(by = ['Count'], ascending  = False)['Genre'][:3]);
```

> ### *What are the top 3 most publishing platform for Nintendo?*

```python
NVG = VGsales1.groupby(['Publisher', 'Platform']).count().reset_index()[['Publisher','Platform', 'Rank']].rename(columns={'Rank':'Count'})
Publisher = NVG[NVG['Publisher'] == 'Nintendo']
sns.barplot(data = Publisher, y = 'Platform',x = 'Count', order =  Publisher.sort_values(by = ['Count'], ascending  = False)['Platform'][:3]);
```
