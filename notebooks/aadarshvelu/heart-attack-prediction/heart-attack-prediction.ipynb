```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
pipeline = make_pipeline(StandardScaler(), SVC())

cross_val_score(pipeline, dataset.iloc[:, :-1], dataset.iloc[:, -1], cv = 10).max()
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.svm import LinearSVC, NuSVC, SVC
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import roc_auc_score, accuracy_score, classification_report
x_train, x_test, y_train, y_test = train_test_split(dataset.iloc[:, :-1], dataset.iloc[: , -1], train_size = 0.7)

abc = AdaBoostClassifier()
```

```python
hyper_parameters = {
    "n_estimators": [50, 100, 150, 200],
    "learning_rate": [1, 1.5, 2, 2.5],
}

gscv = GridSearchCV(abc, hyper_parameters, verbose = 1)
gscv.fit(x_train, y_train)
gscv.best_params_
```

```python
print (classification_report(y_test, gscv.predict(x_test)))
```
