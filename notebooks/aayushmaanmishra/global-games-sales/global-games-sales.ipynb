```python
ax = df.groupby('Genre')['Global_Sales'] \
     .agg(['mean','count']) \
     .sort_values('mean')['mean'] \
     .plot(kind='barh' , figsize=(12,5), title='avg global sales by platform')
ax.set_xlabel('Avg Global Sales')
plt.show
```
