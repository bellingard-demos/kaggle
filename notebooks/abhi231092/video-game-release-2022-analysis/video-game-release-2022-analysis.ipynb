```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#reading the dataset
df = pd.read_csv('/kaggle/input/video-games-released-in-2022/video-games-2022.csv')
df.head()
```

Since we have the data available now, lets look if the data has any anamoly

```python
df.info()
```

There are no missing values. We can proceed further for analysis

```python
months = {'JANUARY':1,'FEBRUARY':2,'MARCH':3,'APRIL':4,'MAY':5,'JUNE':6,'JULY':7,'AUGUST':8,'SEPTEMBER':9,'OCTOBER':10,'NOVEMBER':11,'DECEMBER':12}
df['Month'] = df.Month.map(months)
df.Month.value_counts().sort_index().plot(kind='bar')
```

<b>Obervation 1:</b> It is clear from the above graph that most games are released during May, June, July, August, September during 2022

```python
df.Month.value_counts()*100/df.Month.count()
```

```python
df.Day.value_counts().sort_index().plot(kind='line')
```

It is visible in the above graph that there is a huge spike in new game releases during the end of the month but this could be totally random also and we don't have enough data to support this spike.

```python
#now lets check game releases as per the platform.
plt.figure(figsize=(15,6))
platforms = df['Platform(s)'].apply(lambda x:[i.strip() for i in x.split(',')]).explode().value_counts()
platforms.plot(kind='bar')
plt.show()
```

<b>Observation 2:</b> We can see that Nintendo Switch, Windows, PS4, PS5, XBOX and XSX are the major platforms for which gaming companies released the games in 2022

```python
#lets check genres for released games
plt.figure(figsize=(15,6))
genre = df['Genre(s)'].apply(lambda x:[i.strip() for i in x.split(',')]).explode().value_counts()
genre = pd.Series(genre).where(lambda x:x>10).dropna()
genre.plot(kind='bar')
```

<b>Observation 3:</b> Clearly, adventure, action role-playing, role-playing, action-adventure,visual novel, action, puzzle and first preson shooter games are more prevalent among all the released games

```python
plt.figure(figsize=(15,6))
developers = df['Developer(s)'].apply(lambda x:[i.strip() for i in x.split(',')]).explode().value_counts()
developers = pd.Series(developers).where(lambda x:x>5).dropna()
developers.plot(kind='bar')
plt.show()
```

<b>Observation 4: </b>Most of the games are developer by Square Enix, Nippon, Konami, Mages, Inti Creates, Idea Factory and capcom

```python
plt.figure(figsize=(15,6))
pd.Series(df['Publisher(s)'].value_counts()).where(lambda x:x>5).dropna().plot(kind='bar')
plt.show()
```

<b>Observation 5: </b> Most of the games were published by Square Enix, Bandai Namco, NIS, Devolver Digital, Idea Factory, Nacon, Konami, PQube, Spike Chunsift and some more
