```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#view dataset
df_netflix = pd.read_csv('/kaggle/input/netflix-shows/netflix_titles.csv')
df_netflix.shape
```

#### How many movies and TV shows on Netflix.

```python
df_netflix['type'].value_counts().plot(kind='bar')
```

#### Which year netflix has a surge in number of releases

```python
surges = df_netflix.loc[df_netflix['release_year']>2010,['release_year']].astype('int')
```

```python
surges.groupby('release_year')['release_year'].count().plot(kind='line',figsize=(20,8))
```

```python
#Graph shows the surge in 2018
```

#### which country has how many movies or shows in netflix

```python
df_netflix_temp = df_netflix.copy()
df_netflix_temp['country'] = df_netflix_temp['country'].str.split(',')
df_netflix_temp.explode('country')
```

```python
df_netflix_temp['country'].value_counts()[:5].plot(kind='bar')
```

#### On which day of the week mostly movies/shows releases on netflix

```python
df_netflix['date_added']= pd.to_datetime(df_netflix['date_added'])
df_netflix['day_of_week'] = df_netflix['date_added'].dt.dayofweek
```

```python
df_netflix
```

```python
#convert week day numbers to week day text
def convert_day(x):
    if x==-1:
        return None
    dic = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
    return dic[int(x)]

df_netflix['day_of_week'] = df_netflix['day_of_week'].fillna(-1)
df_netflix['day_of_week'] = df_netflix['day_of_week'].apply(convert_day)
```

```python
df_netflix['day_of_week'].value_counts().plot(kind='bar')
```

```python
#Most of the movies are released on fridays, which is convincing that friday is LWD of the week.
```
