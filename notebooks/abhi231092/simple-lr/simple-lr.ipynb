Simple Linear Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
from sklearn.model_selection import train_test_split
import statsmodels.api as sm
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
#read dataset
df = pd.read_csv('../input/random-linear-regression/train.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
plt.scatter(df['x'],df['y'])
plt.show()
```

```python
sns.regplot(x='x',y='y',data = df)
```

```python
df.corr()
```

```python
sns.heatmap(df.corr(),annot=True)
```

```python
#split the train and test data
df = df.dropna(subset=['y'],axis=0)
x_train,x_test,y_train,y_test = train_test_split(df['x'],df['y'],train_size=0.7)
```

```python
x_train.shape
```

```python
#add constant to the train data
x_train_sm = sm.add_constant(x_train)
x_train_sm
```

```python
model = sm.OLS(y_train,x_train_sm)
lr_model = model.fit()
```

```python
lr_model.params
```

```python
lr_model.summary()
```

```python
y_train_pred = lr_model.predict(x_train_sm)
```

```python
plt.scatter(x_train,y_train)
plt.plot(x_train ,y_train_pred)
plt.show()
```

```python
#residual analysis
res = y_train - y_train_pred
```

```python
plt.scatter(x_train,res)
#residuals does not show any pattern
```

```python
sns.distplot(res)
```

#### residuals follow a normal distribution

```python
#predictions on test data
x_test_sm = sm.add_constant(x_test)
y_test_pred = lr_model.predict(x_test_sm)
```

```python
plt.figure()
plt.scatter(x_test,y_test)
plt.plot(x_test,y_test_pred,'y')
```

```python
r2 = r2_score(y_test,y_test_pred)
```

```python
r2
```

The model performs very well on test data too.
