# Student's T Distribution 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/kaggle/input/data-science-job-salaries/ds_salaries.csv')
df.head()
```

```python
#as we are internest in salaries lets get salaries out of the dataset
salaries = df['salary']
salaries.shape
```

<b>There are 607 salaries, lets assume this is our population as we want to prove Student T's Distribution with various confidence intervals</b>

```python
sns.displot(salaries)
```

This is definelty not a normal distribution or even T distribution. but CLT helps us to get a normal distribution here. So we assume that sampling distribution will give us a normal distribution.
<br>
<br>
#### Take out a sample of 20 observations

```python
sample_salaries = salaries.sample(20)
```

```python
#lets check out population mean 
print("Population mean(salary) = " + str(round(salaries.mean(),2)))
```

```python
import math
#lets check out mean salary of sample
print("Sample mean(salary) = " + str(round(sample_salaries.mean(),2)))
print("Sample Std. Dev.(salary) = " + str(round(sample_salaries.std(),2)))
print("standard error = " + str(round(sample_salaries.std()/math.sqrt(20),2)))
```

Although it a random sample, but the sample size is too small which gives us totally misleading figure of average salary of population.<br>
<b>We can not use z-statistics here as CLT can be implemented only when we have sample size greater than 30. So we will use t-statistics.</b>

T-statistics formula = T(n-1,Aplha) = (x(hat) - Xmean)/std. Error

```python
import scipy.stats
#95% confidence level
t_score_95 = scipy.stats.t.ppf(q=1-.05/2,df=20-1)
#99% confidence level
t_score_99 = scipy.stats.t.ppf(q=1-.01/2,df=20-1)
```

```python
sample_mean = round(sample_salaries.mean(),2)
```

Intervals

```python
print('Intervals for mean salary of population with 95% confidence')
print([sample_mean - t_score_95*round(sample_salaries.std()/math.sqrt(20),2),sample_mean + t_score_95*round(sample_salaries.std()/math.sqrt(20),2)])
```

```python
print('Intervals for mean salary of population with 99% confidence')
print([sample_mean - t_score_99*round(sample_salaries.std()/math.sqrt(20),2),sample_mean + t_score_99*round(sample_salaries.std()/math.sqrt(20),2)])
```

```python
salaries.mean()
```

<b>Clearly, for both 95% and 99% confidence level, we get the population mean salary between the corresponding intervals.</b>
