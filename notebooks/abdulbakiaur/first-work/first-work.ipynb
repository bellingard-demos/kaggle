In this work, we look at the Video Game Sales dataset. First, we need to get the file and examine.

```python
df = pd.read_csv("/kaggle/input/videogamesales/vgsales.csv") # read the dataset
df.head(10) # look at the first ten rows
```

```python
df.info()
```

```python
df.columns # columns of the dataset
```

```python
# correlation map
f,ax = plt.subplots(figsize=(15, 14))
sns.heatmap(df.corr(), annot=True, linewidths=.5, fmt= '.1f',ax=ax)
plt.show()
```

As you may see, the North America sales (NA_Sales) and the global sales (Global_Sales) are positively correlated and also the European sales (EU_Sales) and Global_Sales are positively correlated and both correlations are close to 1.

Now, we look at the total global sales of each year by using line plot.

```python
df.plot(kind = 'line', x = 'Year', y = 'Global_Sales', color = 'blue', label = 'Global', linewidth=1, alpha = 1, grid = True)
plt.xlabel('Year') # name of the x-axis
plt.ylabel('Sales (millions)') # name of the y-axis
plt.title('Global Video Game Sales by Year') # title of plot
plt.show()
```

After 1985, the sales were fairly increased.

We compare NA_Sales and EU_Sales with Global_Sales by using scatter plot.

```python
plt.scatter(df.NA_Sales,  df.Global_Sales, color = 'green', label = "NA_Sales")
plt.scatter(df.EU_Sales, df.Global_Sales, color = 'blue', label = "EU_Sales")
plt.legend()
plt.xlabel("NA_Sales & EU_Sales")
plt.ylabel('Global_Sales')
plt.title('NA_Sales & EU_Sales vs. Global_Sales')
plt.show()
```

We see that most of the sales are made to North America.

We examine the distribution of EU_Sales 

```python
# Histogram
df.EU_Sales.plot(kind = 'hist',bins = 5, figsize = (12,12))
plt.show()
```

Now, we add a column that separates the game which are released before or after 2000. Then, we filter the games for which they have more than 20 millions of global sales.

```python
# add a column called 'Before_After_2000'
df["Before_After_2000"] = ["Before 2000" if(each < 2000) else "After 2000" for each in df.Year]
```

```python
df1 = df['Global_Sales'] > 20 # filter the Global_Sales
df[df1]
```

As we see below, there are 14 games released after 2000.

```python
plt.hist(df[df1].Before_After_2000, bins = 3)
plt.show()
```
