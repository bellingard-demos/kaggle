**Type conversion** - Convert one data type to another

**Implicit**

```python
Revenue_A=20
type(Revenue_A)
```

```python
Revenue_B=20.56
type(Revenue_B)
```

```python
Revenue = Revenue_A + Revenue_B

print(Revenue)
type(Revenue)
```

```python
print("Total revenueRevenue", Revenue)
```

**Explicit conversion**

```python
a=input("Store A sales")
b=input("Store B sales")
c=a+b       #concatinating a and b so the result will be string
print(c)
type(c)
```

```python
d=float(a) + float(b)
print(d)
type(d)
```
