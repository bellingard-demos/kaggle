**Mask filtering**

```python
import pandas as pd
```

```python
a=pd.read_csv('/kaggle/input/novel-corona-virus-2019-dataset/covid_19_data.csv')
a.head()
```

```python
a.info()
```

```python
a.info()
```

**isin() Method**
it is used to check the presence of values

compare multiple values

```python
a=pd.read_csv('/kaggle/input/novel-corona-virus-2019-dataset/covid_19_data.csv')
a.head()
```

```python
mask=a['Country/Region'].isin(['India','Japan'])
mask
```

```python
a[mask]
```

**without isna()**

```python
mask1=a['Country/Region']=='India'
mask2=a['Country/Region']=='Japan'
a[mask1 | mask2]
```

**NULL filtering**

isnull() method returns a boolean series True in case we have a null record**

```python
import pandas as pd
a=pd.read_csv('/kaggle/input/novel-corona-virus-2019-dataset/covid_19_data.csv')
a['ObservationDate']=pd.to_datetime(a['ObservationDate'])
a['Last Update']=pd.to_datetime(a['Last Update'])
a.head()
```

```python
a.info()
```

```python
mask1=a['Province/State'].isnull() #isna() and isnull() are same
```

```python
a[mask1]
```

```python
mask2=a['Province/State'].notnull()
a[mask2]
```

**Use between() method**

```python
import pandas as pd
a=pd.read_csv('/kaggle/input/novel-corona-virus-2019-dataset/covid_19_data.csv')
a.head()
```

```python
mask=a['Confirmed'].between(19,20)
mask
```

```python
a[mask]
```

```python
mask2=a['ObservationDate'].between('01/22/2020','01/23/2020')
a[mask2]
```

**Finding duplicate values**

```python
d={"students":['akash','vikas','akash','swati','vishal','sanjana'],
  "Marks":[22,23,22,24,25,27],
  "Age":[23,17,23,23,22,17]}
a=pd.DataFrame(d)
a
```

```python
a['students'].duplicated()
```

```python
a['students'].duplicated(keep='last')
```

```python
a['Age'].duplicated()
```

```python
m2=a['students'].duplicated(keep=False)
a[m2]
```

```python
m1=a['students'].duplicated(keep='last')
a[m1]
```

**Subset argument**

```python
a.duplicated(subset=['students','Marks','Age'])
```

```python
m3=a.duplicated(subset=['students','Marks','Age'])
a[m3]
```
