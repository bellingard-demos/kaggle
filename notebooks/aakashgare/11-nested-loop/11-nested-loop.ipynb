**Nested Loop**               



```python
i=1
while i<=4:
    j=1
    while j<=3:
        print(j, end=" ")
        j=j+1
    i=i+1
    print("\n")
   
    
```

**Print a pattern**

```python
i=1
while i<=5:
    j=1
    while j<=i:
        print("*",end="")
        j=j+1
    i=i+1
    print("\n")
```

```python
i=5
while i<=5:
    j=1
    while j<=i:
        print("*",end="")
        j=j+1
    i=i-1
    print("\n")
    if i==0:
        break
```

```python

for i in range(1,6):
    j=1
    while j<=i:
        print("*",end="")
        j=j+1
    i=i+1
    print("\n")
```

```python
for i in range(5,0,-1):
    j=1
    while j<=i:
        print("*",end="")
        j=j+1
    i=i-1
    print("\n")
```

**Prime Numbers**
prime number is a number that is divisible by 1 or itslef

```python
i=2
while i<=50:    #range of number which we have to find whether they are prime or not
    j=2
    while j<=i//2+1:      #max multiple will check
        if i==2:
            print(i)
        if i%j==0:
            break
        if j==i//2+1:
            print(i)
        j=j+1
    i=i+1
```

```python
n=int(input("enter the number"))
i=2
while i<=n:    #range of number which we have to find whether they are prime or not
    j=2
    while j<=i//2+1:      #max multiple will check
        if i==2:
            print(i)
        if i%j==0:
            break
        if j==i//2+1:
            print(i)
        j=j+1
    i=i+1
```

```python
n=int(input("enter the number"))
for i in range (1,n):   #range of number which we have to find whether they are prime or not
    j=2
    while j<=i//2+1:      #max multiple will check
        if i==2:
            print(i)
        if i%j==0:
            break
        if j==i//2+1:
            print(i)
        j=j+1
    i=i+1
```
