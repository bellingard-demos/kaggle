**split() method**

```python
a={"Name":["Ajau kumar Jain","Rishi","Aftab Siddique","Rakshit Vig", "Shreya Kumar Jain"],
    "Age":[23,32,12,45,23]
  }
b=pd.DataFrame(a)
b
```

```python
x=b['Name'].str.split(" ",expand=True)
x
```

```python
b[["First name","Second name", "Third name"]]=x
```

```python
b
```

if you want to control the splits then you can n=2 or 3 or 4...

```python
x=b['Name'].str.split(" ",expand=True,n=1)
x
```
