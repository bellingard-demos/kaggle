**Timedelta**-Used to find difference between datetime. It is also used to add/subtract days/hours/minutes from datetime

```python
import datetime 
```

**Add/subtract**

```python
a=datetime.datetime(2021, 12, 3, 5, 34, 56)
b=a + datetime.timedelta(days=30)
b
```

```python
a=datetime.datetime(2021, 12, 3, 5, 34, 56)
b=a + datetime.timedelta(hours=5)
b
```

```python
c=datetime.datetime(2020, 11, 28, 15, 34, 56)
d=c+datetime.timedelta(days=10)
d
```

```python
f=b-a
f
```

```python
g=c-d
g
```

**Time library**-difference from time module in datetime library

```python
import time
```

```python
datetime.time(15,30,20) #this time is relative wrt to 12 AM in the morning
```

```python
a=time.time()  # this represents the number of seconds passed b/w 1 jan 1970 till now
a
```

```python
b=[1,3,5,7,9,11,13,15]
c=[]
first=time.time()
for i in b:
    c.append(i**2)
    second=time.time()
print(c)
print(second-first)
    
```

```python
b=[1,3,5,7,9,11,13,15]
first=time.time()
c=list(map(lambda b:b**2, b))
print(c)
second=time.time()
print(second-first)
```
