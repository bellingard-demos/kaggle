**Assignment**

```python

def countupper(name):
    count=0
    for i in name:
        if i.isupper():
            count=count+1
    return count
        
```

```python
count=countupper("AKash")
print(count)

    
```

**Write a calculator function**

```python
def calc(a,b,c):
    '''
    This is the documentation for the function
    a is the first number
    b is the second number
    c is the operation   
    '''
    #Docstring is used to document the new defined function for better understanding of users
    
    if c=="+":
        return a+b
    elif c=="-":
        return a-b
    elif c=="*":
        return a*b
    elif c=="/":
        return a/b
        
```

```python
num1=int(input("enter a value"))
num2=int(input("enter b value"))
operation=input("enter operation value")
result=calc(num1,num2,operation)
print(result)
```

**Flexible Argument Assignment**

```python
def month_salary(weekly_hours=8,weeks=4,pay_per_hour=500):
    salary=weekly_hours*weeks*pay_per_hour
    return salary
```

```python
salary=month_salary()
print(salary)
```

```python
salary=month_salary(weekly_hours=12,weeks=5,pay_per_hour=500)
print(salary)
```

```python
salary=month_salary(30,2)
print(salary)
```

**Variable number of arguments**

```python
def avg_marks(*marks):
    total,count=0,0
    for i in marks:
        total = total + i
        count = count + 1
    return total/count
```

```python
avg_marks(4,4)
```

```python
def sum_marks(*marks):
    total = 0
    for i in marks:
        total=total+i
    return total
```

```python
sum_marks(1,2,3,4,5,6)
```
