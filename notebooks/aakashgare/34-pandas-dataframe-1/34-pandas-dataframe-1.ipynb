# **Handling null values**


Dropping null records

```python
d={'students':['akash',None,'mahesh','harish'],
  'Marks':[12,None,34,None],
  'Age': [23,None,23,25]}
d
```

```python
import pandas as pd
```

```python
a=pd.DataFrame(d)
a
```

```python
 a.dropna()
```

```python
a.dropna(how='all')
```

**inplace for permananent change**

```python
a.dropna(how='all',inplace=True)
a
```

**Concept axis**

```python
a.dropna(axis=0) #axis =0 or index is same and axis =1 or column is same
```

```python
a.dropna(axis=1)
```

  **Subset**

```python
a.dropna(axis=0,subset=['Marks'])
```

```python
a.dropna(axis=1,subset=[3])
```

**Fillnull values**

```python
d={'students':['akash',None,'mahesh','harish'],
  'Marks':[12,None,34,None],
  'Age': [23,None,23,25]}
d
```

```python
c=pd.DataFrame(d)
c
```

```python
c.dropna(how='all',inplace=True)
c
```

```python
c['Age'].fillna(0)
```

```python
c['Age'].mean()
```

```python
c['Age'].median()
```

```python
c['Age'].fillna(c['Age'].median(),inplace=True)
c
```

```python
c['students'].fillna('vikas',inplace=True)
c
```

```python
c['Marks'].median()
```

```python
c.fillna(c['Marks'].median(),inplace=True)
c
```
