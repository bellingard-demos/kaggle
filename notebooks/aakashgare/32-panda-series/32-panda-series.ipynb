# **Pandas**

**Pandas** is a library/set of codes that we can import to analyze, clean, filter, manipulate, explore, model, etc., the dataset

There are broadly 2 data structures in pandas
* Series
* Dataframes

```python
import pandas as pd
```

**Series**
* Series is a 1 Dimensional array data structure
* Series kind of represents a single column in a table
* It can store data of any data type

**Dataframe**
* Dataframe is a 2 Dimensonal Datastructure
* Data is present in the form of rows and coloumns
* Can be considered as a combination of multiple series 

**Numpy vs Pandas**
* **Numpy** array contains data of same data type || **Pandas** Dataframe can have heterogenous data
* **Numpy** has a better performance for 50k rows or less || **Pandas** has a better performance for 500k rows or more
***Numpy** consumes less memory as compared to Pandas || **Pandas** consume large memory as compared to Numpy
* **Numpy** is prefered while using numbers || **Pandas** is prefered in tabular data

**Sort values with inplace parameter**
