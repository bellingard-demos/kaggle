**Write a python program to map two lists into a dictionary**

```python
name=["quresh","malesh","sudesh"]
age=[26,27,28]
```

```python
d={} #declare an empty cell
j=0
for i in name:
    d[i] = age[j]
    j=j+1
print(d)
```

**write a program to sort the dictionary by keys**

```python
d1={4:"a",
    6:"b",
    3:"c",
    1:"d",
    2:"e"}
s=sorted(d1,reverse=True)

d2={}
for i in s:
    d2[i]=d1[i]
print(d2)
```

Write a pythpon program to get the top three  city by temperaturte are sorted in a dictionary 

```python
temp = {"delhi":25,"banglore":26,"hyderabad":28,"pune":24}
value=sorted(temp.values(),reverse=True)
print(value)
for i in range(0,3):
    for j in temp:
        if value[i] == temp[j]:
            print(j)
    
```

check how many times a word is there in the senetence

```python
w="Fears leads to anger and anger leads to hate and hate leads to cruelty "
q=w.split()
f={}
for i in q:
    if f.get(i)==None:  # to create a fresh entry in our list
        f[i]=1
        print(f)
    else:
        f[i]=f[i]+1
print(f)


```
