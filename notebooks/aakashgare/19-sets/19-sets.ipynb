**Check whether a string is palangram or not**-Palangram is a sentence that contains every letter in english alphabet Eg.The quick brown fox jumps over the lazy dog

```python
a="The quick brown fox jumps over the lazy dog"
b=set(a)
count=0
for i in b:
    count=count+1
print(count)
b
    
```

```python
a=a.lower()
b=set(a)
count=0
for i in b:
    if i.isalpha()==True:
        count=count+1
        print(count)
if count==26:
    print("is palangram")
else:
    print("not palangram")
```
