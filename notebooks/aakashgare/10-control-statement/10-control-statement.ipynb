**Break statement**

```python
i=0
while True:
    print(i)
    i=i+1
    if i==10:
        break
```

****Continue statement:** ****It skips/rejects all the code in current iteration of the loop and moves control to the top of the loop**

```python
i=0
while i<=5:
    i=i+1
    if i==4:
        continue
    print(i)
    
```
