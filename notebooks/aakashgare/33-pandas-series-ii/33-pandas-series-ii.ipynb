**apply() method applies a function on value of series**

```python
a=pd.read_csv("/kaggle/input/videogamesales/vgsales.csv",usecols=['Global_Sales'],squeeze='True')
a
```

```python
def category(num):
    if num>35:
        return "Superhit"
    elif num>=10:
        return "Average"
    else:
        return "BelowAverage"
```

```python
a.apply(category)
```

```python
b=a.apply(category)
b
```

```python
b.value_counts()
```

```python
a.apply(lambda x:x*2)
```
