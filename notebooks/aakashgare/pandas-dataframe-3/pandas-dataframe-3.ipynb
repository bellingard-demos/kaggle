```python
def band(x):
    if x>2000000:
        return 'A'
    elif x>1000000:
        return "B"
    else:
        return "C"
```

**Apply transformation to the particular row**

```python
def flag(x):
    loc=x['Location']
    exp=x['Job Experience']
    if loc in ['Bangalore','Bengaluru','Bangalore/Bengaluru'] and exp=='3-5':
        return "Yes"
    else:
        return 'No'
```

```python
a['apply or not']=a.apply(flag, axis=1)
```

```python
a
```

```python
a['apply or not'].value_counts()
```
