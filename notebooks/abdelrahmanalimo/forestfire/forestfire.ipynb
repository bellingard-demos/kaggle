```python
df['month'] = pd.to_datetime(df["month"], format='%b').dt.month.astype(str)
df['month']
```

```python
df["day"].value_counts()
```

```python
day_to_num={"sun":1,"mon":2,"tue":3,"wed":4,"thu":5,"fri":6,"sat":7}
df["day"]=df["day"].map(day_to_num)
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
df.hist(figsize=(13,11))
plt.show()
```

```python
X=df.drop(columns="FFMC")
y=df["FFMC"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
RF=RandomForestRegressor()
```

```python
RF.fit(X_train,y_train)
```

```python
pred=RF.predict(X_test)
```

```python
print(round(RF.score(X_test,pred)))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style='whitegrid', font_scale=1.2)

# Training data
plt.figure(figsize=(8, 6))
sns.scatterplot(x=y_train, y=RF.predict(X_train), color='blue', label='Training Data')
sns.regplot(x=y_train, y=RF.predict(X_train), color='green', scatter=False)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Training Data vs. Predicted Values')
plt.legend()
plt.show()
```
