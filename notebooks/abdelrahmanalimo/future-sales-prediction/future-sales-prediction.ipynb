```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import plotly.express as px
import plotly.graph_objects as go
import numpy as np 
import pandas as pd
```

```python
data=pd.read_csv("/kaggle/input/sales-prediction/advertising.csv")
data.head()
```

```python
data.isnull().sum()
```

```python
fig=px.scatter(data_frame=data,x="Sales",y="TV",size="TV",trendline="ols")
fig.show()
```

```python
fig=px.scatter(data_frame=data,x="Sales",y="Radio",size="Radio",trendline="ols")
fig.show()
```

```python
corr=data.corr()
corr["Sales"].sort_values(ascending=False)
```

```python
X=data.drop(["Sales"],axis=1)
y=data["Sales"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model=LinearRegression()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
feature=[[220.2,28.5,195]]
print(model.predict(feature))
```
