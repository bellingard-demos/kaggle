```python
import pandas as pd
import pickle
from sklearn.ensemble import IsolationForest

# Load your dataset
# Assuming 'your_dataset.csv' is your dataset file
df = pd.read_excel('/kaggle/input/supervisor-sales/Supervisor Sales.xlsx')
df=df.drop(columns=["Week","Supervisor"],axis=1)
X = df.values

# Initialize the Isolation Forest model
model = IsolationForest(contamination=0.05)  # Adjust the contamination parameter as needed

# Fit the model to the data
model.fit(X)

# Predict outliers
outliers = model.predict(X)

# Add the 'outlier' column to the DataFrame
df['outlier'] = outliers

# Filter out the outliers
filtered_df = df[df['outlier'] == 1]

#save the model
file = open("expense_model.pkl", 'wb')
pickle.dump(model, file)
```
