```python
import pandas as pd 
import yfinance as yf
import datetime
from datetime import date,timedelta
```

```python
today=date.today()
d1=today.strftime("%Y-%m-%d")
end_date=d1
d2=date.today() - timedelta(days=4700)
d2=d2.strftime("%Y-%m-%d")
start_date=d2
data=yf.download("GOOG",start=start_date,end=end_date,progress=False)
data["Date"]=data.index
data=data[["Date","Open","High","Low","Close","Volume"]]
data.reset_index(drop=True,inplace=True)
```

```python
data.tail()
```

```python
import plotly.graph_objects as go
fig=go.Figure(data=[go.Candlestick(x=data["Date"]
                                  ,open=data["Open"],
                                  high=data["High"],
                                  low=data["Low"],
                                  close=data["Close"])])
fig.update_layout(title="google stock",xaxis_rangeslider_visible=False)
```

```python
data.corr()
```

```python
x=data[["Open","High","Low","Volume"]]
y=data[["Close"]]
x=x.to_numpy()
y=y.to_numpy()
y=y.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM
model=Sequential()
model.add(LSTM(128,return_sequences=True,input_shape=(X_train.shape[1],1)))
model.add(LSTM(64,return_sequences=False))
model.add(Dense(25))
model.add(Dense(1))
model.summary()
```

```python
model.compile(optimizer="adam",loss="mean_squared_error")
model.fit(X_train,y_train,batch_size=1,epochs=30)
```

```python
import numpy as np
feature=np.array([[140.750000,158.900000,138.705002,139.279999]])
model.predict(feature)
```
