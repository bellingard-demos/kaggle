```python
df_train=pd.read_csv("/kaggle/input/icr-identify-age-related-conditions/train.csv")
df_train.drop(columns=['Id'],inplace=True)
```

```python
df_train["EJ"]=df_train["EJ"].map({"A":0,"B":1}).astype("float32")
```

```python
from sklearn.impute import SimpleImputer
sm=SimpleImputer(missing_values=np.nan,strategy="mean")
df_train=pd.DataFrame(sm.fit_transform(df_train))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
X=df_train.iloc[:,:-1]
y=df_train.iloc[:,-1]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
lr=LogisticRegression(solver='liblinear',multi_class='ovr')
lr.fit(X_train,y_train)
lr.score(X_test,y_test)
```

```python
svm=SVC(gamma='auto')
svm.fit(X_train,y_train)
svm.score(X_test,y_test)
```

```python
rf=RandomForestClassifier(n_estimators=1000,random_state=1)
rf.fit(X_train,y_train)
rf.score(X_test,y_test)
```

```python
from sklearn.metrics import mean_absolute_error
pred=rf.predict(X_test)
me_ab=mean_absolute_error(pred,y_test,)
print("mea: ",me_ab)
```

```python
import pickle
pickle.dump(rf, open('model.pkl', 'wb'))
```

```python
pickled_model = pickle.load(open('model.pkl', 'rb'))
pickled_model.predict(X_test)
```
