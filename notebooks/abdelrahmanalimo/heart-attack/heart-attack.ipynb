```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
```

```python
data_heart=pd.read_csv("/kaggle/input/heart-attack-risk-analysis/train.csv")
data_heart
```

```python
data_heart.info()
```

```python
fig=px.histogram(data_frame=data_heart,x="Age",y="Heart Attack Risk",color="Sex")
fig.show()
```

```python
data_heart=data_heart.drop(["Country","Continent","Hemisphere"],axis=1)
data_heart["Sex"]=data_heart["Sex"].map({"Male":1,"Female":0})
data_heart["Diet"]=data_heart["Diet"].map({"Healthy":1,"Average":0.5,"Unhealthy":0}).astype('float')
data_heart[["org","press"]]=data_heart["Blood Pressure"].str.split('/', n=1, expand=True).astype('int')
data_heart=data_heart.drop(["Blood Pressure"],axis=1)
data_heart=data_heart.drop(["Patient ID"],axis=1)
```

```python
corr=data_heart.corr()
corr["Heart Attack Risk"].sort_values(ascending=False)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
X=np.array(data_heart[["Diabetes","Cholesterol","org","Exercise Hours Per Week","Triglycerides","Previous Heart Problems","Age","Income"]])
y=np.array(data_heart[["Heart Attack Risk"]])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
lr=LogisticRegression()
lr.fit(X_train,y_train)
```

```python
prediction=lr.predict(X_test)
prediction
```
