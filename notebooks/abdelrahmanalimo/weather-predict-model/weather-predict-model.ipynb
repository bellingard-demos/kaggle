```python
df=pd.read_csv("/kaggle/input/weather-dataset/weather.csv")
```

```python
df.isna().sum()
```

```python
median_sun=df["Sunshine"].median()
df["Sunshine"].fillna(value=median_sun, inplace=True)
```

```python
median_sun=df["WindGustSpeed"].median()
df["WindGustSpeed"].fillna(value=median_sun, inplace=True)
median_sun=df["WindSpeed9am"].median()
df["WindSpeed9am"].fillna(value=median_sun, inplace=True)
```

```python
df[["WindGustDir","WindDir9am","WindDir3pm"]]=df[["WindGustDir","WindDir9am","WindDir3pm"]].fillna("nan")
```

```python
from sklearn.preprocessing import LabelEncoder
df=df.apply(LabelEncoder().fit_transform)
```

```python
df
```

```python
X=df.iloc[:,:-3]
y=df.iloc[:,-3:]
y
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.multioutput import MultiOutputClassifier
clf = MultiOutputClassifier(LogisticRegression(multi_class='multinomial', solver='lbfgs'))
clf.fit(X,y)
```

```python
predict=clf.predict(X)
```

```python
clf.estimators_
```

```python
clf.score(X,y)
```

```python
prediction=pd.DataFrame(predict).to_csv("prediction.csv")
```
