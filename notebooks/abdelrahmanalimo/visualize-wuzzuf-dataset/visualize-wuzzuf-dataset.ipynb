```python
df=pd.read_csv('/kaggle/input/linkedin-csv/linkedin-jobs.csv')
```

```python
df.isna().sum()
```

```python
import re
chars_to_remove = ['b', "'"]
regular_expression = '[' + re.escape (''. join (chars_to_remove)) + ']'

df=df.replace(regular_expression, '', regex=True)
```

```python
df[["city","country"]]=df["Location"].str.split(n=1, expand=True)
```

```python
df["country"]=df["country"].str.split(',').str[1]
```
