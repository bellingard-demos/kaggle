## useing Y Data profiling lib

```python
df = pd.read_csv('/kaggle/input/predict-lifespan-of-a-comet-goldfish/fish_data.csv')
profile = ProfileReport(df, title="Profiling Report")
```

```python
profile
```

```python
val = df[['average_length(inches))', 'average_weight(inches))', 'ph_of_water','color','Gender', 'life_span']]
pd=pd.get_dummies(val)
```

```python
pd=pd.astype(int)
pd
```

```python
x = pd[['average_length(inches))', 'average_weight(inches))', 'ph_of_water','color_Black_and_Orange', 'color_Calico',
       'color_Green_and_Chocolate', 'color_Orange', 'color_Red_and_Silver',
       'color_Red_and_White_Bi_Color', 'color_Reddish_Orange', 'color_White',
       'color_Yellow', 'Gender_False', 'Gender_True']]
y = pd['life_span']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y ,test_size=0.2)
```

```python
xgb = xgboost.XGBRegressor(n_estimators=1000, learning_rate=0.08, gamma=0, subsample=0.75,colsample_bytree=1, max_depth=7)
```

```python
xgb.fit(X_train,y_train)
```

```python
predictions = xgb.predict(X_test)
```

```python
print(f''' The train accuracy : {r2_score(y_train,xgb.predict(X_train))}''')
```
