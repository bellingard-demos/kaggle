```python
import pandas as pd
import matplotlib.pyplot as plt
import pandas_profiling 
from wordcloud import WordCloud
import plotly.express as px
```

# **Read Data** 

```python
df=pd.read_csv("../input/omicron-covid19-variant-daily-cases/covid-variants.csv") 
df.head()
```

```python
df.info()
```

```python
for i in df.columns:
    print(f"{df[i].unique()} \n {'-'*70}")
```

# **Clening Data**

```python
df.isnull().sum()
```

```python
df.variant.unique()
```

```python
#'non_who' it's not necessary 
df=df[df.variant!="non_who"]
df.variant.unique()
```

## Data Analysis With pandas profiling

```python
prof = pandas_profiling.ProfileReport(df)
prof.to_file(output_file='output.html')
prof
```

# Visualization

```python
# word cloud for variant
plt.subplots(figsize=(10,15))
wordcloud=WordCloud(background_color='white',width=512,height=384).generate(" ".join(df.variant))
plt.imshow(wordcloud)
plt.axis('off')
plt.savefig('graph.png')
plt.show()
```

```python
sp  = df.rename(columns={"location":"Location","num_sequences_total":"Number of Case"})

fig = px.treemap(sp, path=[px.Constant('Number of Case'),'Location'], values='Number of Case',
                   hover_data=['Location'])
fig.update_layout(title_text='Total COVID-19 Case')

fig.show()
```
