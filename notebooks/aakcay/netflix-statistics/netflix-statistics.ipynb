# **Data Visualization**

```python
#Correlation Map
f,ax=plt.subplots(figsize=(15,15))
sns.heatmap(data.corr(),annot=True,linewidths=.5,fmt='.1f',ax=ax)
plt.show()
```

```python
#Histogram Plot
def hist(a):
    plt.hist(data[a],bins=10)
    plt.xlabel(a)
    plt.ylabel("Frequency")
    plt.show()

numericvalue=["Total Library Size","No. of Movies","No. of TV Shows",]    
for i in numericvalue:
    hist(i)
    
```

```python
plt.figure(figsize=(15,10))
plt.bar(data["Country"],data["Total Library Size"])
plt.xlabel("Country")
plt.ylabel("Total Library Size")

plt.xticks(rotation=90)
plt.show()

```

```python
plt.figure(figsize=(15,10))
plt.bar(data["Country"],data["No. of TV Shows"])
plt.xlabel("Country")
plt.ylabel("No. of TV Shows")

plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.bar(data["Country"],data["No. of Movies"])
plt.xlabel("Country")
plt.ylabel("No. of Movies")

plt.xticks(rotation=90)
plt.show()
```

```python
#Line plot

plt.figure(figsize=(15,10))
plt.plot(data["Country"],data["Cost Per Month - Basic ($)"],color="red")
plt.plot(data["Country"],data["Cost Per Month - Standard ($)"],color="blue")
plt.plot(data["Country"],data["Cost Per Month - Premium ($)"],color="green")
plt.xticks(rotation=90)
plt.show()
```

```python
#Scatter Plot
plt.figure(figsize=(15,10))
plt.scatter('Country',"Cost Per Month - Basic ($)" , data=data,color="red")
plt.scatter('Country',"Cost Per Month - Standard ($)" , data=data,color="blue")
plt.scatter('Country',"Cost Per Month - Premium ($)" , data=data,color="green")
plt.xticks(rotation=90)

plt.show()
```

```python
#Line plot

plt.figure(figsize=(15,10))
plt.plot(data["Country"],data["No. of TV Shows"],color="red")
plt.plot(data["Country"],data["No. of Movies"],color="blue")

plt.xticks(rotation=90)
plt.show()
```
