# Variable Description
<a id="2" ></a><br>

1. school: student's  school (Gabriel Pereira or Mousinho da Silveira )
1. sex: student's gender (F or M)
1. age: student's age
1. address: student's address (it place in portugal  urban or rural)
1. famsize: student's family size (from 15 to 22)
1. Pstatus: family status (Togather or Apart)
1. Medu: mother's education (0:none,1,2,3,4) 
1. Fedu: father's education (0:none,1,2,3,4) 
1. Mjob: mother's job
1. Fjob: father's job
1. reason: the reason student's choose scholl
1. guardian: student's guardian(mother,father or other)
1. traveltime: travel time from home to school
1. studytime: weekly study time
1. failures: number of failures in school
1. schoolsup: support from school to student 
1. famsup: support from family to student
1. paid: are there paid lesson(yes or no) 
1. activities: extracurricular activity(yes or no)
1. nursery: attended nursery school(yes or no)
1. higher: wants to take higher education(yes or no) 
1. internet: is there internet in home(yes or no) 
1. romantic: romatic relationship(yes or no) 
1. famrel: quality of family relationships
1. freetime: free time after school
1. goout: spending time with friends after school
1. Dalc: daily alcohol consumption
1. Walc: weekend alcohol consumption 
1. health: student health status
1. absences:  number of student absences
1. G1: first period grade
1. G2: second period grade
1. G3: final grade

* int64(16): age, fedu, medu, traveltime, studytime, failures, famrel, freetime, goout, Dalc, Walc ,health, absences, G1, G2, G3       
* object(17):school, sex, address, famsize, Pstatus, Mjob, Fjob, reason, guardian,schoolsup, famsup,  paid, activities, nursery, higher, internet, romantic

# Categorical Description 
<a id="4" ></a><br>

```python
a=["school", "sex", "address", "famsize", "Pstatus", "Mjob", "Fjob", "reason", "guardian","schoolsup", "famsup",  "paid", "activities", "nursery", "higher", "internet", "romantic"]
def count_plot(a):
        plt.figure(figsize=(5,5))
        sns.countplot(x=a,data=data)
       
        plt.show()
        print(a,"\n",data[a].value_counts())
```

```python
for i in a:
        count_plot(i)
                    
```

```python
#age & Dalc ratio
age_list=data.age.unique()
Dalc_ratio=[]
for i in age_list:
    x=data[data.age==i]
    Dalc_rate=sum(x.Dalc)/len(x)
    Dalc_ratio.append(Dalc_rate)
d2=pd.DataFrame({'age':age_list,'Dalc_ratio':Dalc_ratio})  

sns.barplot(x=d2.age,y=d2.Dalc_ratio)
plt.grid()
```

```python
#age & Walc ratio
age_list=data.age.unique()
Walc_ratio=[]
for i in age_list:
    x=data[data.age==i]
    Walc_rate=sum(x.Walc)/len(x)
    Walc_ratio.append(Walc_rate)
d3=pd.DataFrame({'age':age_list,'Walc_ratio':Walc_ratio})  

sns.barplot(x=d3.age,y=d3.Walc_ratio)
plt.grid()
```

```python
#age & goout ratio
age_list=data.age.unique()
goout_ratio=[]
for i in age_list:
    x=data[data.age==i]
    goout_rate=sum(x.goout)/len(x)
    goout_ratio.append(goout_rate)
d4=pd.DataFrame({'age':age_list,'goout_ratio':goout_ratio})  

sns.barplot(x=d4.age,y=d4.goout_ratio)
plt.grid()
```

```python
#age & famrel ratio
age_list=data.age.unique()
famrel_ratio=[]
for i in age_list:
    x=data[data.age==i]
    famrel_rate=sum(x.famrel)/len(x)
    famrel_ratio.append(famrel_rate)
d5=pd.DataFrame({'age':age_list,'famrel_ratio':famrel_ratio})  

sns.barplot(x=d5.age,y=d5.famrel_ratio)
plt.grid()
```

```python

sns.pointplot(x='age',y='Dalc_ratio',data=d2,color="green")
sns.pointplot(x='age',y='Walc_ratio',data=d3,color="blue")
sns.pointplot(x='age',y='goout_ratio',data=d4,color="red")
sns.pointplot(x='age',y='famrel_ratio',data=d5,color="yellow")
plt.text(8,1.2,'famrel ratio',color='yellow',fontsize = 17,style = 'italic')
plt.text(8,0.3,'Dalc ratio',color='green',fontsize = 17,style = 'italic')
plt.text(8,0.6,'Walc ratio',color='blue',fontsize = 18,style = 'italic')
plt.text(8,0.9,'goout ratio',color='red',fontsize = 18,style = 'italic')
plt.ylabel('Values',fontsize = 15,color='blue')
plt.xlabel('Age',fontsize = 15,color='blue')
plt.grid()
```
