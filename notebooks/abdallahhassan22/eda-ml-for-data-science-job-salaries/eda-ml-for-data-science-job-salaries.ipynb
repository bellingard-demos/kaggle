# Build ML Model To Predict Salaries in USD
* select most frequented jobs to predict their salaries
* encode the data 
* scale the data
* split our data to train and test 
* fit the model
* evalute our results
* note we don't have enough data or good features so the results not the best

```python
new_df=df[df['job_title'].isin(['Data Scientist','Data Analyst','Data Engineer','Machine Learning Engineer'])]
new_df=new_df[['work_year','remote_ratio','experience_level','employment_type','job_title','employee_residence','company_location','company_size','salary_in_usd']].copy()
new_df
```

```python
new_df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
new_df.iloc[:,2:-1]=new_df.iloc[:,2:-1].apply(LabelEncoder().fit_transform)
new_df
```

```python
y=new_df.iloc[:,-1]
from sklearn.preprocessing import StandardScaler
X=StandardScaler().fit_transform(new_df.iloc[:,:-1])
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
import xgboost as xg
from sklearn.metrics import mean_absolute_error
xgb_r = xg.XGBRegressor(objective ='reg:squarederror', n_estimators = 10)
xgb_r.fit(X,y) ## fitting the data

xgb_r.fit(X_train, y_train)
pred=xgb_r.predict(X_test)
# Predicting the Test set results
print('Mean Absolute Error = '+str(mean_absolute_error(y_test,pred))+' $')
xgb_r.score(X_test,y_test)
```
