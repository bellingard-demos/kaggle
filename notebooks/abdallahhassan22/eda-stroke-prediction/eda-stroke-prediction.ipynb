# From here we will start to build our ml model
* select our columns 
* here I selected the categorical columns on the left side for easy encoding
* encode categorical columns

```python

df=df[['gender','ever_married','work_type','Residence_type','age','hypertension','heart_disease','avg_glucose_level','bmi','stroke']]
df
```

```python
from sklearn.preprocessing import LabelEncoder
df.iloc[:,0:4]=df.iloc[:,0:4].apply(LabelEncoder().fit_transform)
df
```

# Undersampling technique
* <img src='https://datascience.aero/wp-content/themes/yootheme/cache/handleimbalanceddata-53752054.jpeg'>
* as i mentioned before there is imbalance in our data so I will use this technique
* you can choose any technique you want 
* I will equalize the two classes
* the first class has 288 sample 
* so the second class will be 288 to be balanced
* connect the tow classes in one dataframe

```python
from sklearn.utils import resample
patient=df[df['stroke']==1]
not_patient = resample(df[df['stroke']==0],
             replace=True,
             n_samples=288,
             random_state=42)
df=pd.concat([patient,not_patient])
df
```

# Divide the data to features and label
* X is our features
* y is our label

```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

# Scale our data

```python
from sklearn.preprocessing import StandardScaler
X=StandardScaler().fit_transform(X)
```

# Split our data to train and test sets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 17)
```

# Build the model
* <img src='https://gaussian37.github.io/assets/img/ml/concept/random-forest/randomforest.PNG'>
* I selected the random forest classifier with 10 trees
* fit the data into the model 
* predict the test data 
* compare with y_test to show our metrics 
* print the confusion matrix
* print the classifaction report 
* the results looks good 
* here there is no matter to evaluate the model by accuracy metric because the data is balanced

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
rf=RandomForestClassifier(n_estimators=10,random_state=17)
rf.fit(X_train,y_train)
pred=rf.predict(X_test)
cm=confusion_matrix(pred,y_test)
print(cm)
print(accuracy_score(pred,y_test))
print(classification_report(pred,y_test))
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report

from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
rf=KNeighborsClassifier(n_neighbors=13)
rf.fit(X_train,y_train)
pred=rf.predict(X_test)
cm=confusion_matrix(pred,y_test)
print(cm)
print(accuracy_score(pred,y_test))
print(classification_report(pred,y_test))
KNeighborsClassifier
```
