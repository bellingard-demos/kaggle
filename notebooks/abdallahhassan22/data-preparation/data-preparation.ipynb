# load 1 million record from the data 
# you can change this number based on your needs

```python
f=open('/kaggle/input/otto-recommender-system/train.jsonl', 'r').readlines()[:1000]
```

# make a loop on the data to arrnge it 
# add every single object in a list 
# all you will need in a recommendation system model these features and you can change to what you need

```python
from tqdm import tqdm
records=[]
for line in tqdm(f):
        # Parse the JSON object
    obj = json.loads(line)
    session=obj['session'] 
    for event in obj['events']:
        # Get the aid for the event
        aid = event['aid']
        type=event['type']
        record=[session,aid,type]
        records.append(record)
```

# convert the list into dataframe

```python
df=pd.DataFrame(records,columns=['session_id','aid','type'])
```

```python

from sklearn import preprocessing
  
label_encoder = preprocessing.LabelEncoder()
  
df['type']= label_encoder.fit_transform(df['type'])
  
df['type'].unique()
```

```python
matrix = df.pivot_table(values='type', index='session_id', columns='aid', fill_value=4)
```

```python
matrix.head()
```

```python
matrix.shape
```

```python
X=matrix.T
```

```python
from sklearn.decomposition import TruncatedSVD
SVD = TruncatedSVD(n_components=10)
decomposed_matrix = SVD.fit_transform(X)
decomposed_matrix.shape
```

```python
correlation_matrix = np.corrcoef(decomposed_matrix)
correlation_matrix.shape
```

```python
X.index[100]
```

```python
i = 4284

aids = list(X.index)
product_ID = aids.index(i)
product_ID
```

```python
correlation_product_ID = correlation_matrix[product_ID]
correlation_product_ID.shape
```

```python
Recommend = list(X.index[correlation_product_ID >= 0.99])

# Removes the item already bought by the customer
Recommend.remove(i) 

Recommend[0:9]
```
