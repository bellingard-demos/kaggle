```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('/kaggle/input/home-data-for-ml-course/train.csv')
```

```python
# features = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']

features = [
    'MSSubClass',
    'LotArea',
    'OverallQual',
    'OverallCond',
    'YearBuilt',
    'YearRemodAdd',
    'BsmtFinSF1',
    'BsmtFinSF2',
    'BsmtUnfSF',
    'TotalBsmtSF',
    '1stFlrSF',
    '2ndFlrSF',
    'LowQualFinSF',
    'GrLivArea',
    'BsmtFullBath',
    'BsmtHalfBath',
    'FullBath',
    'HalfBath',
    'BedroomAbvGr',
    'KitchenAbvGr',
    'TotRmsAbvGrd',
    'Fireplaces',
    'GarageCars',
    'GarageArea',
    'WoodDeckSF',
    'OpenPorchSF',
    'EnclosedPorch',
    '3SsnPorch',
    'ScreenPorch',
    'PoolArea',
    'MiscVal',
    'MoSold',
    'YrSold']

```

```python
X = data[features]
y = data.SalePrice
x_train, x_val, y_train, y_val = train_test_split(X,y,random_state=1)
```

```python
rf_model = RandomForestRegressor(random_state=1, n_estimators=700)
rf_model.fit(x_train,y_train)
rf_val_preds = rf_model.predict(x_val)

gbm_model = GradientBoostingRegressor(random_state=1, n_estimators=500)
gbm_model.fit(x_train,y_train)
gbm_val_preds = rf_model.predict(x_val)


preds = (rf_val_preds + gbm_val_preds)/ 2
```

```python
np.sqrt(mean_squared_error(preds,y_val))
```

```python
rf_model_fulldata = RandomForestRegressor(random_state=1, n_estimators=700)
rf_model_fulldata.fit(X,y)
```

```python
gbm_model_fulldata = GradientBoostingRegressor(random_state=1, n_estimators=500)
gbm_model_fulldata.fit(X,y)
```

```python
test_data = pd.read_csv("/kaggle/input/home-data-for-ml-course/test.csv")
test_data = test_data.fillna(-1)


x_test = test_data[features]
test_preds1 = rf_model_fulldata.predict(x_test)
test_preds2 = gbm_model_fulldata.predict(x_test)

test_preds = (test_preds1 + test_preds2)/2

output = pd.DataFrame({'Id':test_data.Id, 'SalePrice':test_preds})
```

```python
output.to_csv("my_submissions.csv",index=False)
```
