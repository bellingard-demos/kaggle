## Bivariate Analysis

```python
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
flights = sns.load_dataset('flights')
titanic = pd.read_csv("/kaggle/input/titanic/train.csv")
```

## 1. Scatterplot - Numerical Numerical

```python
#Bivariate
sns.scatterplot(x=tips["total_bill"], y=tips["tip"])
```

Info - Linear Relation between total bill and tip

```python
#Multivariate
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex", style="smoker",size="size")
```

## 2. Bar Plot - Num Catergorical

```python
#Bivariate
sns.barplot(data=titanic, x="Pclass", y="Age")
```

Info - Younger people were in lower class, while Older people in Upper class

```python
#Multivariate
sns.barplot(data=titanic, x="Pclass", y="Age", hue="Sex")
```

## 3. Box Plot - Num Cat

```python
#Bivariate
sns.boxplot(data=titanic,x="Sex", y="Age")
```

Info - Age group was mostly between 20 to 40, very low number of older people, more number of younger people

```python
#Multivariate
sns.boxplot(data=titanic, x="Sex",y="Age",hue="Survived")
```

## 4. Distplot - Num Cat

```python
#Biviarte
sns.kdeplot(titanic[titanic['Survived'] == 0]['Age'])
sns.kdeplot(titanic[titanic['Survived'] == 1]['Age'])
```

Info - Younger People were priotized, age of 30 weren't

## 5. HeatMap - cat cat

```python
data = pd.crosstab(titanic["Pclass"],titanic["Survived"])
data
```

```python
sns.heatmap(data)
```

Info - For dead people most are 3rd class, least in 1st class <br>
For Alive, most in 1st class

## 6. ClusterMap (Categorical - Categorical)

```python
#Bivariate
data = pd.crosstab(titanic["Pclass"],titanic["Survived"])
sns.clustermap(data)
```

```python
#Multivariate
sns.clustermap(flights.pivot_table(values='passengers',index='month',columns='year'))
```

## 7. Pairplot

```python
#Bivariate
sns.pairplot(iris)
```

```python
#Multivariate
sns.pairplot(iris,hue='species')
```

## 8. Lineplot (Numerical - Numerical)

```python
df = flights.groupby('year').sum().reset_index()
df.head()
```

```python
sns.lineplot(data=df,x="year",y="passengers")
```

Info - growth in company year by year

```python
sns.lineplot(data=flights,x="year",y="passengers",hue="month")
```

Info - In Summers people travel more
