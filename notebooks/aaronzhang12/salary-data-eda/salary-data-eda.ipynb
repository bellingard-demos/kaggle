```python
def category(data):
    blanks=[]
    for x in data['Job Title']:
        if x in Software_Engineering:
            blanks.append('Software_Engineering')
        elif x in Data_Science:
            blanks.append('Data_Science')
        elif x in Management:
            blanks.append('Management')
        elif x in Marketing:
            blanks.append('Marketing')
        elif x in Product_Management:
            blanks.append('Product_Management')
        elif x in Human_Resources:
            blanks.append('Human_Resources')
        elif x in Sales:
            blanks.append('Sales')
        elif x in Finance:
            blanks.append('Finance')
        elif x in IT:
            blanks.append('IT')
        elif x in Customer_Service:
            blanks.append('Customer_Service')
        elif x in Research_Science:
            blanks.append('Research_Science')
        elif x in Web_Development:
            blanks.append('Web_Development')
        elif x in Design:
            blanks.append('Design')
        elif x in Administrative:
            blanks.append('Administrative')
        elif x in HR:
            blanks.append('HR')
        elif x in Writing:
            blanks.append('Writing')
        else:
            blanks.append('Event_Coordination')
            
    return blanks
```

```python
data['Job Category'] = category(data)
```

```python
def race(data):
    blanks = []
    for x in data['Race']:
        if x == 'Korean' or x == 'Chinese':
            blanks.append('Asian')
        elif x == 'Welsh' or x == 'Australian':
            blanks.append('White')
        elif x == 'African American':
            blanks.append('Black')
        else:
            blanks.append(x)
    return blanks 
        
```

```python
data['Ethnicity'] = race(data)
```

```python
#Function return a dataframe of avg salary of each job category in each country

def country_job_salary(country):
    if country in data.Country.unique():
        df = data.loc[data['Country']==country].groupby('Job Category')['Salary'].mean()\
.sort_values(ascending=False).reset_index().set_index('Job Category').rename(columns={'Salary':country})
        return df
    else:
        print('The country is not available.')
```

```python
country_job_salary('China')
```

```python
for x in data.Country.unique():
    df = country_job_salary(x)
    plt.figure(figsize=(16, 6))
    sns.barplot(data=df.reset_index(), x='Job Category', y=x)
    plt.xticks(rotation=30)
    plt.title(f'Avg Salary of each Job Category in {x}');
```

```python
#Return average value of salary of each job category in each country

avg_salary_country = pd.DataFrame()

for x in data.Country.unique():
    df = country_job_salary(x)
    avg_salary_country = pd.concat([
        avg_salary_country, df
    ], axis=1)
    
avg_salary_country
```

```python
avg_salary_country.plot(kind='bar', figsize=(18, 8), rot=35)
plt.title('Avge Salary of each Job Category in each Country');
```

```python
#Avg Salary of Different Education Level in different Country

def education_salary(country):
    if country in data['Country'].unique():
        df = data.loc[data['Country'] == country].groupby('Education Level')['Salary'].mean()\
             .sort_values(ascending=False).reset_index()\
             .rename(columns={'Salary':country}).set_index('Education Level')
        return df 
    
    else:
        print('The country is not available from the dataset.')
```
