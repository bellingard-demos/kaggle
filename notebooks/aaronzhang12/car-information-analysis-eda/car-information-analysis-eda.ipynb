```python
#Sklearn
```

```python
df = pd.concat([
    data, pd.get_dummies(data['origin'])
],axis=1)

df.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
import xgboost
```

```python
from sklearn.model_selection import train_test_split

X = df.drop(['name', 'mpg', 'origin'], axis=1)
y = data.loc[:, 'mpg']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
tree = DecisionTreeRegressor()
tree.fit(X_train, y_train)
```

```python
y_hat_train = tree.predict(X_train)
```

```python
mean_squared_error(y_hat_train, y_train)
```

```python
forest = RandomForestRegressor()
forest.fit(X_train, y_train)
```

```python
y_hat_train = forest.predict(X_train)
```

```python
mean_squared_error(y_hat_train, y_train)
```

```python
#Cross Validated Score
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(forest, X_train, y_train,
                        scoring='neg_mean_squared_error', cv=10)

tree_rmse_in_sample_estimates = np.sqrt(-scores)

tree_rmse_in_sample_estimates
```

```python
def display_scores(scores):
    print('===============================================')
    print('Scores: {}'.format(scores))
    print('===============================================')
    print('Mean Score: {}'.format(scores.mean()))
    print('===============================================')
    print('Standard Deviation of Scores: {}'.format(scores.std()))
    print('===============================================')
    
    return None
```

```python
display_scores(tree_rmse_in_sample_estimates)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
forest.get_params()
```

```python
# # 12 hyperparameters
grid_1 = {'n_estimators': [3,10,30],
         'max_features': [2,4,6,8]}

# 6 hyperparameters
grid_2 = {'bootstrap':[False],
         'n_estimators':[3,10],
         'max_features':[2,3,4]}

param_grid = [grid_1, grid_2]
```

```python
grid_search = GridSearchCV(forest,
                          param_grid,
                          cv=5,
                          scoring='neg_mean_squared_error',
                          return_train_score=True,
                          refit=True)

grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_estimator_
```

```python
feature_importances = grid_search.best_estimator_.feature_importances_
feature_importances
```

```python
plt.barh(X_train.columns, feature_importances)
```

```python
sorted(zip(X_train.columns, feature_importances))
```

```python
final_model = grid_search.best_estimator_
```

```python
final_model.fit(X_test, y_test)
```

```python
y_hat_test = final_model.predict(X_test)
```

```python
mean_squared_error(y_hat_test, y_test)
```

```python
def name_to_brand(data, feature):
    blanks = []
    for x in data[feature]:
        blanks.append(x.split()[0])   
    return blanks
```

```python
data['brand'] = name_to_brand(data, 'name')
```

```python
for i in range(len(data.brand)):
    if data.brand[i] == 'chevrolet' or data.brand[i] == 'chevroelt':
        data.brand[i] = 'chevy'
    elif data.brand[i] == 'volkswagen' or data.brand[i] == 'vw':
        data.brand[i] = 'volkswagen'
    elif data.brand[i] == 'vokswagen' or data.brand[i] == 'vw':
        data.brand[i] = 'volkswagen'
    elif data.brand[i] == 'maxda':
        data.brand[i] = 'mazda'
    elif data.brand[i] == 'toyouta':
        data.brand[i] = 'toyota'
    elif data.brand[i] == 'mercedes':
        data.brand[i] = 'mercedes-benz'
```
