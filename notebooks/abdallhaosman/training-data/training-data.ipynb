```python
import re

def extract_battery_capacity(battery_str):
    match = re.search(r'\d+', battery_str)
    if match:
        return int(match.group())
    return None

df['Battery_mAh'] = df['battery'].apply(extract_battery_capacity)

print(df[['battery', 'Battery_mAh']].head())
```

```python
#del.df["battery"]
```

```python
def extract_storage_ram(Ram_str):
    match = re.search(r'\d+', Ram_str)
    if match:
        return int(match.group())
    return None

df['Ram'] = df['storage_ram'].apply(extract_battery_capacity)

print(df[['storage_ram', 'Ram']].head())
```

```python
#del.df["storage_ram"]
```

```python
ax = sns.scatterplot(x='price',
                y='Battery_mAh',
                hue='ratings',
                data=df)
ax.set_title('price  vs. Battery_mAh')
plt.show()
```
