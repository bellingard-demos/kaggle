<h4> youtube channel Data Dataset contains <mark>27 columns</mark>, each are:</h4>


* <b> <mark>rank:</mark></b> Position of the YouTube channel based on the number of subscribers
* <b> <mark>Youtuber:</mark></b> Name of the YouTube channel
* <b> <mark>subscribers:</mark></b> Number of subscribers to the channel
* <b> <mark>video views:</mark></b> Total views across all videos on the channel
* <b> <mark>category:</mark></b> Category or niche of the channel
* <b> <mark>Title:</mark></b> Title of the YouTube channel
* <b> <mark>uploads:</mark></b> Total number of videos uploaded on the channel
* <b> <mark>Country:</mark></b> Country where the YouTube channel originates
* <b> <mark>Abbreviation:</mark></b> Abbreviation of the country
* <b> <mark>channel_type:</mark></b> Type of the YouTube channel (e.g., individual, brand)
* <b> <mark>video_views_rank:</mark></b> Ranking of the channel based on total video views
* <b> <mark>country_rank:</mark></b> Ranking of the channel based on the number of subscribers within its country
* <b> <mark>channel_type_rank:</mark></b> Ranking of the channel based on its type (individual or brand)
* <b> <mark>video_views_for_the_last_30_days:</mark></b> Total video views in the last 30 days
* <b> <mark>lowest_monthly_earnings:</mark></b> Lowest estimated monthly earnings from the channel
* <b> <mark>highest_monthly_earnings:</mark></b> Highest estimated monthly earnings from the channel
* <b> <mark>lowest_yearly_earnings:</mark></b> Lowest estimated yearly earnings from the channel
* <b> <mark>highest_yearly_earnings:</mark></b> Highest estimated yearly earnings from the channel
* <b> <mark>subscribers_for_last_30_days:</mark></b> Number of new subscribers gained in the last 30 days
* <b> <mark>created_year:</mark></b> Year when the YouTube channel was created
* <b> <mark>created_month:</mark></b> Month when the YouTube channel was created
* <b> <mark>created_date:</mark></b> Exact date of the YouTube channel's creation
* <b> <mark>Gross tertiary education enrollment (%):</mark></b> Percentage of the population enrolled in tertiary education in the country
* <b> <mark>Population:</mark></b> Total population of the country
* <b> <mark>Unemployment rate:</mark></b> Unemployment rate in the country
* <b> <mark>Urban_population:</mark></b> Percentage of the population living in urban areas
* <b> <mark>Latitude:</mark></b> Latitude coordinate of the country's location
* <b> <mark>Longitude:</mark></b> Longitude coordinate of the country's location

```python
# Fill missing values in categorical columns with 'Unknown'
categorical_columns = ['category', 'Country', 'channel_type']
df[categorical_columns] = df[categorical_columns].fillna('Unknown')

# Calculate mean and median for specific columns
mean_video_views_rank = df['video_views_rank'].mean()
median_Population = df['Population'].median()
median_Unemployment_rate = df['Unemployment rate'].median()
median_Urban_population = df['Urban_population'].median()
median_Latitude = df['Latitude'].median()
median_Longitude = df['Longitude'].median()

# Fill missing values in specific columns with calculated values
df['video_views_rank'].fillna(mean_video_views_rank, inplace=True)
df['Population'].fillna(median_Population, inplace=True)
df['Unemployment rate'].fillna(median_Unemployment_rate, inplace=True)
df['Urban_population'].fillna(median_Urban_population, inplace=True)
df['Latitude'].fillna(median_Latitude, inplace=True)
df['Longitude'].fillna(median_Longitude, inplace=True)

# Drop specific columns
df.drop(columns=['subscribers_for_last_30_days','Title','category'], inplace=True)

# Drop rows with any remaining missing values
df = df.dropna()

# Display the shape of the DataFrame after changes
print(df.shape)
```

```python
def Freq_df(cleanwordlist):
    Freq_dist_nltk = nltk.FreqDist(cleanwordlist)
    df_freq = pd.DataFrame.from_dict(Freq_dist_nltk, orient='index')
    df_freq.columns = ['Frequency']
    df_freq.index.name = 'Term'
    df_freq = df_freq.sort_values(by=['Frequency'],ascending=False)
    df_freq = df_freq.reset_index()
    return df_freq

def Word_Cloud(data, color_background, colormap, title):
    plt.figure(figsize = (20,15))
    wc = WordCloud(width=1200, 
               height=600, 
               max_words=50,
               colormap= colormap,
               max_font_size = 100,
               random_state=88, 
               background_color=color_background).generate_from_frequencies(data)
    plt.imshow(wc, interpolation='bilinear')
    plt.title(title, fontsize=20)
    plt.axis('off')
    plt.show()
```

```python
freq_df = Freq_df(df['Youtuber'].values.tolist())
data = dict(zip(freq_df['Term'].tolist(), freq_df['Frequency'].tolist()))
data = freq_df.set_index('Term').to_dict()['Frequency']

Word_Cloud(data ,'black','RdBu', 'WordCloud of Youtuber')
```

```python
models = {
    'Random Forest': RandomForestRegressor(random_state=42),
    'Gradient Boosting': GradientBoostingRegressor(random_state=42),
}
best_model = None
best_r2 = 0

for model_name, model in models.items():
    model.fit(X_train, y_train)
    y_pred= model.predict(X_test)

    # Evaluate the model
    r2 = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    submit = pd.DataFrame()
    submit['Actual_video_views'] = y_test
    submit['Predict_video_views'] = y_pred
    submit = submit.reset_index()
    r2 = r2_score(y_test, y_pred)
    if r2 > best_r2:
        best_r2 = r2
        best_model = model.__class__.__name__

    print(f'{model_name}:')
    print(f'R2 Score: {r2:.2f}')
    print(f'Mean Absolute Error (MAE): {mae:.2f}')
    print(f'Root Mean Squared Error (RMSE): {rmse:.2f}')
    print(submit.head(5))

    print('----------------------------------------')
print(f"The best performing model is: {best_model} with accuracy: {best_r2:.2f}")
```

```python
importances = model.feature_importances_

feature_names = X.columns

feature_importance_dict = dict(zip(feature_names, importances))

sorted_feature_importance = sorted(feature_importance_dict.items(), key=lambda x: x[1], reverse=True)

for feature, importance in sorted_feature_importance:
    print(f"{feature}: {importance:.2f}")

plt.figure(figsize=(12, 7))
plt.barh(*zip(*sorted_feature_importance), alpha=0.9, color='#6495ED')
plt.title('Feature Importance', fontsize=15)
plt.xlabel("Importance")
plt.ylabel("Feature")
plt.show()
```
