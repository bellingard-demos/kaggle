## Data clean


**Price**

```python
df['Price'].describe()
```

```python
#distribution of price before filling missing values
sns.displot(data, kind="kde", x="Price")
```

```python
df['Price'].fillna(df['Price'].mean(), inplace=True)
df.isnull().sum()
```

```python
#distribution of Price after filling missing values
sns.displot(df, kind="kde", x="Price")
```

```python
df = data[data['Price'] > 1000000]
plt.scatter(df['Price'] , df['BuildingArea'])
plt.xlabel("Price")
plt.ylabel("BuildingArea")
```

**distance**

```python
df['Distance'].fillna(df['Distance'].median(), inplace=True)
```

```python
df['Distance'].isnull().sum()
```

**postcode**

```python
df['Postcode'].fillna(df['Postcode'].mean(), inplace=True)
```

```python
df['Postcode'].isnull().sum()
```

**Council area**

```python
df.dropna(subset=['CouncilArea'], inplace=True)
```

```python
df['CouncilArea'].isnull().sum()
```

```python
#YearBuilt has no significant meaning in our data, so it is dropped
df = df.drop('YearBuilt', axis=1)
df.info()
```

```python
sns.kdeplot(data, x='Lattitude', fill=True)
```

```python
df['Lattitude'].interpolate(method='linear', inplace=True)
df['Lattitude'].isnull().sum()
```

```python
sns.kdeplot(df, x='Lattitude', fill=True)
```

**LONGITUDE**

```python
df['Longtitude'].interpolate(method='spline', order=2, inplace=True)
df['Longtitude'].isnull().sum()
```

```python
sns.displot(data, kind='kde', x='Longtitude')
sns.displot(df, kind="kde", x='Longtitude')
```

**Bedroom**

```python
df['Bedroom2'].describe()
```

```python
df['Bedroom2'].fillna(method='bfill', inplace=True)
df.isnull().sum()
```

**Bathroom**

```python
df['Bathroom'].describe()
```

```python
df['Bathroom'].fillna(method='ffill', inplace=True)
df.isnull().sum()
```

**car**

```python
df['Car'].describe()
```

```python
df['Car'].interpolate(method='linear', inplace=True)
df['Car'].isnull().sum()
```

**Landsize**

```python
df['Landsize'].describe()
```

```python
df['Landsize'].fillna(df['Landsize'].mean(), inplace=True)
df['Landsize'].isnull().sum()
```

**BuildingArea**

```python
df['BuildingArea'].describe()
```

```python
df['BuildingArea'] = rou.fit_transform(df[['BuildingArea']])
sns.boxplot(df, x='BuildingArea')
```

```python
df['BuildingArea'].fillna(method='bfill', inplace=True)
```
