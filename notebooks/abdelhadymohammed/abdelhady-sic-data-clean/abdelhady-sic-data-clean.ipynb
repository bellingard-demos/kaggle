## housing_median_age

we have 258 missing values in this column

```python
data['housing_median_age'] = data['housing_median_age'].fillna(data['housing_median_age'].mean())
```

```python
data['housing_median_age'].isnull().sum()
```

## total_bedrooms

we have 4882 missing values in this column

```python
data['total_bedrooms'] = data['total_bedrooms'].fillna(data['total_bedrooms'].median())
```

```python
data['total_bedrooms'].isnull().sum()
```

## population 

we have 44 missing values in this column

```python
data.dropna(subset=['population'], inplace=True)
```

```python
data['population'].isnull().sum()
```

## households


```python
data['households'] = data['households'].replace('no', np.nan)
```

```python
data['households'] = pd.to_numeric(data['households'])
data['households'] = data['households'].fillna(data['households'].max())
```

```python
data['households'].isnull().sum()
```

## median_income   

```python
data['median_income'] = data['median_income'].ffill()
```

```python
data['median_income'].isnull().sum()
```

## gender      

```python
data['gender'].fillna('Unknown', inplace=True)
```

```python
data['gender'].isnull().sum()
```
