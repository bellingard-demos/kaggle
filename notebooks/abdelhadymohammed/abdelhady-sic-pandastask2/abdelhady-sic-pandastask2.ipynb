1. **what is the number of flights per class**

```python
flights_class = df['class'].value_counts()
flights_class
```

```python
plt.pie(flights_class, labels = flights_class.index, explode=(0.05,0.01) ,autopct='%0.1f%%')
plt.show()
```

*Conclusion: economy class tickets are more than double of buisness class*

2. **Distribution of ticket prices**

```python
df['price'].agg(['max', 'min', 'std', 'mean'])
```

```python
price = sns.displot(x=df['price'], data=df, kde=True)
price.fig.set_figwidth(10)
price.fig.set_figheight(6)
```

*Conclusion: the tickets in range 4000 to 15000 are the most sold*


3. **What's the total number of flights per airline**

```python
df['airline'].value_counts()
```

```python
sns.countplot(x=df['airline'], order=df['airline'].value_counts().index)
```

*Conclusion: Vistara airline has the most number of flights*

4. **average price by airline**

```python
avg = df.groupby('airline')['price'].mean()
avg
```

```python
plt.figure(figsize=(15,7))
sns.boxplot(data=df, x='airline',y='price')
```

*conclusion: Vistara airline flights are more expensive than others *

5. **How is the price affected when tickets are bought less than 2 days before departure?**

```python
avg_prices = df.groupby('days_left')['price'].mean()
avg_prices
```

```python
avg_prices = df.groupby('days_left')['price'].mean()

plt.figure(figsize=(10, 6))
plt.plot(avg_prices.index, avg_prices.values, marker='o')
plt.xlabel('Days Left Before Departure')
plt.ylabel('Average Price')
plt.show()
```

*conclusion: tickets sold in two days or less before departure are more expensive than the tickets sold in the rest*

6. **How does the ticket price vary between Economy and Business class?**

```python
df.groupby('class')['price'].mean()
```

```python
sns.kdeplot(data = df, x='price', hue = 'class')
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='class', y='price')
plt.xlabel('class')
plt.ylabel('price')
plt.show()
```

*conclusion: the buisness class tickets are much more expensive than economy class*
