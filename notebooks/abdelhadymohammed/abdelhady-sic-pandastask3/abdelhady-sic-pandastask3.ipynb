```python
def kdeplot(feature):
    plt.figure(figsize=(9, 4))
    plt.title(f"KDE for {feature}")
    ax0 = sns.kdeplot(df[df['Churn'] == 0][feature], color= 'navy', fill=True,label= 'Churn: No')
    ax1 = sns.kdeplot(df[df['Churn'] == 1][feature], color= 'orange', fill=True,label= 'Churn: Yes')
    plt.legend(['Not','Churn'])

kdeplot('tenure')
kdeplot('MonthlyCharges')
kdeplot('TotalCharges')
```

* gender does not affect churn rate
* Senior citizen are much more probable to churn than customers who are not Senior citizens
* Customers who do not have partners or do not have dependents are more probable to churn
* Customers opting for Electronic payment are more probable churn.
* Short term contract customers are more likely to churn.
* Customers with Paperless Billing are more probable to churn.
* Customers with Fiber Optics internet service have high churn rate
* customers who have total charges of less 2000 have very high churn rate
* Customers of tenure less than 20 are more likely to churn

18. **We want to give offers according to the monthly charge categories, can you explain that? Using Visulization**

```python
def offer(row):
    if row['Contract'] == "Month-to-month" and row['SeniorCitizen'] == 1:
        return("Flex_offer")
    elif row['Contract'] == "Month-to-month" and row['SeniorCitizen'] == 0:
        return("small offer")
    elif row['Contract'] == "Two year" and row['SeniorCitizen'] == 1:
        return("coin_offer")
    elif row['Contract'] == "Two year" and row['SeniorCitizen'] == 0:
        return("coin_flex_offer")
    elif row['Contract'] == "One year" and row['SeniorCitizen'] == 1:
        return("2X")
    elif row['Contract'] == "One year" and row['SeniorCitizen'] == 0:
        return("1.5X")
    else:
        return "No_offer"
```

```python
df['offer'] =df.apply(offer , axis = 1)
```

```python
df['offer'].value_counts()
```

```python
df['offer'].value_counts().plot(kind='pie', autopct='%1.2f%%', explode=(0,0.02,0.02,0,0.01,0.03))
```

***Observation***
* majority of customers who have short term contract not senior citizens have small offer عشان ميبقاش خراب بيوت

19. **We want to present offers by gender and the Senior Citizen, could you explain that? Using Visulization**

```python
sns.kdeplot(data=df, x='MonthlyCharges', hue='gender')
```

```python
def get_offers(df):
    categories = [
        {'category': 'Medium', 'min_charge': 0, 'max_charge': 50, 'offer': 'Offer 1'},
        {'category': 'Low', 'min_charge': 50, 'max_charge':100, 'offer': 'Offer 2'},
        {'category': 'High', 'min_charge': 100, 'max_charge': np.inf, 'offer': 'Offer 3'}
    ]

    df['category'] = ''
    df['tooffer'] = ''

    for category in categories:
        call = (df['MonthlyCharges'] >= category['min_charge']) & (df['MonthlyCharges'] < category['max_charge'])
        df.loc[call, 'category'] = category['category']
        df.loc[call, 'tooffer'] = category['offer']

    return df
```

```python
offers = get_offers(df)
df['tooffer'].value_counts()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=offers, x='category')
plt.title("Offers")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()
```

***Observation***
* Customers of Offer 2 have Lowest offer
