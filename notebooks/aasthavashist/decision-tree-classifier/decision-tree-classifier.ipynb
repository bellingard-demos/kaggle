# Decision Tree Classifier

```python
#importing model and dataset
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
```

```python
iris = pd.read_csv("../input/iris/Iris.csv")
iris.head()
```

```python
iris.count()
```

```python
Data =iris.iloc[:,:-1]
Label =iris.iloc[:, -1]
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(Data, Label ,test_size=0.3, random_state=5)
```

# Trainning The Model

```python
model = DecisionTreeClassifier(criterion = "entropy", random_state = 0)
```

```python
our_model = model.fit(X_train,Y_train)
```

```python
#prediction
pred=our_model.predict(X_test)
print(pred)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn import tree
```

```python
print("ACCURACY OF THE MODEL: ", accuracy_score(Y_test, pred))
print("Confusion Matrix: ",confusion_matrix(Y_test, pred))
print("Report : ",classification_report(Y_test, pred))
tree.plot_tree(model, filled=True)
```
