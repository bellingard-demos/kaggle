# **Series and DataFrame**


### **Series**

A **Series** is a one dimensional labelled array of a single data type in pandas. It can hold data of any data type such as integers, floats, strins or even python object. <br>
We can create a series as follows -

```python
data = [12, 45, 22, 37]
series = pd.Series(data)
series
```

By default the index values are 0,1,2,and so on. We can change the index values as follows -

```python
index = ['A', 'B', 'C', 'D']
series = pd.Series(data, index = index)
series
```

### **DataFrame**

A **DataFrame** is a 2 dimensional table where each column contains the data of same type. In simple words, it is a collection of series that share the same index labels. <br>
There are various methods to make a dataframe. Some are as follows -

**Creating DataFrame using list**

```python
data = [25, 30, 35]
df = pd.DataFrame(data, columns = ['Age'])
df
```

**Creating DataFrame from list of list**

```python
data = [['Alice',25], ['Bob',30], ['Charlie',35]]
df = pd.DataFrame(data, columns = ['Name', 'Age'])
df
```

**Creating DataFrame from dict of narray/lists**

```python
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 35]}
df = pd.DataFrame(data)
df
```

**Creating a DataFrame by proving index label explicitly.**

```python
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 35]}
index = ['Person 1', 'Person 2', 'Person 3']
df = pd.DataFrame(data, index = index)
df
```

**Creating Dataframe from list of dicts**

```python
data = [{'Person 1': 'Alice', 'Person 2': 'Bob', 'Person 3': 'Charlie'},
        {'Person 1': 25, 'Person 2': 30, 'Person 3': 35}]
df = pd.DataFrame(data)
df
```

**Creating dataframe from series**

```python
data = pd.Series([25, 30, 35])
df = pd.DataFrame(data)
df
```

**Creating DataFrame from Dictionary of series.**

```python
data = {'Name': pd.Series(['Alice', 'Bob', 'Charlie']), 
       'Age': pd.Series([25, 30, 35])}
df = pd.DataFrame(data)
df
```

**Creating DataFrame using zip() function.**

```python
Name = ['Alice', 'Bob', 'Charlie']
Age = [25, 30, 35]
list_of_tuples = list(zip(Name, Age))
df = pd.DataFrame(list_of_tuples, columns=['Name', 'Age'])
df
```

# **Sorting**

Sorting your dataframe is one of the basic step we perform to get a better understanding of the data we are working with. <br>
Sorting can be easily performed using **.sort_values() method** (https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.sort_values.html) .It can be used to sort data along either axis (columns or rows). Mostly, we sort our dataframe rows on the basis of one or more columns (it is similar to how we sort in a spreadsheet).<br>
There is one more function commonly used for sorting the index values or assigning custom index values. It is **.sort_index() method** (https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.sort_index.html) . Unlike sort_values(), here we sort on the basis of row index or column names and not by the values.

**Sorting in Ascending order**

```python
# By default, sort_values sort in ascending order. Here we are sorting the values using the Age column.
df.sort_values('Age')
```

**Sorting in Descening Order**

```python
# To sort in descending order, we will put the parameter ascending as False
df.sort_values('Age', ascending = False)
```

**Type of sorting**<br>
As you might be aware, there are various sorting algorithms available. By default, sort_values() uses quick sort, but there are other algorithms also available such as heap sort, merge sort and stable. We can select the most appropriate one according to our need.

```python
# Let's use heap sort to sort in ascending order
df.sort_values('Age', kind = 'heapsort')
```

**Sorting on the basis of multiple columns**<br>
When sorting on the basis of multiple columns, we can decide on how we want to sort according to each row. That is we can sort the dataframe in ascending order according to one column and in descending order using other column.

```python
# For example, we want to see the largest area available at same price. So we can sort price in ascending order and area in descending order
sorted_df = df.sort_values(by = ['Age', 'Sleep Duration'], ascending = [True, False])
sorted_df
```

**Sort Index**<br>
You must have noticed that when sorting our dataframe, the index values get jumbled. To get the dataframe back to original order, we can use sort_index() method.

```python
# Default is ascending but we can also sort in descending order by putting ascending as False
sorted_df.sort_index()
```

**Assigning custom index values**

```python
new_index_df = sorted_df.set_index(['Gender', 'Occupation'])
new_index_df
```

Here, we have set the index to furnishing status and air conditioning. This type of indexing is called hierarchical indexing or Multi-index.

```python
# sorting this new index based dataframe
new_index_df.sort_index()
```

# **Indexing and Retrieving Data**
Knowing how to index and retrieve data is highly useful in answering few questions regarding the data. <br>
The basic syntax to index a dataframe is df[start:end:step]. Step indicates the number of steps to move forward. We can retrieve the data without using step. df[start:end]. In this case, the value of step is assumed to be 1. <br>
Another method to retrieve column is by mentioning the column name - df['Name']. In this, we can also add some logical condition. The result of such indexing is the DataFrame consisting only of the rows that satisfy the condition on the column - df[C(df['Name'])]. Here C is the consition.

```python
# Q) What is the proportion of people suffering from Sleep Apnea?
df['Sleep Disorder'].value_counts()
```

```python
# Q) What is the average values for all numerical features for people suffering from Sleep Apnea?
import numpy as np
df.select_dtypes(include = np.number)[df['Sleep Disorder'] == 'Sleep Apnea'].mean()
```

```python
# What is the average quality sleep do people with sleep apnea have?
df[df['Sleep Disorder'] == 'Sleep Apnea']['Quality of Sleep'].mean()
```

**Slicing**

```python
# Retrieving values of index 0 to 5 and columns lableled from Age to Sleep Duration
df.loc[0:5, "Age": "Sleep Duration"]
```

```python
# Instead of column names we can also use column number
df.iloc[0:5, 2:5]
```

# **Grouping**
Group by method is used for grouping. It is yet another useful method to draw inference from the dataset by applying some functions. It is similar to grouping used in SQL.<br>
In python, group by method includes 3 steps -
1. Split: Divide the tables into groups
2. Apply: apply some operations
3. Combine: combine the result



```python
# Lets apply corr() to find the correlation between various parameters for sleep disorder
df.groupby('Sleep Disorder')[['Age', 'Sleep Duration', 'Quality of Sleep', 'Stress Level']].corr()
```

# **Summary Table**
We can summarize our dataset using crosstab() method to see the distribution in context to two variables. Aside from this, we can also use pivot_table() which is similar to the pivot tables in excel.

```python
pd.crosstab(df["Sleep Disorder"], df["BMI Category"])
```

```python
df.pivot_table(['Age', 'Sleep Duration', 'Quality of Sleep', 'Stress Level'],['Sleep Disorder'], aggfunc='mean')
```
