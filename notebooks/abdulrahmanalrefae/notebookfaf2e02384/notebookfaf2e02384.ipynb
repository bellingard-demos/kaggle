```python
class ChessBoard:
    def __init__(self):
        self.board = np.ones((8,8,3))

        black = (0,0,0)
        white = (1,1,1)

        for i in range(8):
            for j in range(8):
                if (i%2==0 and j%2==0) or (i%2==1 and j%2==1):
                    self.board[i][j]=black
    
    def add_red(self, row, column):
        self.row1 = row
        self.column1 = column
        red = (1,.2,0)
        self.board[row][column] = red
    
    def add_blue(self, row, column):
        self.row2 = row
        self.column2 = column
        blue = (0,0,1)
        self.board[row][column] = blue
    
    def render(self):
        plt.imshow(self.board)
    
    def is_under_attack(self):
        # if red and blue are in the same column --> vertical
        # if red and blue are in the same row --> horizontal
        # if the sum of row and column positions for blue equal the sum of row and column positions for red --> diagonal
        # if row minus column positions for blue equal the minus of row and column positions for red --> diagonal
        if self.row1 == self.row2 or self.column1 == self.column2 or self.row1 - self.column1 == self.row2 - self.column2 or self.row1 + self.column1 == self.row2 + self.column2:
            return True
        else:
            return False
    
```

**Horizontal** 

```python
aa = ChessBoard()
aa.add_blue(0,7)
aa.add_red(7,7)
aa.render()
print(aa.is_under_attack())
```

**Vertical**

```python
bb = ChessBoard()
bb.add_blue(0,7)
bb.add_red(0,0)
bb.render()
print(bb.is_under_attack())
```

**Diagonal 1**

```python
cc = ChessBoard()
cc.add_blue(0,7)
cc.add_red(6,1)
cc.render()
print(cc.is_under_attack())
```

**Diagonal 2**

```python
dd = ChessBoard()
dd.add_blue(1,2)
dd.add_red(5,6)
dd.render()
print(dd.is_under_attack())
```

**Not Under Attack**

```python
ee = ChessBoard()
ee.add_blue(1,2)
ee.add_red(3,6)
ee.render()
print(ee.is_under_attack())
```
