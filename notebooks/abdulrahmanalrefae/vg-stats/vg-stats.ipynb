```python
df = pd.read_csv("../input/videogamesales/vgsales.csv")
df
```

## 1- Which company is the most common video game publisher?

```python
print(df['Publisher'].value_counts().idxmax())
print(df['Publisher'].value_counts().max())
```

## 2- What’s the most common platform?

```python
print(df['Platform'].value_counts().idxmax())
```

## 3- What about the most common genre?

```python
print(df['Genre'].value_counts().idxmax())
```

## 4- What are the top 20 highest grossing games?

```python
df.nlargest(20, 'Global_Sales')
```

## 5- For North American video game sales, what’s the median?

```python
df['NA_Sales'].median()
```

## 5a- Provide a secondary output showing ten games surrounding the median sales output.

```python
mid = int(len(df['NA_Sales'])/2)
df[mid-5:mid+5]
```

## 5b- Assume that games with same median value are sorted in descending order.

```python
val = df.sort_values(by='NA_Sales', axis=0, ascending = False)
mid = int(len(df['NA_Sales'])/2)
val[mid - 5: mid + 5]
```

## 6- For the top-selling game of all time, how many standard deviations above/below the mean are its sales for North America?

```python
naMean = df['NA_Sales'].mean()
naStd = df['NA_Sales'].std()
wiiNaSales = df.iloc[0]['NA_Sales']
output = (wiiNaSales - naMean) / naStd
output
```

## 7- The Nintendo Wii seems to have outdone itself with games. How does its average number of sales compare with all of the other platforms?

```python
print(df[df['Platform'] == 'Wii']['Global_Sales'].mean())
print(df[df['Platform'] != 'Wii']['Global_Sales'].mean())
```
