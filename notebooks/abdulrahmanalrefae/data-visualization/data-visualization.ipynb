## Gammas 
This dataset have shows data about gammas

```python
df_gammas
```

## Flights
When we visualize the data of flights we notice that it increased since 1950, and people started to trust airlines more and to travel by plains.

```python
sns.relplot(x='year', y= 'passengers', data = flights)
```

By comparing the number of passengers with the months, we notice that number of passengers increases in July and August due to the incease of tempreture so people decides to travel and spend time in other places.

```python
sns.barplot(x = 'month', y = 'passengers', data = flights)
print()
```

# Gammas
- The relationship between BOLD and time point arises from temporally sparse events.
- The relationship is dominated by the high-amplitude events.


```python
sns.relplot(x='timepoint', y= 'BOLD signal', data = gammas)
```

by visualizing relation between ROI and subject we find that all of the ROI have the same number of subjects

```python
sns.barplot(x = 'subject', y = 'ROI', data = gammas)
print()
```
