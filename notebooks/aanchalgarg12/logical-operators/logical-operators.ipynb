**And Operator**

```python
a = 3
b = -2
if a>0 and b>0 :                  #Both the conditions should be satisfied
    print("Positive")     
else :
    print("Not Positive")
```

****OR operator****

```python
a = 2
b = -2
if a>0 or b>0 :                     #Either first or the second condition should be satisfied.
    print("ok")
else:
    print("not ok")
```

**Not Operator**

```python
a = 2
b = 5
a > b
```

```python
a = 2
b = 5
not a > b
```

**And + Or operator in single statement**

```python
project_A = True
project_B = False
CGPA = 6.3
if project_A == True or project_B == True and CGPA > 6:
    print("Pass")
else : 
    print("Improve")
```

```python
project_A = True
project_B = False
CGPA = 5.2
if( project_A == True or project_B == True )and CGPA > 6 :  #use bracets according to the python precedence.
    print("Pass")
else : 
    print("Improve")
```

```python
project_A = False
project_B = False
CGPA = 6.3
if project_A == True or project_B == True and CGPA > 6:
    print("Pass")
else : 
    print("Improve")
```
