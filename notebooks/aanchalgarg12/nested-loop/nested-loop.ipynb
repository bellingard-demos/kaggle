```python
i = 1
while i <= 4:
    j = 1                                        #While loop
    while j <= 3:
        print(j, end = " ")
        j = j + 1
    i = i + 1
    print("\n")
```

```python

for i in range(1,5):
    j = 1                                           #For loop 
    while j <= 3:
        print(j, end = " ")
        j = j + 1
    i = i + 1
    print("\n")
```

**print a pattern**

```python
i = 1
while i <= 5:
    j = 1
    while j <= i :
        print("*", end = " ")
        j = j + 1
    i = i + 1
    print("\n")
```

**prime numbers between 1 - 50**

```python
i = 2
while i <= 50 :
    j = 2                         #range of number which we have to find whether they are prime or not.
    while j <= i//2+1:
        if i == 2:
            print(i)                     #max multiple we will check
        if i % j == 0:
            break
        if j == i//2+1:
            print(i)
        j = j + 1
    i = i + 1
```
