**type conversion - implicit**

```python
Revenue_A = 20
Revenue_B = 20.3
```

```python
type(Revenue_A)
```

```python
type(Revenue_B)
```

```python
Revenue = Revenue_A + Revenue_B
print(Revenue)
type(Revenue)      #Implicit conversion
```

```python
print("total revenue is", Revenue)
```

**Explicit conversion**

```python
a = input("sales of A store")
b = input("sales of B store")
```

```python
c = a+b       #concatenating a and b
print(c)
type(c)
```

```python
d = float(a)+float(b)    #adding by converting both the values into float.
                        #explicit conversion
print(d)
```
