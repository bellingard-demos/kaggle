**Elif statement**

```python
a = int(input("Enter the number"))
if a > 0 :
    print("The number is positive")
elif a < 0 :                             #elif work only if the the first condition prove false.
    print("The number is negative")
```

**elif and if together**


```python
a =int(input("Enter the first number"))
b = int(input("Enter the second number"))
if a > 0 :                                      #second statement is not get print because the first condition was true.
    print("The first number is positive")
elif b > 0 :
    print("The second number is positive")
```

```python
Place = input("Enter the place") 
Bestfriend = input("Is my bestfriend going?")
Sponsor = input("Are you Sponsoring my trip?")
if Place == "Hill Station":
    print("I may go!!")
    if Bestfriend == "yes":
        print("let's go!!")
        print("This is the inner if block")
elif Sponsor == "yes" :
    print("Definately going")
```

**Else**
