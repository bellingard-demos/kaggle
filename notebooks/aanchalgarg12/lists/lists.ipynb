**how many times a particular item in a list**

```python
l = [2,3,1,2,7,9]
num = int(input("Enter the number"))
count = 0

for i in l:
    if num == i :
        count = count + 1
print(num, "is occuring",count, "times")
```

**Remove occurence of a particular item form the list**

```python
b = [23,22,56,36,96,3,2,23]
num1 = int(input("Enter the number that you want to remove"))

for i in b:
    if i == num1:
        b.remove(i)
print(b)
```
