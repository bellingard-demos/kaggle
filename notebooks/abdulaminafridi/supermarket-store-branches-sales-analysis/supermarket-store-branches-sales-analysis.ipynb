```python
km = KMeans(n_clusters=4,init='k-means++',max_iter=300,random_state=42)
km.fit(train_df)
y_pred = km.predict(train_df)
df_clusters = pd.DataFrame(train_df,columns=data.columns)
df_clusters['clusters'] = y_pred
df_clusters.head(10)
```

```python
df_clusters['clusters'].value_counts()
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=16, random_state=0)
    kmeans.fit(train_df)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=4, init='k-means++', max_iter=300, n_init=10, random_state=0)
pred_y = kmeans.fit_predict(train_df)
plt.scatter(train_df[:,0], train_df[:,1])
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red')
plt.show()
```
