# Discover and prepare data for our EDA process

```python
imdb=pd.read_csv("/kaggle/input/imdb-5000-movies-multiple-genres-dataset/IMDb_All_Genres_etf_clean1.csv")
```

## We removing "$" and "M stands for million" from 'Total_Gross' column to be able to switch that column to float

```python
Total_Gross=[]
for value in imdb.Total_Gross:
   if value=="Gross Unkown":
    Total_Gross.append(np.nan)
   else:
    new_value= value.replace('$',"").replace('M',"")
    Total_Gross.append(new_value)
```

```python
imdb.Total_Gross=Total_Gross
imdb
```

```python
imdb.Total_Gross.astype("float")
```

# _**1-Most Rated Movies?**_

```python
most_rated_movies=imdb[['Movie_Title','Rating']].sort_values("Rating",ascending=False).head(20)
px.bar(x=most_rated_movies['Movie_Title'],y=most_rated_movies['Rating'],color_discrete_sequence=px.colors.sequential.Blackbody)
```

# _**2-Movies with the highest grossing total**_

```python
highest_grossing_total=imdb[['Movie_Title','Total_Gross']].sort_values("Total_Gross",ascending=False).head(30)
highest_grossing_total
fig = go.Figure(data=[go.Table(header=dict(values=['Movie_Title', 'Total_Gross']),
                 cells=dict(values=[highest_grossing_total.Movie_Title,"$"+highest_grossing_total.Total_Gross+"M"]))
                     ])
fig.show()
```

# _**3-The directors with the largest number of films**_

```python
highest_director=imdb.Director.value_counts().sort_values(ascending=False).head(10)
fig=px.bar(highest_director,orientation='h',color_discrete_sequence=px.colors.sequential.Blackbody)
fig.update_layout({
"plot_bgcolor": "rgba(0, 0, 0, 0)",
"paper_bgcolor": "rgba(0, 0, 0, 0)",
})
```

# _**4.1-Which genre has the most movies?**_

```python
highest_main_genre=imdb.main_genre.value_counts().sort_values(ascending=False).head(10)
px.pie(highest_main_genre,values=highest_main_genre.values,names=highest_main_genre.index, color_discrete_sequence=px.colors.sequential.YlOrBr_r)
```

# _**4.2-What are the most popular Genres of movies?**_

```python
highest_genre=imdb.groupby(['main_genre','side_genre']).size().sort_values(ascending=False)
highest_genre
```

# _**5-Top Censor Certification Across Movies**_

```python
# Top 20 - Rating Across Movies

highest_cesors = imdb['Censor'].value_counts().sort_values(ascending=False).head(10)
fig=px.bar(highest_cesors,color_discrete_sequence=px.colors.sequential.Blackbody)
fig.update_layout({
"plot_bgcolor": "rgba(0, 0, 0, 0)",
"paper_bgcolor": "rgba(0, 0, 0, 0)",
})
```

# _**6-How long are most movies?**_

```python

most_durations=imdb['Runtime(Mins)'].value_counts().sort_index()
px.line(most_durations,color_discrete_sequence=px.colors.sequential.Blackbody)
```

<div style="  background-color: #e7f3fe;

# _**7-How is the production of movies changed over the years?**_

```python
years=imdb.Year	.value_counts().sort_index(ascending=False).head(30)
px.line(years,color_discrete_sequence=px.colors.sequential.Blackbody)
```
