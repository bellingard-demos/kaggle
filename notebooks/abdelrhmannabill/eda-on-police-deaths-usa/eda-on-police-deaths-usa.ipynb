### as we know from the description of the dataset that "K9_unit" is boolean type and year should be string (object)

```python
df["K9_Unit"]=df["K9_Unit"].astype("boolean")
df["Year"]=df["Year"].astype("object")
```

```python
df.info()
```

### Function works for Bar plot with values above each bar

```python
def bar_with_values_on_the_top(dataframe):
    index=dataframe.keys()
    values=dataframe.values
    plt.bar(index,values, color ='maroon',width = .3)
    plt.rcParams['xtick.major.pad']='1'
    for index, value in enumerate(dataframe):
        plt.text(index-.2, value+20,str(value))
    plt.show()
```

### see a summary of our dataset

```python
df.describe(include=["object"])
```

# from the summary we have made , we see that:
#### 1- the highest rank of death is "Patrolman" rank 

```python
highest_rank=df["Rank"].value_counts().head(5)
bar_with_values_on_the_top(highest_rank)
```

### 2-the highest Cause of Death is Gunfire 

```python
highest_cause=df['Cause_of_Death'].value_counts().head()

plt.pie(highest_cause,labels=highest_cause.keys(),autopct='%1.1f%%')
plt.xlabel("Cause Of Death")
```

### 3-the highest year is  2021

```python
killings_in_the_year = {}

for year in df['Year']:
    if year in killings_in_the_year:
        killings_in_the_year[year] += 1
    else:
        killings_in_the_year[year] = 1
```

```python
years = []
kills = []
for key in killings_in_the_year:
    years.append(key)
    kills.append(killings_in_the_year[key])
plt.plot(years, kills)
plt.title("Police deaths since 1791")
plt.ylabel('Deaths')
plt.show()
plt.clf()
```

```python
last_years = years[-23:]
last_kills = kills[-23:]

plt.plot(last_years, last_kills)
plt.title("Police deaths since 2000")
plt.ylabel('Deaths')
plt.grid(visible=True)
plt.show()
plt.clf()
```

### 4- the highest month of death is September

```python
highest_month=df['Month'].value_counts().head()
plt.pie(highest_month,labels=highest_month.keys(),autopct='%1.1f%%')
plt.xlabel("Month")
```

### 5- highest day in killing is saturday

```python
highest_day=df['Day'].value_counts().head()
bar_with_values_on_the_top(highest_day)
```

### 6-the highest state of death is Texas

```python
highest_state=df['State'].value_counts().head()
bar_with_values_on_the_top(highest_state)
```
