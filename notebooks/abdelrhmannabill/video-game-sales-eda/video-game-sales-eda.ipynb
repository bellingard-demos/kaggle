# 1-Which platform has the most releases?

```python
highest_platform=vg_sales['Platform'].value_counts()
px.bar(highest_platform.head(10))
```

# 2-Which Genre has the most releases?

```python
highest_genre=vg_sales['Genre'].value_counts()
px.bar(highest_genre.head(10),text_auto='.4s')
```

# 3-Which publisher has the most releases ?

```python
highest_publisher=vg_sales['Publisher'].value_counts()
px.bar(highest_publisher.head(15),orientation='h' )
```

# 4-is there a relation between specific  (platform & publisher)?

```python
re=vg_sales.groupby(['Platform','Publisher']).Global_Sales.sum()
re.sort_values().tail(20)
```

### wii & nintendo made the highest sells 
### after that came DS & nintendo 
#### we can conclude that Nintendo is powerful publisher even though in releasing games it's not number one according to Question number 3 

# 5-How is the publishing of video games in the market changed over the years?

```python
highest_year=vg_sales['Year'].value_counts().head(15).sort_index()
px.line(highest_year,markers=True)
```

# 6-The difference between JP, NA, EU, and Other places in sales by years?

```python
sales_by_na=vg_sales.groupby('Year').NA_Sales.sum()
sales_by_eu=vg_sales.groupby('Year').EU_Sales.sum()
sales_by_jp=vg_sales.groupby('Year').JP_Sales.sum()
sales_by_other=vg_sales.groupby('Year').Other_Sales.sum()

fig = px.line(y=[sales_by_na.values,sales_by_eu.values,sales_by_jp.values,sales_by_other.values],x=sales_by_na.index,markers=True,title='Sales by region')
newnames = {'wide_variable_0':'NA','wide_variable_1':'Europe','wide_variable_2':'Japan','wide_variable_3':'Other'}
fig.for_each_trace(lambda t: t.update(name = newnames[t.name],
                                      legendgroup = newnames[t.name],
                                      hovertemplate = t.hovertemplate.replace(t.name, newnames[t.name])
                                     )
                  )
```

#  7-Difference between JP, NA, EU, and other places in sales with respect to genre?

```python
genre_jp=vg_sales.groupby('Genre').JP_Sales.sum()
genre_eu=vg_sales.groupby('Genre').EU_Sales.sum()
genre_na=vg_sales.groupby('Genre').NA_Sales.sum()
genre_other=vg_sales.groupby('Genre').Other_Sales.sum()

fig = px.line(x=genre_eu.index,y=[genre_jp.values,genre_eu.values,genre_na.values,genre_other.values],markers=True)
newnames = {'wide_variable_0':'Japan','wide_variable_1':'Europe','wide_variable_2':'North America','wide_variable_3':'Other'}
fig.for_each_trace(lambda t: t.update(name = newnames[t.name],
                                      legendgroup = newnames[t.name],
                                      hovertemplate = t.hovertemplate.replace(t.name, newnames[t.name])
                                     )
                  )

fig.show()
```

#  8-Profits of every publisher in the four countries?

```python
all=vg_sales.groupby('Publisher').Global_Sales.sum()
all.sort_values(ascending=False)
fig=px.bar(all.sort_values(ascending=False).head(15))
fig.show()
```

# 9-Profits of every Game in the four countries?

```python
all=vg_sales.groupby('Name').Global_Sales.sum()
all.sort_values(ascending=False)
fig=px.bar(all.sort_values(ascending=False).head(15))
fig.show()
```
