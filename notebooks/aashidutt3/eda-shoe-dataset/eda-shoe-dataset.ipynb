```python
# Replace special characters 

df['Current_Price'] = df['Current_Price'].str.replace('‚Çπ', '')
df['Current_Price'] = df['Current_Price'].str.replace(',', '')
df['How_Many_Sold'] = df['How_Many_Sold'].str.replace(',', '')
```

# **<span style="color:#E64848;">ü•ΩLet's see what all we can explore in this dataset</span>**

1. Which Brand has max rating?

2. Which brand is sold the max?

3. Which is the most and least priced Product?

4. Which Brand sells costliest shoes?

## **<span style="color:#FEB139;">1Ô∏è‚É£Which Brand has max rating?</span>**

```python
max_rating = df['RATING'].max()
max_rating_brand = df.loc[df['RATING'] == max_rating, 'Brand_Name'].iloc[0]

print(max_rating_brand, ' has maximum rating of ', max_rating)
```

```python
# --------------------Rating VS Brand Name------------------------

plt.figure(figsize = (40,20))
sns.catplot(x = 'RATING',
            y = 'Brand_Name',
            kind = "bar",
            height = 7,
            aspect = 2,            
            data = df.head(100)).set(title = "Rating vs Brand Name");
plt.xticks(rotation = 60)
plt.show()
```

## **<span style="color:#FEB139;">2Ô∏è‚É£Which brand is sold the max?</span>**

```python
df['How_Many_Sold'] = df['How_Many_Sold'].astype(int)
```

```python
max_shoes_sold = df['How_Many_Sold'].max()
max_sold_brand = df.loc[df['How_Many_Sold'] ==  max_shoes_sold, 'Brand_Name'].iloc[0]

print(max_sold_brand, ' sold maximum ', max_shoes_sold, 'Shoes')
```

```python
min_shoes_sold = df['How_Many_Sold'].min()
min_sold_brand = df.loc[df['How_Many_Sold'] ==  min_shoes_sold, 'Brand_Name'].iloc[0]

print(min_sold_brand, ' sold least', min_shoes_sold, 'Shoes')
```

```python
# --------------------Shoes sold VS Brand Name------------------------

plt.figure(figsize = (40,20))
sns.catplot(x = 'How_Many_Sold',
            y = 'Brand_Name',
            kind = "bar",
            height = 7,
            aspect = 2,            
            data = df.head(100)).set(title = "Rating vs Brand Name");
plt.xticks(rotation = 60)
plt.show()
```

## **<span style="color:#FEB139;">3Ô∏è‚É£Which is the most and least priced Product?</span>**

```python
df['Current_Price'].unique()
```

```python
df.dropna(axis = 0, inplace = True)
```

```python
df['Current_Price'] = df['Current_Price'].astype(int)
```

```python
min_shoes_price = df['Current_Price'].min()
min_priced_product = df.loc[df['Current_Price'] ==  min_shoes_price, 'Product_details'].iloc[0]

print(min_priced_product, ' are the least priced shoes, priced at - $',min_shoes_price )
```

```python
max_shoes_price = df['Current_Price'].max()
max_priced_product = df.loc[df['Current_Price'] ==  max_shoes_price, 'Product_details'].iloc[0]

print(max_priced_product, ' are the least priced shoes, priced at - $',max_shoes_price )
```

```python
# --------------------Product VS Price------------------------

plt.figure(figsize = (40,20))
sns.catplot(x = 'Current_Price',
            y = 'Product_details',
            kind = "bar",
            height = 7,
            aspect = 2,            
            data = df.head(100)).set(title = "Product vs Price");
plt.xticks(rotation = 60)
plt.show()
```

## **<span style="color:#FEB139;">4Ô∏è‚É£Which Brand sells costliest shoes?</span>**

```python
max_shoes_price = df['Current_Price'].max()
max_priced_brand = df.loc[df['Current_Price'] ==  max_shoes_price, 'Brand_Name'].iloc[0]

print(max_priced_brand, ' have the most priced shoes at - $',max_shoes_price )
```
