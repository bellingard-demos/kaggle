## **<span style="color:#6CB4EE;">1Ô∏è‚É£ Passengers travel to which city (Arrival city) more often from Banglore?</span>**

```python
import matplotlib.pyplot as plt
```

```python
most_travelled = df['Arrival_City'].value_counts()
most_travelled
```

```python
plt.figure(figsize=(15,8))
most_travelled.plot(kind = 'bar',color = my_colors[0])
plt.show()
```

## **<span style="color:#6CB4EE;">2Ô∏è‚É£ Which flight is usually ariving late?</span>**

```python
late_arrival = df['Flight_Time_status'].value_counts()
late_arrival
```

```python
plt.figure(figsize= (8,8))
late_arrival.plot(kind = 'bar',color = my_colors[1])
plt.show()
```

## **<span style="color:#6CB4EE;">3Ô∏è‚É£ What is the maximum delayed time for a flight?</span>**

```python
Max_delayed_time = df['Flight_delay_by_time'].value_counts()
Max_delayed_time
```

**<span style="color: #FBCB0A;">From above we can infer that most of the time, flights are delayed max by about 16 minutes.</span>**

## **<span style="color:#6CB4EE;">4Ô∏è‚É£ Which Airlines has most flights ariving at Banglore?</span>**

```python
most_airlines = df['Airlines_Name'].value_counts()
most_airlines
```

```python
plt.figure(figsize= (15,8))
most_airlines.plot(kind = 'barh',color = my_colors[2])
plt.show()
```

```python
plt.figure(figsize= (15,8))
most_airlines.plot(kind = 'pie')
plt.show()
```

**<span style="color: #FBCB0A;">Indigo Airlines have most flights going in and out Banglore. Precisely, because Banglore is the IT hub and people from across India work there. Hence, most of the people travel to and out of Banglore go to various cities accross India as compared to International flights by Airlines like KLM or Turkish Airlines. Thus, people prefer Indigo Ailines the most</span>** 

## **<span style="color:#6CB4EE;">5Ô∏è‚É£ Which flight took most time to cover and from where?</span>**

```python
Max_time = df['Flight_time'].max()
Max_time_flight = df.loc[df['Flight_time'] == Max_time, 'Arrival_City'].iloc[0]
print("Flight from ", Max_time_flight, " took maximum time of ", Max_time, " to reach Banglore.")
```

## **<span style="color:#6CB4EE;">6Ô∏è‚É£ Which Airlines travel maximum distance?</span>**

```python
sns.catplot(x = 'Airlines_Name',
            y = 'Flight_distance',
            kind = "strip",
            height = 7,
            aspect = 2,            
            data = df).set(title = "Airlines vs Flight Distance");
```

```python
#Which airlines flight is delayed my maximum time
plt.figure(figsize= (15,8))
sns.stripplot(x="Flight_delay_by_time", y="Airlines_Name", data=df, jitter=0.05).set(title = "Airlines vs Delayed time");
```

**<span style="color: #FBCB0A;">IndiGo flights are most frequent travellers to Banglore and thus have highest delay time as well.</span>**
