## **<span style="color:#B56576;"> 1Ô∏è‚É£ When was most movies released ?</span>**

```python
# When was most Movies released
max_release = df['Release_Date'].value_counts()
max_release
```

```python
plt.figure(figsize=(15,8))
year_club = df.groupby(['Release_Date', 'Title']).Title.size().head(20).plot(kind = 'barh',color = my_colors[2])
plt.title('Titles Vs their Release Dates')
plt.legend()
plt.show()
```

**<span style="color:#E56B6F;">Most movie releases took place on 10 Sept 2021, followed by 12 Feb 2021.</span>**

## **<span style="color:#B56576;"> 2Ô∏è‚É£ Relation between Genre and Titles</span>**

```python
plt.figure(figsize=(15,8))
year_club = df.groupby(['Genre', 'Title']).Genre.size().head(20).plot(kind = 'barh',color = my_colors[3])
plt.title('Genre VS Titles')
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (40,20))
sns.catplot(x = 'Genre',
            y = 'Title',
            kind = "strip",
            height = 7,
            aspect = 2,            
            data = df.head(100)).set(title = "Genre vs Title");
plt.show()
```

## **<span style="color:#B56576;"> 3Ô∏è‚É£ Movie Release Date VS Titles</span>**

```python
plt.figure(figsize= (15,8))
sns.stripplot(x="Release_Date", y="Title", data=df.head(50), jitter=0.5).set(title = "Release Date vs Title");
```

## **<span style="color:#B56576;"> 4Ô∏è‚É£ Which Genre movies are most prefered?</span>**

```python
from wordcloud import WordCloud

genre = df['Genre']

# Generate a word cloud 
wordcloud1 = WordCloud(collocation_threshold = 2,
                      collocations= True,
                      colormap="afmhot").generate(' '.join(df['Genre'].astype(str)))

# Generate plot
plt.figure(figsize = (15,8))
plt.imshow(wordcloud1, interpolation="bilinear")
plt.axis("off")
plt.show()
```

## **<span style="color:#B56576;"> 5Ô∏è‚É£ Which movie title made big name?</span>**

```python
wordcloud2 = WordCloud(collocation_threshold = 2,
                      collocations= True,
                       colormap = 'RdYlGn'
                      ).generate(' '.join(df['Title']))

# Generate plot
plt.figure(figsize = (15,8))
plt.imshow(wordcloud2, interpolation="bilinear")
plt.axis("off")
plt.show()
```

## **<span style="color:#B56576;"> 6Ô∏è‚É£ Which Production Company made big name?</span>**

```python
wordcloud2 = WordCloud(collocation_threshold = 2,
                      collocations= True,
                       colormap="Set2"
                      ).generate(' '.join(df['Production company'].astype(str)))

# Generate plot
plt.figure(figsize = (15,8))
plt.imshow(wordcloud2, interpolation="bilinear")
plt.axis("off")
plt.show()
```
