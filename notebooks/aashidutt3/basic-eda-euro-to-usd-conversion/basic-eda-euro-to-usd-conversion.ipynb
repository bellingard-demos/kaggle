## **<span style="color:#957DAD;">👀 Visualization</span>**

```python
#Plots
df.hist(bins = 50, figsize= (20,15), color = '#C689C6')
```

## **<span style="color:#957DAD;">⚖ Comparisons</span>**

```python
# Open Value vs Close Value
plt.figure(figsize= (15,8))
sns.stripplot(x="Open", y="Close", data=df.head(100), jitter=0.5).set(title = "Open Value vs Close Value");
plt.show()
```

```python
plt.figure(figsize = (40,20))
sns.catplot(x = 'Open',
            y = 'Close',
            kind = "point",
            height = 7,
            aspect = 2,            
            data = df.head(100), color = '#E8A0BF').set(title = "Open Value vs Close Value");

plt.show()
```

```python
#Low Value vs High Value
plt.figure(figsize = (40,20))
sns.catplot(x = 'Low',
            y = 'High',
            kind = "point",
            height = 7,
            aspect = 2,            
            data = df.head(100), color = '#FFD124').set(title = "Low Value vs High Value");

plt.show()
```

```python
# Close price vs Adjusted close price
plt.figure(figsize = (40,20))
sns.catplot(x = 'Close',
            y = 'Adj Close',
            kind = "point",
            height = 7,
            aspect = 2,            
            data = df.head(100), color = '#E56B6F').set(title = "Close price vs Adjusted close price");

plt.show()
```

From the above patterns, we observe that almost all the parameters are propotional to each other. While we do see some dip in the open vs close or high vs low chart but the over all cheart is exponential i.e with increase in open values the close value mostly tend to increase as well and same applies for high vs low.

## **<span style="color:#957DAD;">🤝Correlation</span>**

```python
plt.figure(figsize = (15, 8), facecolor = "#F7F4F4")
sns.heatmap(df.corr(), annot = True, cmap = "summer");
```

The patterns shown in above charts set justified as seen in the correlation heatmap. There is a strong correlation among all attributes.
