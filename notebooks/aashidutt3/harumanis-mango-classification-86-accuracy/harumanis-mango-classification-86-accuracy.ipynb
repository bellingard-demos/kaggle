```python
# Imports
import sklearn
from sklearn import metrics
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
#Environment check
import os
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('/kaggle/input/harumanis-mango-physical-measurement/Data.csv')
```

```python
df.describe()
```

```python
df.info()
```

```python
df.head()
```

```python
df.columns
```

```python
df.drop('No',axis = 1, inplace = True)
```

```python
df['Grade'] = df['Grade'].str.replace('A', '0')
df['Grade'] = df['Grade'].str.replace('B', '1')
```

```python
df['Grade'] = df['Grade'].astype(int)
```

## **<span style="color:#cd486b;">ü™ì Let's split the dataset</span>**

```python
X = df.iloc[:,:-1]
y = df['Grade']
```

```python
X.shape
```

```python
y.shape
```

```python
#Check for imbalanced data
y.value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42, test_size = 0.2, stratify = y)
```

## **<span style="color:#cd486b;">üéé Define your Model</span>**

```python
model = RandomForestClassifier()

# tried n_estimators = 100, but did not make a great difference
```

```python
#Fit model
model.fit(X_train, y_train)
```

## **<span style="color:#cd486b;">üîÆ Predictions</span>**

```python
#Make predictions
predictions = model.predict(X_test)
```

```python
type(predictions[0])
```

## **<span style="color:#cd486b;">üî¢Metrics</span>**

## **<span style="color:#FFC300;">Mean Absolute Error</span>**

```python
#Mean absolute error

mae = mean_absolute_error(predictions, y_test)
print("Mean Absolute Error with  Random Forest classifier:" , mae)
```

## **<span style="color:#FFC300;">Accuracy</span>**

```python
#Accuracy

accuracy = metrics.accuracy_score(y_test,predictions)
print('Accuracy for Random Forest classifier model is - ', accuracy)
```

## **<span style="color:#FFC300;">Precision</span>**

```python
#Precision

precision_score = sklearn.metrics.precision_score(y_test, predictions, labels=model.classes_)
print("Precision score for Random Forest Classifier is  ", precision_score)
```

## **<span style="color:#FFC300;">Classification Report</span>**

```python
#Classification Report

clf_report = classification_report(y_test, predictions)
print(clf_report)
```

## **<span style="color:#FFC300;">Confusion Matrix</span>**

```python
#Confusion Matrix

cm = confusion_matrix(y_test, model.predict(X_test), labels=model.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                              display_labels=model.classes_)
disp.plot()
```

**From the above classification report and confusion matrix, we found that our model is performing well on both training and test sets. Also, the confusion matrix looks quite appropriate for a small dataset like used in this notebook.**

# **<span style="color:#cd486b;">ü§òConclusion</span>**
