## **<span style="color:#003892;"> 1Ô∏è‚É£ Correlation among all parameters</span>**

```python
plt.figure(figsize = (15, 8), facecolor = "#F7F4F4")
sns.heatmap(df.corr(), annot = True, cmap = "gist_heat");
```

**<span style="color:#00D7FF;">There is no corelation between any of these parameters.</span>**

## **<span style="color:#003892;"> 2Ô∏è‚É£ Find the number of categories and subcategories.</span>**

```python
no_of_categories = df.category.unique()
no_of_subcategories = df.sub_category.unique()
print("There are ", len(no_of_categories), " categories and ", len(no_of_subcategories), " sucategories")
```

```python
plt.plot(no_of_categories, color = my_colors[1])
```

```python
plt.figure(figsize=(15,8))
year_club = df.groupby(['category', 'sub_category']).category.value_counts().head(20).plot(kind = 'bar',color = my_colors[3])
plt.title('Items and their prices')
plt.legend()
plt.show()
```

## **<span style="color:#003892;"> 3Ô∏è‚É£ Which is the most priced product</span>**

```python
max_price = df['price'].max()
max_price_item = df.loc[df['price'] == max_price, 'items'].value_counts().plot(kind = 'barh', color = my_colors[2])
plt.title('Product with highest price')
plt.show()
```

```python
print("Most costly item on Jio Mart are Pinak stores Black Ear wired Earphones with mic having a price of Rs",max_price)
```

**<span style="color:#00D7FF;">Pinak stores Black Ear wired Earphones with mic are apparently the most costly product on Jio Mart.</span>**
