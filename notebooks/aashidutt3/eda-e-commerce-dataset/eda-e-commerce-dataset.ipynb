# **<span style="color:#FFC54D;">ü•ΩLet's see what all we can explore in this dataset</span>**

```python
#First lets check for Null values

df.isna().any()
```


**<span style="color:#9EB23B;">There are no Null values in our dataset.</span>**

## <span style="color:#9EB23B;">Before we move further, Let's drop or replace few features that we feel won't contribute much to our analysis - such as 
    
> Unnamed column

> 'Rs\n' from the MRP column
    
> 'Size:' from Sizes column    

>  Converting all prices into integer type.

</span>


```python
df.drop(columns = ['Unnamed: 0'], inplace = True)
```

```python
df.columns
```

```python
df.replace('Nan',np.nan, inplace = True)
```

```python
df.dropna(axis = 0, inplace = True)
```

```python
df['MRP'] = df['MRP'].str.replace('Rs\n', '')
df['Sizes'] = df['Sizes'].str.replace('Size:', '')
df['MRP'] = df['MRP'].astype(int)
df['SellPrice'] = df['SellPrice'].astype(int)
```

## **<span style="color:#BD4291;">1Ô∏è‚É£ Which brand is most costly / affordable to the customer?</span>**

```python
max_price = df['SellPrice'].max()
max_price_brand = df.loc[df['SellPrice'] == max_price, 'BrandName'].iloc[0]
max_category = df.loc[df['SellPrice'] == max_price, 'Category'].iloc[0]
print(max_price_brand ,  'is the costliest brand selling ', max_category, 'at Rs', max_price)
```

```python
df.plot.hist(max_category, bins=12, alpha=0.5)
plt.xlabel('prices')
plt.show()
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(df['MRP'],df['SellPrice'],hue=df['Category'])
```

```python
min_price = df['SellPrice'].min()
min_price_brand = df.loc[df['SellPrice'] == min_price, 'BrandName'].iloc[0]
min_category = df.loc[df['SellPrice'] == max_price, 'Category'].iloc[0]
print(min_price_brand ,  'is the most affordable brand selling ',min_category ,'at Rs', min_price)
```

```python
max_prices = df['SellPrice'].value_counts()
plt.figure(figsize=(15,8))
plt.plot(max_prices.head(100), '.', alpha=0.6, markersize=10, color=my_colors[4])
plt.xlabel('Selling prices')
```

```python
plt.figure(figsize=(15,8))
yearly_opening_rank = df.groupby(['SellPrice','BrandName']).BrandName.count().head(20).plot(kind = 'bar', color = my_colors[1])
plt.show()
```

**<span style="color:#9EB23B;">The above plots gives us an estimated range of clothes i.e between Rs 500 to 2500. While the plot in Yellow has slight lower range ofvalues certainly due to brands having Lingerie type of women wear.</span>**

## **<span style="color:#BD4291;"> 2Ô∏è‚É£ Which category has maximum options in women wear</span>**

```python
max_for_category = df['Category'].value_counts()
max_for_category
```

```python
plt.figure(figsize=(15,8))
plt.plot(max_for_category.head(100), '.', alpha=0.6, markersize=50, color=my_colors[4])
plt.plot(max_for_category.head(100), color=my_colors[1])
plt.xlabel('Categories')
```

**<span style="color:#9EB23B;">From above analysis, we can say that we have maximum clothings in Western wear section. While, items for Watches section remain very limited. </span>**

## **<span style="color:#BD4291;"> 3Ô∏è‚É£ How much Discount does each brand gives to customers?</span>**


> So, it's basic math right! 
> Profit = Selling Price - Cost Price which here is MRP.

```python
Profit = df['SellPrice'] - df['MRP']
print(Profit)
```

> Turns out almost every value is negative that means brands are selling at discounted price.

> Lets calculate discount for each brand

```python
discounted =  df['MRP'] - df['SellPrice'] 
plt.plot(discounted, '.', alpha=0.6, markersize=10, color=my_colors[4])
plt.xlabel('discounted prices')
```

```python
plt.figure(figsize = (15,8))
df['Discount'].value_counts().plot(kind='bar', color = my_colors[2])
```

**<span style="color:#9EB23B;">From the above plot, we can say that Most brands landed discounted prices in the range of Rs 10k to 20k tuning it in about max of 50% to min of 10%.</span>**

## **<span style="color:#BD4291;"> 4Ô∏è‚É£ How many option in sizes do women have?</span>**

```python
df.groupby(['Sizes','BrandName']).Sizes.count().head(20).plot(kind = 'bar', color = my_colors[1])
```

```python
plt.figure(figsize = (15,8))
df['Sizes'].value_counts().head(20).plot(kind='bar', color = my_colors[2])
plt.legend()
```

**<span style="color:#9EB23B;">As fashion industry is progressing, the brands are trying their best to fit into every customer size from small to XL sizes as well. Thus, from the above analysis it is clear that though the gap is shortneing but their is still high demand for sizes in Small,X-Small, Medium, Large and X-Large</span>**
