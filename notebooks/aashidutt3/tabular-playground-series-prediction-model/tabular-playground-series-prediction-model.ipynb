```python
train = pd.read_csv('/kaggle/input/tabular-playground-series-jan-2021/train.csv')
test = pd.read_csv('/kaggle/input/tabular-playground-series-jan-2021/test.csv')
```

```python
sub = pd.read_csv('/kaggle/input/tabular-playground-series-jan-2021/sample_submission.csv')
```

```python
train.head()
```

```python
train.drop("id", axis = 1, inplace = True)
```

```python
from sklearn.ensemble import RandomForestRegressor

forest_reg = RandomForestRegressor(max_depth = 2, n_estimators = 100,random_state = 42)
forest_reg.fit(train.drop("target", axis = 1), train.target)
```

```python
from sklearn.metrics import mean_squared_error
train_pred = forest_reg.predict(train.drop("target", axis = 1))
mean_squared_error(train_pred, train['target'])
```

```python
pred = forest_reg.predict(test.drop('id', axis = 1))
output = pd.DataFrame({'id':test['id'],'target':pred})
output.set_index('id',inplace=True)
output.to_csv('output.csv')
```
