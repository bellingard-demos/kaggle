## **<span style="color:#DA291C;">Which Menu Item has least and max Energy (kCal)?</span>**

```python
min_energy = df['Energy (kCal)'].min()
min_energy_item = df.loc[df['Energy (kCal)'] == min_energy, 'Menu Items'].iloc[0]
print("Menu item: ", min_energy_item, " has least energy of: ", min_energy, "kCal")
```

```python
max_energy = df['Energy (kCal)'].max()
max_energy_item = df.loc[df['Energy (kCal)'] == max_energy, 'Menu Items'].iloc[0]
print("Menu item: ", max_energy_item, " has most energy of: ", max_energy, "kCal")
```

## **<span style="color:#FFC72C;">Which item has min / max serve size?</span>**

```python
df['Per Serve Size'] = df['Per Serve Size'].map(lambda x: x.lstrip(' g').rstrip(' g'))
df['Per Serve Size'] = df['Per Serve Size'].map(lambda x: x.lstrip(' ml').rstrip(' ml'))
```

```python
df['Per Serve Size'].head()
```

```python
df['Per Serve Size'].unique()
```

```python
df['Per Serve Size'] = df['Per Serve Size'].astype(float)
```

```python
min_serve_size = df['Per Serve Size'].min()
min_serve_item = df.loc[df['Per Serve Size'] == min_serve_size, 'Menu Items'].iloc[0]
print("Menu item: ", min_serve_item, " has least serve size of: ", min_serve_size, "g")
```

```python
max_serve_size = df['Per Serve Size'].max()
max_serve_item = df.loc[df['Per Serve Size'] == max_serve_size, 'Menu Items'].iloc[0]
print("Menu item: ", max_serve_item, " has max serve size of: ", max_serve_size, "ml")
```
