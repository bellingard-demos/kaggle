# **<span style="color:#cd486b;">üìÅAbout the files</span>**

Collection of training data is performed with the help of IOT devices since the goal is to develop a AI based smoke detector device.
Many different environments and fire sources have to be sampled to ensure a good dataset for training. A short list of different scenarios which are captured:

Normal indoor
Normal outdoor
Indoor wood fire, firefighter training area
Indoor gas fire, firefighter training area
Outdoor wood, coal, and gas grill
Outdoor high humidity
etc.
The dataset is nearly 60.000 readings long. The sample rate is 1Hz for all sensors. To keep track of the data, a UTC timestamp is added to every sensor reading

## **<span style="color:#cd486b;">ü™ìSplit the Dataset</span>**

```python
X = df.iloc[:,:-1]
y = df['Fire Alarm']  
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42, test_size = 0.2)
```

## **<span style="color:#cd486b;">‚öñScaling Data</span>**

```python
from sklearn.preprocessing import MinMaxScaler
scale = MinMaxScaler()
X_train_scaled=pd.DataFrame(scale.fit_transform(X_train),columns=X_train.columns)
X_train_scaled
```

```python
X_test_scaled = pd.DataFrame(scale.fit_transform(X_test),columns=X_test.columns)
X_test_scaled
```

## **<span style="color:#cd486b;">üåüModel</span>**

```python
from sklearn import svm
model = svm.SVC()
```

```python
model.fit(X_train_scaled, y_train)
```

## **<span style="color:#cd486b;">üìÉErrors and Reports</span>**

```python
from sklearn.metrics import mean_absolute_error

predictions = model.predict(X_test_scaled)
mae = mean_absolute_error(predictions, y_test)

print("Mean Absolute Error with Support Vector Machine: {:,.0f}".format(mae))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay


cm = confusion_matrix(y_test, predictions, labels=model.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                              display_labels=model.classes_)
disp.plot()
```

```python
#Classification Report
from sklearn.metrics import classification_report

clf_report = classification_report(y_test, predictions)
print(clf_report)
```
