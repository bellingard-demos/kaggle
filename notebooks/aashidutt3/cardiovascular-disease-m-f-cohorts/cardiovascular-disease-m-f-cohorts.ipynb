**About Dataset:**

It is population weighted worldwide data on male and female cohorts ages 15-69 years including cardiovascular disease early death and associated dietary, metabolic and other risk factors.


The purpose of creating this formatted database is to explore the univariate and multiple regression correlations of cardiovascular early deaths and other health outcomes with risk factors.

Find Dataset [here](https://www.kaggle.com/ishadss/kkkkkk)
