## **<span style="color:#00165e;">ðŸ‘€Let's Analyse</span>**

## **<span style="color:#9c1203;">1. Which book was the best seller in UK? (Max. copies sold in UK)</span>**

```python
max_copies_uk = df['Copies Sold in UK'].max()
book_uk = df.loc[df['Copies Sold in UK'] == max_copies_uk, 'Book Name'].iloc[0]

print(book_uk, 'is the bestseller book in UK, selling over', max_copies_uk, 'Million copies')
```

```python
plt.figure(figsize=(15,8))
sns.barplot(x = df['Copies Sold in UK'], y = df['Book Name'])
plt.ylabel('Book Name')
plt.xlabel('Copies Sold in UK (in Million)')
plt.title('Number of copies sold for each book in UK')
plt.show()
```

## **<span style="color:#9c1203;">2. Which book was the best seller Worldwide? (Max. copies sold in Worldwide)</span>**

```python
max_copies_worldwide = df['Copies Sold WorldWide'].max()
book_worldwide = df.loc[df['Copies Sold WorldWide'] == max_copies_worldwide, 'Book Name'].iloc[0]

print(book_worldwide, 'is the bestseller book worldwide, selling over', max_copies_worldwide, 'Million copies')
```

```python
plt.figure(figsize=(15,8))
sns.barplot(x = df['Copies Sold WorldWide'], y = df['Book Name'])
plt.ylabel('Book Name')
plt.xlabel('Copies Sold WorldWide (in Million)')
plt.title('Number of copies sold for each book WorldWide')
plt.show()
```

## **<span style="color:#9c1203;">3. Timeline of books published</span>**

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x = df['Publish Year'], y = df['Book Name'],color = my_colors[0])
plt.ylabel('Book Name')
plt.xlabel('Publish Year')
plt.title('When were these books Published')
plt.show()
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x = df['Publish Date'], y = df['Book Name'], color = my_colors[1])
plt.ylabel('Book Name')
plt.xlabel('Publish Date')
plt.title('When were these books Published')
plt.show()
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x = df['Publish Month'], y = df['Book Name'],color = my_colors[2])
plt.ylabel('Book Name')
plt.xlabel('Publish Month')
plt.title('When were these books Published')
plt.show()
```

## **<span style="color:#9c1203;">4. Books Vs Page Count</span>**

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x = df['Pages'], y = df['Book Name'],color = my_colors[3])
plt.ylabel('Book Name')
plt.xlabel('Pages')
plt.title('Books Vs Page count')
plt.show()
```

## **<span style="color:#9c1203;">5. Books Vs Words</span>**

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x = df['Words'], y = df['Book Name'], color = my_colors[4])
plt.ylabel('Book Name')
plt.xlabel('Words')
plt.title('Books Vs Words')
plt.show()
```

## **<span style="color:#9c1203;">6. Audiobook durations</span>**

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x = df['Audiobook'], y = df['Book Name'], color = my_colors[0], s= 100)
plt.ylabel('Book Name')
plt.xlabel('Audiobook')
plt.title('Audiobook durations')
plt.show()
```

## **<span style="color:#9c1203;">7. Books and Awards won</span>**

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x = df['Total Awards Won'], y = df['Book Name'], color = my_colors[1])
plt.ylabel('Book Name')
plt.xlabel('Total Awards Won')
plt.title('Books and Awards won')
plt.show()
```
