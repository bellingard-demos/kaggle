# **<span style="color:#E64848;">ü•ΩLet's see what all we can explore in this dataset</span>**

> 1. Maximum income of people of the area

> 2. Maximum and minimum house age

> 3. Max and Min number of rooms

> 4. Average area population

> 5. Maximum and minimum price of a house

## **<span style="color:#554994;">1Ô∏è‚É£Maximum and minimum income of people</span>**


```python
income_min = df['Avg. Area Income'].min()
income_max = df['Avg. Area Income'].max()
print('Min income of people is - $', round(income_min),' while, Max income is - $', round(income_max))
```

```python
# --------------------Price VS Avg Area Income------------------------

plt.figure(figsize = (40,20))
sns.catplot(x = 'Price',
            y = 'Avg. Area Income',
            kind = "bar",
            height = 7,
            aspect = 2,            
            data = df.head(100)).set(title = "Price vs Area Income");
plt.xticks(rotation = 60)
plt.show()
```

## **<span style="color:#554994;">2Ô∏è‚É£Maximum and minimum house age</span>**


```python
age_min = df['Avg. Area House Age'].min()
age_max = df['Avg. Area House Age'].max()

print('Min age of house is -', round(age_min), 'Years. Max age of house is - ', round(age_max), 'Years')
```

```python
plt.figure(figsize = (15,8))
House_age = round(df['Avg. Area House Age'])
House_age.head(20).plot(kind = 'bar', color = my_colors[2])
plt.xticks(rotation = 0)
plt.show()
```

```python
# --------------------Price VS House Age------------------------

plt.figure(figsize = (40,20))
sns.catplot(x = 'Price',
            y = 'Avg. Area House Age',
            kind = "bar",
            height = 7,
            aspect = 2,            
            data = df.head(100)).set(title = "Price vs House Age");
plt.xticks(rotation = 60)
plt.show()
```

## **<span style="color:#554994;">3Ô∏è‚É£Max and Min number of rooms</span>**


```python
rooms_min = df['Avg. Area Number of Rooms'].min()
rooms_max = df['Avg. Area Number of Rooms'].max()

print('Min rooms per house are -', round(age_min), 'while, Max rooms per house are - ', round(age_max))
```

```python
# ------------------Price VS Number of Rooms----------------------------

plt.figure(figsize = (40,20))
sns.catplot(x = 'Price',
            y = 'Avg. Area Number of Rooms',
            kind = "bar",
            height = 7,
            aspect = 2,            
            data = df.head(100)).set(title = "Price vs Number of Rooms");
plt.xticks(rotation = 60)
plt.show()
```

## **<span style="color:#554994;">4Ô∏è‚É£Average area population</span>**


```python
area_population = df['Area Population']
plt.figure(figsize=(15,8))
plt.plot(area_population, color= my_colors[4])
```

```python
# ------------------Price VS Area Population----------------------------

plt.figure(figsize = (40,20))
sns.catplot(x = 'Price',
            y = 'Area Population',
            kind = "bar",
            height = 7,
            aspect = 2,            
            data = df.head(100)).set(title = "Price vs Area Population");
plt.xticks(rotation = 60)
plt.show()
```

## **<span style="color:#554994;">5Ô∏è‚É£Maximum and minimum price of a house</span>**


```python
price_min = df['Price'].min()
price_max = df['Price'].max()

print('Min price of house is $ -', round(price_min), 'while, Max price of house is - ', round(price_max))
```

```python
price = round(df['Price'])

plt.figure(figsize = (15,8))
price.head(20).plot(kind = 'bar', color = my_colors[3])
plt.xticks(rotation = 0)
plt.show()
```
