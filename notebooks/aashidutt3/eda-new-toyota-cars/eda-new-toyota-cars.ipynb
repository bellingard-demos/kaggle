## **<span style="color:#B270A2;">üßπData Cleaning</span>**

```python
df['price'].unique()
```

```python
df['price'] = df['price'].str.replace('$', '')
df['price'] = df['price'].str.replace(',', '')
```

```python
df['price'] = df['price'].str.replace('Not Priced', '0')
```

```python
#Drop all cars priced at $ 0
df = df[(df[['price']] != 0).all(axis=1)]
```

```python
df['price'].head()
```

```python
#Convert from Series object to int 
df['price'] = df['price'].astype(int)
```

```python
#df['reviews'].unique()
```

```python
df['reviews'] = df['reviews'].str.replace('(', '')
df['reviews'] = df['reviews'].str.replace(')', '')
df['reviews'] = df['reviews'].str.replace(',', '')
df['reviews'] = df['reviews'].str.replace('reviews', '')
```

```python
df['reviews'] = df['reviews'].str.replace(' ', '')
```

```python
df['reviews'] = df['reviews'].str.replace('review', '')
```

```python
#Convert from Series object to int 
df['reviews'] = df['reviews'].astype(int)
```

```python
df.head()
```

## **<span style="color:#9C9EFE;">1Ô∏è‚É£ What is the average price of a new car?</span>**

```python
average_price = df['price'].mean()
print("Average price of Toyota Cars is about $",average_price)
```

```python
plt.figure(figsize= (15,8))
plt.plot(df['price'], color = my_colors[2])
plt.show()
```

## **<span style="color:#9C9EFE;">2Ô∏è‚É£ Which car is the most costliest and most affordable?</span>**

```python
min_price = df['price'].min()
afford_car = df.loc[df['price'] == min_price, 'car names'].iloc[0]
print('The most affordable Toyota is', afford_car,'around - $', min_price)
```

```python
max_price = df['price'].max()
lux_car = df.loc[df['price'] == max_price, 'car names'].iloc[0]
print('The most priced Toyota is', lux_car,'around - $', max_price)
```

```python
plt.figure(figsize = (40,20))
sns.catplot(x = 'car names',
            y = 'price',
            kind = "bar",
            height = 7,
            aspect = 2,            
            data = df.head(30)).set(title = "Car vs Price");
plt.xticks(rotation=60)
plt.show()
```

## **<span style="color:#9C9EFE;">3Ô∏è‚É£ Which car has most and least reviews?</span>**

```python
min_review = df['reviews'].min()
min_review_car =df.loc[df['reviews'] == min_review, 'car names'].iloc[0]
print('The least reviewed Toyota is', min_review_car,' with ', min_review, ' reviews') 
```

```python
max_review = df['reviews'].max()
max_review_car =df.loc[df['reviews'] == max_review, 'car names'].iloc[0]
print('The most reviewed Toyota is', max_review_car,' with ', max_review , ' reviews') 
```

```python
plt.figure(figsize = (40,20))
sns.catplot(x = 'car names',
            y = 'reviews',
            kind = "bar",
            height = 7,
            aspect = 2,            
            data = df.head(30), color = '#E8A0BF').set(title = "Car vs Reviews");
plt.xticks(rotation=60)
plt.show()
```

## **<span style="color:#9C9EFE;">4Ô∏è‚É£ What would you say about a car having maximum an minimum rating?</span>**

```python
min_rate = df['ratings'].min()
low_rate_car = df.loc[df['ratings'] == min_rate, 'car names'].iloc[0]

print(low_rate_car, "has least rating of ", min_rate, " - Not Recommended to buy!")
```

```python
max_rate = df['ratings'].max()
high_rate_car = df.loc[df['ratings'] == max_rate, 'car names'].iloc[0]

print(high_rate_car, "has best rating of ", max_rate, " - Highly Recommended to buy!")
```

## **<span style="color:#9C9EFE;">5Ô∏è‚É£Let's check the average ratings of Toyota Cars</span>**


```python
avg_rate = df['ratings'].mean()
count = 0
for i in range(0, len(df['ratings'])):
    if (i >= avg_rate):
        count = count + 1
        
print('Average rating for Toyota cars is about ', round(avg_rate))
```

```python
# Check how many cars have rating above average rating

print("About", count, " out of ", len(df['ratings']),"Toyota Cars have 5 star ratings")
```

```python
plt.figure(figsize= (15,8))
sns.stripplot(x="car names", y="ratings", data=df.head(100), jitter=0.5).set(title = "Car vs Rating");
plt.xticks(rotation= 60)
plt.show()
```

**<span style="color:#E56B6F;"> Almost all cars have rating of 4 or above.</span>** 

## **<span style="color:#cd486b;">üåü Most Optimum Toyota Car üåü</span>**

```python
# Let's find the most optimum car

optimum_car_rating = df.loc[df['price'] == min_price, 'ratings'].iloc[0]
opti_review = df.loc[df['car names'] == afford_car, 'reviews'].iloc[0]

print(afford_car, 'is the most affordable car costing about $', min_price, ", having ",optimum_car_rating, "star rating and ", opti_review," reviews." )
```
