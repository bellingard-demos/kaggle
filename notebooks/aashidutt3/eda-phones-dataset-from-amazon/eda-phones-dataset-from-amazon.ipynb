**<span style="color:#957DAD;">We need to convert 2 columns - Rated_people and Offered_price_in_rupee from object into numerics (int or float) to use their values</span>**

```python
df['Rated_people'] = df['Rated_people'].str.replace(',', '')
df['Rated_people'] = df['Rated_people'].astype(int)
```

```python
df['Offered_price_in_rupee'] = df['Offered_price_in_rupee'].str.replace(',', '')
df['Offered_price_in_rupee'] = df['Offered_price_in_rupee'].astype(float)
```

# **<span style="color:#E0BBE4;">ü•ΩLet's see what all we can explore in this dataset</span>**

> 1. Which brand has highest rating?

> 2. Which phone has highest rating by people?

> 3. Highest and lowest priced brand

> 4. Highest and lowest discount

## **<span style="color:#957DAD;">1Ô∏è‚É£Which brand has highest rating?</span>**



```python
Brand_rating = df['ratting_out_of_five'].max()
brand = df.loc[df['ratting_out_of_five'] == Brand_rating, 'Brand'].iloc[0]
phone = df.loc[df['Brand'] == brand, 'phone_name'].iloc[0]

print(brand,phone, ' has highest rating of ', Brand_rating, 'Out of 5')
```

```python
Brand_rating_low = df['ratting_out_of_five'].min()
brand_low = df.loc[df['ratting_out_of_five'] == Brand_rating_low, 'Brand'].iloc[0]
phone = df.loc[df['Brand'] == brand_low, 'phone_name'].iloc[0]

print(brand_low,phone, ' has lowest rating of ', Brand_rating_low, 'Out of 5')
```

```python
# --------------------Brand VS Rating------------------------

plt.figure(figsize = (40,20))
sns.catplot(x = 'ratting_out_of_five',
            y = 'Brand',
            kind = "bar",
            height = 7,
            aspect = 2,            
            data = df.head(100)).set(title = "Brand VS Rating");
plt.xticks(rotation = 0)
plt.show()
```

## **<span style="color:#957DAD;">2Ô∏è‚É£Which phone has highest rating by people?</span>**



```python
Brand_rating = df['ratting_out_of_five'].max()
brand = df.loc[df['ratting_out_of_five'] == Brand_rating, 'Brand'].iloc[0]
phone = df.loc[df['Brand'] == brand, 'phone_name'].iloc[0]

print(phone, ' has highest rating of ', Brand_rating, 'Out of 5')
```

```python
# --------------------Phone VS Rating------------------------

plt.figure(figsize = (40,20))
sns.catplot(x = 'ratting_out_of_five',
            y = 'phone_name',
            kind = "bar",
            height = 10,
            aspect = 5,            
            data = df.head(50)).set(title = "Phone VS Rating");
plt.xticks(rotation = 0)
plt.show()
```

## **<span style="color:#957DAD;">3Ô∏è‚É£Highest and lowest priced brand</span>**



```python
price = df['Offered_price_in_rupee'].max()
brand_per_price = df.loc[df['Offered_price_in_rupee'] == price, 'Brand'].iloc[0]

print(brand_per_price, ' is the most priced brand with max price of Rs', price)
```

```python
price = df['Offered_price_in_rupee'].min()
brand_per_price = df.loc[df['Offered_price_in_rupee'] == price, 'Brand'].iloc[0]

print(brand_per_price, ' is the most priced brand with min price of Rs', price)
```

```python
# --------------------Brand VS Price------------------------

plt.figure(figsize = (40,20))
sns.catplot(x = 'Offered_price_in_rupee',
            y = 'Brand',
            kind = "bar",
            height = 7,
            aspect = 2,            
            data = df.head(100)).set(title = "Brand VS Price");
plt.xticks(rotation = 0)
plt.show()
```

## **<span style="color:#957DAD;">4Ô∏è‚É£Highest and lowest discount</span>**



```python
# Which brand offers highest and lowest discounts

discount = df['Percent_off'].min()
brand = df.loc[df['Percent_off'] == discount, 'Brand'].iloc[0]

print(brand, ' phones have least discount of ', discount, '%')
```

```python
# Which brand offers highest and lowest discounts

discount = df['Percent_off'].max()
brand = df.loc[df['Percent_off'] == discount, 'Brand'].iloc[0]

print(brand, ' phones have max discount of ', discount, '%')
```

```python
# --------------------Brands VS Discount------------------------

plt.figure(figsize = (40,20))
sns.catplot(x = 'Percent_off',
            y = 'Brand',
            kind = "bar",
            height = 7,
            aspect = 2,            
            data = df.head(100)).set(title = "Brand VS Discount");
plt.xticks(rotation = 0)
plt.show()
```
