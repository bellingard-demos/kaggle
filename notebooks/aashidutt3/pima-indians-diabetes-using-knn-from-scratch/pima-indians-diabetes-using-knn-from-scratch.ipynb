**Now we have our feature columns and target column, we can use split them into training and testing set using train test split**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.4, random_state = 42, stratify = Y)
```

**This is a classification problem, where given the features we need to classify whether the given features belong to class 0 or 1 of the target. Here 1 means patient has diabetes and 0 means the patient does not have diabetes.
Let's use K nearest Neighbor classification to classify for diabetes symptoms in a patient.**

# **<span style="color:#00008B;"> üî¥KNN from scratch</span>**

This code has been taken from Akashay Pachaar tweet üôå

```python
# KNN FROM SCRATCH

class KNN:
    def __init__(self, k = 3, task = 'classification'):
        self.k = k
        self.task = task
        
    def fit(self,X, y):
        self.X_train = X
        self.y_train = y
        
    def euclidean_distance(self,a,b):
        return np.sqrt(np.sum((a-b)**2, axis = 1))
    
    def predict(self,X):
        y_pred = [self.pred_one(x) for x in X]
        return np.array(y_pred)
    
    def pred_one(self,x):
        distances = self.euclidean_distance(x, X_train) # predict for single point
        
        k_indices = np.argsort(distances)[:self.k] # find k closet data point
        nn = [self.X_train[i].tolist() for i in k_indices] # get nearest neighbors
#         print("Nearest neighbors: ", nn)
        
        k_nearest_labels = [self.y_train[i].tolist() for i in k_indices]
#         print("Labels of nearest neighbors: ", k_nearest_labels)
        
        if self.task == 'classification':
            return np.argmax(np.bincount(k_nearest_labels))
        
        
        
```

```python
knn = KNN(k = 3, task = 'classification')
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

# print("Predicted labels: ", y_pred)
```

```python
# loss function
from sklearn.metrics import confusion_matrix, classification_report

confusion_matrix(y_test, y_pred)
```

```python
pd.crosstab(y_test, y_pred, rownames = ['True'], colnames = ['Predicted'], margins = True)
```

```python
print(classification_report(y_test, y_pred))
```

```python
# Lets try differnt k value

knn = KNN(k = 7, task = 'classification')
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```
