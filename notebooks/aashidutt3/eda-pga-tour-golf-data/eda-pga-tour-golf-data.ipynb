# **<span style="color:#75c9b7;">üì∞About the Dataset</span>**

Raw PGA tour data pulled from: https://www.advancedsportsanalytics.com/pga-raw-data

This data includes player level information and results for each tournament through 2022.

# **<span style="color:#75c9b7;">üìÅAbout the files</span>**

The file has total of 37 columns of which top 10 are described below.

> 1. Player_initial_last - player name short

> 2. tournament id - tourney id

> 3. player id - unique player id

> 4. hole_par - number of par 

(A typical golf course has 18 holes. So, if you add up all 18 par numbers, you could get around par 69 to 74 total. A regulation golf course has a par of 72 which is equivalent to four (4) par-3, ten (10) par-4 and four (4) par-5.)

> 5. strokes - player strokes

( total number of strokes is counted over one or more rounds of 18 holes)

> 6. hole_DKP - DKP hole

> 7. hole_FDP - FDP hole

> 8. hole_SDP - SDP for the hole

> 9. streak_DKP - streak DKP

> 10. streak_FDP - streak FDP

## **<span style="color:#BD4291;"> 1Ô∏è‚É£ Which player scored first position maximum times</span>**

```python
plt.figure(figsize=(15,8))
df.loc[df['pos'] == 1]['player'].value_counts().head(20).plot(kind = 'bar', color = my_colors[0])
plt.title('Tournament Wins of Players from 2015 to 2022')
plt.show()
```

**<span style="color:#9EB23B;">Dustin Johnson won title most of the time from year 2015 to 2022 followed by Justin Thomas and Bryson.</span>**

## **<span style="color:#BD4291;"> 2Ô∏è‚É£ Which player has max streak DKP, FDP and SDP.</span>**

```python
plt.figure(figsize=(15,8))
max_streak_DKP = df['streak_DKP'].max()
df.loc[df['streak_DKP'] == max_streak_DKP, 'player'].value_counts().plot(kind = 'barh', color = my_colors[1])
plt.title('Players with maximum score in DKP streak')
plt.show()
```

**<span style="color:#9EB23B;"> Only 3 players managed to maintain highest streak DKP</span>**

```python
plt.figure(figsize=(15,8))
max_streak_FDP = df['streak_FDP'].max()
df.loc[df['streak_FDP'] == max_streak_FDP, 'player'].value_counts().plot(kind = 'barh', color = my_colors[0])
plt.title('Players with maximum score in FDP streak')
plt.show()
```

**<span style="color:#9EB23B;"> Jon Rahm is the only player with highest streak FDP</span>**

```python
plt.figure(figsize=(15,8))
max_streak_SDP = df['streak_SDP'].max()
df.loc[df['streak_SDP'] == max_streak_SDP, 'player'].value_counts().plot(kind = 'barh', color = my_colors[1])
plt.title('Players with maximum score in SDP streak')
plt.show()
```

**<span style="color:#9EB23B;"> Scott Piercy is the only player with highest streak FDP</span>**

## **<span style="color:#BD4291;"> 3Ô∏è‚É£ Which player made maximum strokes</span>**

```python
max_strokes = df['strokes'].max()
df.loc[df['strokes'] == max_strokes, 'player'].value_counts().plot(kind = 'barh', color = my_colors[2])
plt.title('Player with maximum strokes')
plt.show()
```

```python
plt.figure(figsize=(15,8))
year_club = df.groupby(['strokes', 'player']).player.value_counts().head(10).plot(kind = 'bar',color = my_colors[4])
plt.title('Players and their number of strokes')
plt.legend()
plt.show()
```

## **<span style="color:#BD4291;"> 4Ô∏è‚É£ Strokes VS Hole Par per tournament</span>**

```python
plt.figure(figsize=(15,8))
df.plot(x='hole_par', y='strokes', style='|',alpha=0.6, markersize=10, color=my_colors[0], title='Strokes vs Hole Par')
plt.legend()
plt.show()
```

**<span style="color:#9EB23B;"> Greater the number of Strokes, greater is the number of hole par.</span>**

## **<span style="color:#BD4291;"> 5Ô∏è‚É£ Comparision between number of rounds and number of strokes</span>**

```python
plt.figure(figsize=(15,8))
df.groupby(['n_rounds', 'strokes']).strokes.count().plot(kind = 'bar',color = my_colors[1])
plt.title('Strokes vs number of rounds')
plt.legend()
plt.show()
```

**<span style="color:#9EB23B;">4th round mared maximum number of strokes throughout the tournaments from 2015 to 2022.</span>**

## **<span style="color:#BD4291;"> 6Ô∏è‚É£ Correlation among parameters</span>**

```python
plt.figure(figsize = (15, 8), facecolor = "#F7F4F4")
sns.heatmap(df.corr(), annot = True, cmap = "Greens");
```
