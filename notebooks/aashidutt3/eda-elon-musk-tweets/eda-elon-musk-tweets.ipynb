## **<span style="color:#1DA1F2;">1Ô∏è‚É£Which tweet has maximum / minimum number of likes?</span>**

```python
max_likes = df['Number of Likes'].max()
liked_tweet = df.loc[df['Number of Likes'] == max_likes, 'Tweets'].iloc[0]

print('"', liked_tweet , '" tweet got most likes i.e', max_likes, 'likes')
```

```python
min_likes = df['Number of Likes'].min()
least_liked_tweet = df.loc[df['Number of Likes'] == min_likes, 'Tweets'].iloc[0]

print('"', least_liked_tweet , '" tweet got least likes i.e', min_likes, 'likes')
```

```python
sorted_df = df.sort_values(by = "Number of Likes", ascending = False)
sorted_df = sorted_df.head(25)

fig = px.bar(sorted_df, x = "Tweets", y = "Number of Likes", color = "Number of Likes",
             color_continuous_scale = ["#E80F88", "#AF0171", "#790252", "#4C0033"])

fig.update_traces(marker = dict(line = dict(width = 2, color = "#F2F2F2")))
fig.update_layout(template = "plotly_dark", font = dict(family = "PT Sans", size = 16, color ="#EBC7E8"))
fig.update_xaxes(showticklabels = False) # Hide x axis ticks
fig.update_coloraxes(showscale=False)
fig.show()

## This snippet of code has been contributed by Matt OP
```

## **<span style="color:#1DA1F2;">2Ô∏è‚É£Which sources are most popular for tweeting?</span>**

```python
df['Source of Tweet'].value_counts()
```

```python
df['Source of Tweet'].value_counts().plot(kind="bar",figsize=(15,8),title="Source of Tweet", color = my_colors[4])
plt.show()
```

## **<span style="color:#1DA1F2;">3Ô∏è‚É£What are the least and most number of likes?</span>**

```python
sorted_df['Number of Likes'].head(25).plot(kind = 'bar', figsize = (15,8), title = "Most number of likes", color = my_colors[3])
plt.show()
```

```python
sorted_df = df.sort_values(by = "Number of Likes", ascending = True)
sorted_df['Number of Likes'].head(25).plot(kind = 'bar', figsize = (15,8), title = "Least number of likes", color = my_colors[4])
plt.show()
```
