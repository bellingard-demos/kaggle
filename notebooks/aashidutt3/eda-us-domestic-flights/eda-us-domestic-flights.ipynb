# **<span style="color:#F7B2B0;">📰About the Dataset</span>**

There are over 3.5 million monthly domestic flight records from 1990 to 2009. The dataset has following parameters to explore:

# **<span style="color:#F7B2B0;">📑Data</span>**

> Origin - String -  Three letter airport code of the origin airport

> Destination - String - Three letter airport code of the destination airport

> Origin City - String - Origin city name

> Destination City - String - Destination city name

> Passengers - Integer - Number of passengers transported from origin to destination

> Seats - Integer - Number of seats available on flights from origin to destination

> Flights - Integer - Number of flights between origin and destination (multiple records for one month, many with flights > 1)

> Distance - Integer - Distance (to nearest mile) flown between origin and destination

> Fly Date - Integer - The date (yyyymm) of flight

> Origin Population - Integer - Origin city's population as reported by US Census Destination Population

# **<span style="color:#F7B2B0;">🥽Let's see what all we can explore in this dataset</span>**

> 1. Which destination is the most visited place?

> 2. Which flight has maximum passengers and how does that relate to destination they are travelling to?

> 3. Which fly date has maximum booking?

> 4. Relation between Destination Population and Origin Population

> 5. Relation between Distance and Flights


```python
most_visited_destination = df['Destination'].value_counts()
most_visited_destination
```

```python
# Which destinations were most visited 

most_visited_destination.head(20).plot(kind = 'bar', figsize = (15,8), color = 'coral')
```

**ORD is the airport code for O'Hare International Airport, Illinois, USA. Thus, it was the most visited destination from 1990 to 2009.**

```python
#Which Destination city flight had maximum passengers

Max_passenger = df['Passengers'].max()
Destination = df.loc[df['Passengers'] == Max_passenger, 'Destination_City'].iloc[0]

print('Destination City of '+ Destination + ' had maximum passengers')
```

**Honoluluhad is a city in Hawaii. Since, Hawaii is the most popular destination for holidays therefore it had max passengers.**

```python
#Which Destination city flight had minimum passengers

Min_passenger = df['Passengers'].min()
Destination = df.loc[df['Passengers'] == Min_passenger, 'Destination_City'].iloc[0]

print('Destination City of '+ Destination + ' had least passengers')
```

**Bend is a small city on the Deschutes River, in Oregon and had least number of visitors.**

```python
#Which fly date had maximum booking

Max_fly_date = df['Fly_Date'].value_counts()
Max_fly_date
```

**From the above data we can say that maximum flights were booked in month of December 2007.**

```python
# Relation between Flights and Fly_Date

df.plot(x="Flights", y="Fly_Date", kind="hist", figsize=(15, 8), color = 'coral')
plt.show()
```

```python
#Which Destination State had max passengers
Max_Destination_State = df['Destination_State'].value_counts()
Max_Destination_State[0]
```

```python
Max_Destination_State.head(20).plot(kind = 'bar', figsize = (15,8), color = 'coral')
```

**The above graph shows that Texas was the most visted destination state.**

```python
# On which dates did passengers fly the most

Max_fly_date.head(20).plot(kind = 'bar', figsize = (15,8), color = 'coral')
```

```python
#Maximum distance traveled by a flight

Max_distance_covered = df['Distance'].max()
print('Max distance covered was ', Max_distance_covered)
```

```python
# Relation between Distance and Flights

df.plot(x="Flights", y="Distance", kind="hist", figsize=(15, 8), color = 'coral')
plt.show()
```

**The above histogram shows that the lower distances had more frequency of flights.**

```python
#Relation between Origin Population and Destination Population

df.plot(x="Origin_Population", y="Destination_Population", kind="hist", figsize=(15, 8), color = 'coral')
```
