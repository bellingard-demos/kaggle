## Using a regression model, split your data into train and test portions.

```python
from sklearn.model_selection import train_test_split
```

```python
X = x.reshape(-1,1)
x_train, x_test,  y_train, y_test = train_test_split(X,y, test_size=.3, train_size=.7)
```

## Fit your training split to the regression model

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
y_predicted = lr.predict(x_test)
```

## Show your regression model’s score.

```python
print("Regression Model’s Score:",lr.score(x_test,y_test) * 100)
```

## Draw at least three conclusions from your regression model.

### 1) Relationship: in my model i have positive linear relationship so the increase in that independent variable is associated with an increase in the dependent variable

```python
plt.scatter(x_test,y_test)
plt.plot(x_test, y_predicted)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("relation between height and weight of human body ")
plt.plot
```

### 2) outliers: I don't have any outliers in my model as the shown in the upper graph 

### 3) Strength of the Relationship: A higher R-squared value indicates a stronger Relationship between the dependent variable and independent variable.
### In my model the relationship is strong.
#### Note: the value of R-squared is between one and zero.

```python
import statsmodels.api as sm
X = sm.add_constant(X)
model = sm.OLS(y, X)
results = model.fit()
r_squared = results.rsquared
print("R-squared:", r_squared)
```
