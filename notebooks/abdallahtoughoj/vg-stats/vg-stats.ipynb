## 1) Which company is the most common video game publisher?


```python
most_common_publisher =  df['Publisher'].value_counts().idxmax()
most_common_publisher
```

## 2) What’s the most common platform?

```python
most_common_platform =  df['Platform'].value_counts().idxmax()
most_common_platform
```

## 3) What about the most common genre?

```python
most_common_genre =  df['Genre'].value_counts().idxmax()
most_common_genre
```

## 4) What are the top 20 highest grossing games?

```python
highest_grossing_games =  df.sort_values('Global_Sales', ascending=False).head(20)
highest_grossing_games
```

## 5) For North American video game sales, what’s the median?
* Provide a secondary output showing ten games surrounding the median sales output.
* Assume that games with same median value are sorted in descending order.

```python
north_american_sales_median =  df['NA_Sales'].median() 
north_american_sales_median
```

```python
above_median_games = df[df['NA_Sales'] > north_american_sales_median].sort_values(by='NA_Sales', ascending=False).head(10)
above_median_games
```

```python
below_median_games = df[df['NA_Sales'] < north_american_sales_median].sort_values(by='NA_Sales', ascending=False).head(10)
below_median_games
```

## 6) For the top-selling game of all time, how many standard deviations above/below the mean are its sales for North America?

```python
north_american_sales_mean =  df['NA_Sales'].mean() 
north_american_sales_std = df['NA_Sales'].std()
```

## 7) The Nintendo Wii seems to have outdone itself with games. How does its average number of sales compare with all of the other platforms?

```python
wii_average_sales = df[df['Platform'] == 'Wii']['Global_Sales'].mean()
wii_average_sales
```

```python
other_platforms_average_sales = df[df['Platform'] != 'Wii']['Global_Sales'].mean()
other_platforms_average_sales
```

```python
if wii_average_sales > other_platforms_average_sales:
        print("wii average sales is more than other platforms average sales" )
else:
    print("wii average sales is less than other platforms average sales" )
```
