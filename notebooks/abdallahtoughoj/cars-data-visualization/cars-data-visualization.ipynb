```python
import seaborn as sns
```

```python
cars = pd.read_csv('/kaggle/input/automobile-dataset/Automobile.csv')
cars
```

## This graph shows that when the car hours power is increased, the weight of the cars will increased 

```python
sns.regplot(x="horsepower",y="weight", data=cars)
```

## This graph shows that the most manufactured cars are four cylinders cars

```python
sns.violinplot(cars['cylinders'])
```

## conclusion:
### The analyzed graphs offer valuable insights into the automotive industry. The first graph demonstrates a positive correlation between car horsepower and weight, indicating that as the power of car engines increases, the weight of the cars also tends to increase. This suggests that higher power output is associated with heavier vehicles. The second graph reveals that four-cylinder cars dominate the manufacturing landscape, indicating that a majority of cars produced are equipped with four-cylinder engines. These findings shed light on both the relationship between power and weight in cars, as well as the prevalent use of four-cylinder engines in the industry. Together, they provide valuable information about design and production trends within the automotive sector.
