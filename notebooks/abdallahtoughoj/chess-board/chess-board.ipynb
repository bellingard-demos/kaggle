```python
class ChessBoard:
    def __init__(self):
        self.board = np.zeros([8,8,3])
        self.white = [1,1,1]
        for i in range(8):
            for j in range(8):
                if (j%2 != 0 and i%2 != 0) or (j%2 == 0 and i%2 == 0):
                    self.board[i][j] =self.white

        self.red_queen = None
        self.blue_queen = None
    def add_red(self):
        red_row = int(input("Enter the row number for red: "))
        red_column = int(input("Enter the column number for red: "))
        self.red_queen = [red_row,red_column]
        self.board[red_row][red_column] = [1, 0, 0]

    def add_blue(self):
        blue_row = int(input("Enter the row number for blue: "))
        blue_column = int(input("Enter the column number for blue: "))
        self.blue_queen = [blue_row,blue_column]
        self.board[blue_row][blue_column] = [0, 0, 1]

    def render(self):
        plt.imshow(self.board)

    def is_under_attack(self):
        # Diagonals check
        if (abs(self.red_queen[0] - self.blue_queen[0]) == abs(self.red_queen[1] - self.blue_queen[1])):
            print ("Diagonals")
            return True 
        # Horizontal check
        elif (self.red_queen[0]== self.blue_queen[0]):
            print ("Horizontal")
            return True 
        # Vertical check
        elif (self.red_queen[1]== self.blue_queen[1]):
            print ("Vertical")
            return True
        else:
            print("Not under attack")
            return False
```

## Horizontal (aka same row).

```python
chess_board = ChessBoard()
chess_board.add_red()
chess_board.add_blue()  
chess_board.render() 
chess_board.is_under_attack()
```

## Vertical (aka same column).

```python
chess_board = ChessBoard()
chess_board.add_red()
chess_board.add_blue()  
chess_board.render() 
chess_board.is_under_attack()
```

## Four Diagonals.

```python
chess_board = ChessBoard()
chess_board.add_red()
chess_board.add_blue()  
chess_board.render() 
chess_board.is_under_attack()
```

## A “Not Under Attack” scenario.

```python
chess_board = ChessBoard()
chess_board.add_red()
chess_board.add_blue()  
chess_board.render() 
chess_board.is_under_attack()
```
