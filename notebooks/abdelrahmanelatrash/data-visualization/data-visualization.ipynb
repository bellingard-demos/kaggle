```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### analysis data 

```python
data = pd.read_csv('/kaggle/input/employee-names-salaries-and-position-titles/Current_Employee_Names__Salaries__and_Position_Titles.csv')

data.head()
```

###  visualize the distribution of salaries using a histogram

```python
sns.histplot(data=data, x='Annual Salary')

plt.title('Distribution of Salaries')
plt.xlabel('Salary')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```

### Visualize the count of employees in each department.

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data, y='Department')
plt.title('Count of Employees by Department')
plt.xlabel('Count')
plt.ylabel('Department')
plt.show()
```

### Visualize the average annual salary by department

```python
avg_salary_by_dept = data.groupby('Department')['Annual Salary'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(data=avg_salary_by_dept, x='Annual Salary', y='Department')
plt.title('Average Annual Salary by Department')
plt.xlabel('Average Annual Salary')
plt.ylabel('Department')
plt.show()
```

### Visualize the relationship between hourly rate and typical hours for hourly employees.

```python
hourly_employees = data[data['Salary or Hourly'] == 'HOURLY']

plt.figure(figsize=(10, 6))
sns.scatterplot(data=hourly_employees, x='Hourly Rate', y='Typical Hours')
plt.title('Hourly Rate vs. Typical Hours')
plt.xlabel('Hourly Rate')
plt.ylabel('Typical Hours')
plt.show()
```

# the second dataset from url 


```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
data_2=pd.read_csv(url)
data_2.info()
```

```python
data_2.head()
```

### Visualize the count of passengers by their passenger class.

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data_2, x='Pclass')
plt.title('Passenger Count by Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()
```

### Visualize the proportion of passengers by their survival status.

```python
survival_count = data_2.Survived.value_counts()

plt.figure(figsize=(8, 8))
plt.pie(survival_count, labels=survival_count.index, autopct='%1.1f%%', startangle=90)
plt.title('Proportion of Passengers by Survival Status')
plt.show()
```

### Visualize the distribution of passenger ages.

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=data_2, x='Age', bins=20)
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```
