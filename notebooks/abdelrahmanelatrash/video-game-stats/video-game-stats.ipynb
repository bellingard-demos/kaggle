## Which company is the most common video game publisher?

```python
game_puplish=df.Publisher
game_puplish.value_counts().head(1)
```

## What’s the most common platform?

```python
platform_counts = df['Platform'].value_counts().head(1)
platform_counts
```

## What about the most common genre?

```python
most_common=df.Genre
most_common.value_counts().head(1)
```

## What are the top 20 highest grossing games?

```python
top_20=df.sort_values('Global_Sales', ascending=False).head(20)
top_20
```

## For North American video game sales, what’s the median?
>-Provide a secondary output showing ten games surrounding the median sales output.

>-Assume that games with same median value are sorted in descending order.

```python
na_s_median = df.NA_Sales.median()
print(" NA sale: " ,na_s_median)

ten_games = df[df.NA_Sales == na_s_median].head(10)
ten_games
```

## For the top-selling game of all time, how many standard deviations above/below the mean are its sales for North America?

```python
top_game = df.sort_values('Global_Sales', ascending=False).iloc[0]
na_mean = df['NA_Sales'].mean() # calculate the mean
na_std = df['NA_Sales'].std() # calculate standard deviation

standard_deviation_number = (top_game['NA_Sales'] - na_mean) / na_std

print("The standard_deviation_number : ", standard_deviation_number)
 
print("absolute value  : " ,abs(standard_deviation_number))
```

## The Nintendo Wii seems to have outdone itself with games. How does its average number of sales compare with all of the other platforms?

```python
nintendo_sales = df[df.Platform == 'Wii'].Global_Sales  # nintendo_sales only
nintendo_sales_mean = nintendo_sales.mean()
print(" average of global sales for Nintendo : ",nintendo_sales_mean)

other_sales = df[df['Platform'] != 'Wii']['Global_Sales']
other_sales_mean=other_sales.mean()
print("average of global sales for other :" ,other_sales_mean )
```
