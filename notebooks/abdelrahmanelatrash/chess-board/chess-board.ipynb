```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class ChessBoard:
    
    def __init__(self):
        """
        Initializes a new chessboard with alternating black and white cells.
        """
        
        self.board = np.zeros((8, 8, 3))
        
        for i in range(8):
            for j in range(8):
                if(i%2==1 and j%2==0) or (i%2==0 and j%2==1):
                        self.board[i][j] = (1,1,1)
                    
        self.red_positions = []
        self.blue_positions = []
        
    def render(self):
        """
        Displays the current state of the chessboard using Matplotlib.
        """
        plt.imshow(self.board)
        plt.axis('off')
        plt.show()
        
    def add_red(self, row, col):
        """
        Adds a red piece to the given row and column on the chessboard.

        Args:
            row (int): The row of the cell to place the red piece on.
            col (int): The column of the cell to place the red piece on.
        """
        self.board[row, col] = (1,0.2,0) # red
        self.red_positions.append((row, col))
    
    def add_blue(self, row, col):
        """
        Adds a blue piece to the given row and column on the chessboard.

        Args:
            row (int): The row of the cell to place the blue piece on.
            col (int): The column of the cell to place the blue piece on.
        """
        self.board[row, col] = (0,1,1) # blue
        self.blue_positions.append((row, col))
        
    def is_under_attack(self):
        """
        Returns a boolean indicating whether a red piece is under attack by a blue piece
        horizontally, vertically or diagonally on the board.

        Returns:
            bool: True if a red piece is under attack, False otherwise.
        """
        for red_pos in self.red_positions:
            for blue_pos in self.blue_positions:
                # Horizontal attack
                if red_pos[0] == blue_pos[0]:
                    return True
                # Vertical attack
                elif red_pos[1] == blue_pos[1]:
                    return True
                # Diagonal attacks
                elif abs(red_pos[0]-blue_pos[0]) == abs(red_pos[1]-blue_pos[1]):
                    return True
        return False
```

```python
board = ChessBoard()
board.render()
```

```python

board.add_blue(6, 6)
board.add_red(0, 0)
board.add_red(0, 6)
board.add_red(6, 0)
board.render()
board.is_under_attack()
```
