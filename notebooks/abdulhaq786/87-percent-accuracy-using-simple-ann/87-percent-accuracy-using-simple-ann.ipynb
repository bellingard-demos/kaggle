```python
import pandas as pd
train_data=pd.read_csv('/kaggle/input/simulation-generated-dataset-binary-class/generated_train.csv')
test_data=pd.read_csv('/kaggle/input/simulation-generated-dataset-binary-class/generated_test.csv')
print('Train Shape',train_data.shape)
test_data.shape
```

```python
train_data.columns=['feature1','feature2','target']
test_data.columns=['feature1','feature2','target']
```

```python
train_data.head()
```

```python
X_train=train_data.drop('target',axis=1)
y_train=train_data.target
X_test=test_data.drop('target',axis=1)
y_test=test_data.target
```

```python
import keras
from keras.layers import Dense,Flatten
from keras.models import Sequential
```

```python
model=Sequential()
model.add(Dense(128,activation='relu',input_shape=(None,2)))
model.add(Dense(64,activation='relu'))
model.add(Dense(28,activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(16,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10)
```
