```python
data.Class=data.Class.replace(['AGGRESSIVE', 'NORMAL', 'SLOW'],[1,2,3])
```

```python
x=data[['AccX', 'AccY', 'AccZ', 'GyroX', 'GyroY', 'GyroZ','Timestamp']]
y=data.Class
```

```python
x_train,x_test,y_trian,y_test=train_test_split(x,y,random_state=40,test_size=0.3)
```

```python
ModelD=DecisionTreeClassifier()
```

```python
ModelD.fit(x_train,y_trian)
y_predict=ModelD.predict(x_test)
```

```python
ModelD.score(x_test,y_test)
```

```python
confusion_matrix(y_test,y_predict)
```

```python
#finding optimal value of k
k=0
M=-1
for i in range (1,101):
    ModelK=KNeighborsClassifier(n_neighbors=i)
    ModelK.fit(x_train,y_trian)
    S=ModelK.score(x_test,y_test)
    if S>M:
        M=S
        k=i
print(k)
```

```python
ModelK=KNeighborsClassifier(n_neighbors=k)
```

```python
ModelK.fit(x_train,y_trian)
y_predict=ModelK.predict(x_test)
```

```python
ModelK.score(x_test,y_test) 
```

```python
accuracy_score(y_test,y_predict)
```

```python
confusion_matrix(y_test,y_predict)
```

```python
sns.distplot(y_test,hist=False)
sns.distplot(y_predict,hist=False)
```
