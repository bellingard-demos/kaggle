Cleaning The Dataset: In order to perform numerical calculations on the Dataset we have to change the datatypes of numerical columns from "object" to "int". 

```python
#printing columns names
data.columns
```

```python
data['Average price of 1GB (USD  at the start of 2021)']=data['Average price of 1GB (USD  at the start of 2021)'].str.replace('NO PACKAGES','0')
data['Average price of 1GB (USD – at start of 2020)']=data['Average price of 1GB (USD – at start of 2020)'].str.replace('NO PACKAGES','0')
```

```python
#changing Object data types of numerical columns to int
ls=  ['Average price of 1GB (USD)', 'Cheapest 1GB for 30 days (USD)','Most expensive 1GB (USD)','Average price of 1GB (USD  at the start of 2021)',
       'Average price of 1GB (USD – at start of 2020)']
for i in ls:
    data[i]=data[i].str.replace('$','').astype('float')
```

If we compare the histograms 'Average price of 1GB (USD – at start of 2020)' vs 'Average price of 1GB (USD  at the start of 2021)' vs present average price. We can see than most of countries internet prices are decreased from 2020 to 2022

```python
#we can cross validate our above conclusion
def divisionToGroups(t):
    t=t.tolist()
    for i in range(len(t)):
        if t[i]<=5:
            t[i]=1
        elif t[i]>5 and t[i]<=10:
            t[i]=2
        elif t[i]>10 and t[i]<=15:
            t[i]=3
        elif t[i]>15 and t[i]<=20:
            t[i]=4
        else:
            t[i]=5
    return pd.Series(t)
```

```python
data.columns
```

```python
d=data.copy()
d['Average price of 1GB (USD)']=divisionToGroups(data['Average price of 1GB (USD)'])
d['Average price of 1GB (USD  at the start of 2021)']=divisionToGroups(data['Average price of 1GB (USD  at the start of 2021)'])
d['Average price of 1GB (USD – at start of 2020)']=divisionToGroups(data['Average price of 1GB (USD – at start of 2020)'])
l1=d['Average price of 1GB (USD – at start of 2020)'].value_counts()
l1=l1.tolist()
l2=d['Average price of 1GB (USD  at the start of 2021)'].value_counts()
l2=l2.tolist()
l3=d['Average price of 1GB (USD)'].value_counts()
l3=l3.tolist()
```
