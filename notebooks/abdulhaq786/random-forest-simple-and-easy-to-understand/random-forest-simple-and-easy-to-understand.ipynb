```python
S=StandardScaler()
numericalData=S.fit_transform(numericalData)
```

```python
catagoricalData=pd.get_dummies(data.ocean_proximity)
```

```python
numericalData=pd.DataFrame(numericalData)
```

```python
data.columns
```

```python
numericalData.columns=['longitude', 'latitude', 'housing_median_age', 'total_rooms','total_bedrooms', 'population', 'households', 'median_income','median_house_value']
```

```python
data=numericalData.join(catagoricalData)
```

```python
data.isnull().sum()
```

```python
data.total_bedrooms=data.total_bedrooms.replace(np.NaN,data.total_bedrooms.mean())
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
x=data.drop(['median_house_value'],axis=1)
y=data.median_house_value
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=101,test_size=0.4)
```

```python
Model=RandomForestRegressor()
Model.fit(x_train,y_train)
Model.score(x_test,y_test)

```

```python
Prediction=Model.predict(x_test)
Prediction=pd.Series(Prediction)
```

```python

y_test.plot(kind='kde',label='Actual Values')
Prediction.plot(kind='kde',label='Predicted Values')
plt.legend()
```
